```python

import csv
import numpy as np
from scipy.interpolate import interp1d
import matplotlib


# Define the function that loads the files
def readfile(fname,N=1800,start=100):
    NewDT = 0.002;
    
    Araw = np.genfromtxt(fname, comments="#", delimiter=" ", dtype=float)        # read CSV
    Araw = Araw[start:start+N,:];                      # discard a few samples at the beginning, for the tap to start recording
    Time = np.cumsum(Araw[:,3]) * 1e-3;            # create a time axis, seconds
    Araw[:,3] = sqrt(sum(Araw[:,:3]**2,axis=1));   # find total accelleration, create a fourth column
    Araw = Araw-mean(Araw,axis=0);                 # remove DC in each column

    TimeNew = np.arange(Time[0],0.005*N,NewDT);    # new fine-time vector

    Fgroup=np.zeros([shape(TimeNew)[0]/2,shape(Araw)[1]]);
    for a in range(shape(Araw)[1]):
    
#        print 'in filename: ',fname,' Column ',a,'   Araw shape:',shape(Araw[:,a]),'   Time shape: ',shape(Time);
        F = interp1d(Time, Araw[:,a], kind='nearest');  # do the interpolation
        
    #    print 'in filename: ',fname,' Time shape: ',shape(Time),'   maximum is : ',Time[-1],' TimeNew: ',shape(TimeNew),'   maximum is : ',TimeNew[-1];
        if shape(TimeNew)[0] % 2 == 1:  TimeNew = TimeNew[0:-1];  # ensure the number of samples is even
        A = F(TimeNew);                                # do the interpolation
        Nnew = shape(TimeNew)[0];                      # new number of samples expected
        
        # do the FFT etc
        A = A * np.kaiser(Nnew,6) / mean(np.kaiser(Nnew,6));    # window
        F = abs(numpy.fft.fft(A))/Nnew;                         # FFT
        F=F[range(Nnew/2)]+r_[0,F[range(Nnew-1,Nnew/2,-1)]];    # add positive and negative frequency power together
        F[0:3]=0;                                               # erase the DC
        
        Fgroup[:,a] = F[:];

    freq = numpy.fft.fftfreq(Nnew,NewDT);
    freq=freq[range(Nnew/2)];
    
#    print 'in filename: ',fname,' A shape:',shape(A),'   Time shape: ',shape(TimeNew),'   Freq shape: ',shape(freq),'   F shape: ',shape(F);
    
    return freq,Fgroup;
```
