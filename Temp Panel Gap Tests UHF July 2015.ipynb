```python
# Create a time axis for both channels, with a small offset to allow synchronised plotting
T = D.timestamps - D.timestamps[0];


# Create a flipped frequency axis for plotting purposes
F=1712-D.channel_freqs/1e6;

# create a cropped frequency axis being only those allowed through the horn and the filters
freqrange=[500,870];  # one neat elevation cut
sanechannels = [i for i, v in enumerate(F) if ((v<=freqrange[1])&(v>=freqrange[0])) ];
FS = F[sanechannels];

# find the channels which would contain the integer multiples of 10 MHz
combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
# remove combchannels which are outside the chosen frequency range, not pretty but it works
temp = combchannels[:];
for a in temp:
    if (not(a in sanechannels)):
        combchannels.remove(a);
FC = F[combchannels];

# create a frequency axis which does not contain the comb channels
sanechannelsnocomb = sanechannels[:];
temp = sanechannelsnocomb[:];
for a in temp:
    if (a in combchannels):
        sanechannelsnocomb.remove(a);
FSNC = F[sanechannelsnocomb];


print 'Dump period = ',D.dump_period,'seconds'
print 'Start Time = V:',D.start_time,'Z'
print 'End Time = V:',D.end_time,'Z'
print 'Az range = V',np.min(D.az),' to ',np.max(D.az)
print 'El range = V',np.min(D.el),' to ',np.max(D.el)

print 'Found ', size(combchannels), ' comb channels'
print '  Channels ',combchannels[0],combchannels[1],combchannels[2], '...',combchannels[-1]
print '  Freqs ',FC[0],FC[1],FC[2], '...',FC[-1], ' MHz'
```

```python
# load the CSV of comb output powers and select only the comb channels we measured
temp = np.genfromtxt('panelgap/150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv', comments="#", delimiter=",", dtype=float)        # read CSV
combpower=np.zeros(shape(FC[:]));
for a in range(size(FC)):
    combpower[a]=temp[(  np.abs( temp[:,0] - np.round(FC[a]/10)*10 )  ).argmin(),1];
combpower = 10**((combpower-30)/10.0);  # convert to watts
figure(); plot(FC,combpower);
xlabel('Frequency, MHz');  ylabel('Comb line power, W');  title('Comb generator, 10 MHz setting, including 10 dB attenuator and cables');
```
