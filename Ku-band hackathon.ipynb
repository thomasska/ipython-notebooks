### Finding the absolute power - assuming noise floor is 20 Kelvin

Finds a smooth noise floor by taking elevation minimum of non-comb channels and smoothing.
Creates new autocorrelation matrices that are in W and not arbitrary units.

```python
# create a full-size visibility matrix with no comb channels (use neighbour to the left)
nocombh = squeeze(np.real(dsh.vis));
nocombv = squeeze(np.real(dsv.vis));
for i in combchannels:    # cloning of the neighbours
    nocombh[:,i,:] = 1.0*nocombh[:,i-1,:];
    nocombv[:,i,:] = 1.0*nocombv[:,i-1,:];

# find the noise floor using min over time - which mostly eliminates RFI etc
noisefloorh = squeeze(np.amin(  np.add( nocombh[:,:,0], nocombv[:,:,0] )/2,  axis=0 ));
noisefloorv = squeeze(np.amin(  np.add( nocombh[:,:,1], nocombv[:,:,1] )/2,  axis=0 ));

# filter this over frequency
N=300;  W = numpy.hamming(N);  WS = sum(W);
snfh = convolve(W,noisefloorh, 'same') / WS;
snfv = convolve(W,noisefloorv, 'same') / WS;

# how much power at 20 K in each channel bandwidth?
channelnoisepower = 1.38e-23 * 20 * dsh.channel_width;
print 'Noise power in a %i Hz channel at 20 K is %6.3g W' % (dsh.channel_width,channelnoisepower)

# multiply this by the smoothed noise power.  AbsPowerSignal = LevelSignal / Level20K * AbsPower20K
vish = (squeeze(np.real(dsh.vis[:,:,0])) + squeeze(np.real(dsh.vis[:,:,1]))) * channelnoisepower / snfh;
visv = (squeeze(np.real(dsv.vis[:,:,0])) + squeeze(np.real(dsv.vis[:,:,1]))) * channelnoisepower / snfh;

print 'Average power in a channel is %6.3g W, in a comb channel (1598) is %6.3g W' % (mean(mean(vish)),mean(vish[:,1598]));
print 'Created vish and visv with shape ',shape(vish)


# plot, for interest
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(14,5));
axes.plot(dsh.channel_freqs/1e6,squeeze(mean(real(dsh.vis[0:199,:,0]),axis=0)), label='HP, mean power',alpha=0.3,lw=1);
axes.plot(dsh.channel_freqs/1e6,noisefloorh, label='HP',alpha=0.7);
axes.plot(dsh.channel_freqs/1e6,snfh, label='smoothed',alpha=0.7,lw=3);
#axes.plot(noisefloor, label='Smoothed average, assume 20 K');
axes.set_xlim([870, 1650]);  axes.set_ylim([0, 220]);
axes.legend(loc=8);  axes.set_title('Noise floor power over frequency, showing derivation of the power figures');
```
