Plotting of some of the results
----------------

```python
# load the file
S=int(numpy.genfromtxt('SiteGTR_noise_m063_test1-V.npy.timestamp'))              # the timestamp
D=numpy.load('SiteGTR_noise_m063_test1-V.npy', mmap_mode='r')                    # the data
samplerate = 1712e6
start=samplerate-S%samplerate;
```

```python
# Plot all the samples, decimated to save the plotter
skip=2**15

figure(figsize=(12,5))
plot(np.linspace(0, D.shape[0]/samplerate*1e3, num=D.shape[0]/skip+1),D[::skip])
xlabel('Time after start, ms');    ylabel('Sample voltage');  grid(True);  title('Raw digitiser voltages over many seconds');  xlim([0,4000]);
```

```python
# Plot some samples, near the event
plotlength = 40e3            # samples

figure(figsize=(12,5))
plot(np.linspace(0, plotlength/samplerate*1e6, num=plotlength),D[(start):(start+plotlength)])
xlabel('Time after true 1PPS, us');    ylabel('Sample voltage');  grid(True);   xlim([0,20]);
```

```python
# average the power in a bunch of samples, plot it over time.  Plot this for each event.

averagesize = 128             # samples, ~= 100 ns
plotlength = 65536            # samples

figure(figsize=(12,6))

# make a time vector
T = np.linspace(0, plotlength/samplerate*1e6, num=plotlength/averagesize)       # time vector in microseconds

# make a total power vector
totalpower=T*0.0;

for pulsenum in range(4):                # plot each pulse seen
    Pow = (D[(pulsenum*samplerate+start):(pulsenum*samplerate+start+plotlength)]**2).reshape(-1,averagesize).mean(axis=1)
    totalpower = totalpower + Pow
    plot(T,10*log10(Pow))

plot(T,10*log10(totalpower/4),lw=2,color='k')

xlabel('Time after true 1PPS, us');    ylabel('Power, dB');  grid(True);   xlim([10,18]);
```

```python
# Spectrum of the file, off and on
figure(figsize=(12,6))

fftlen=128
fftnum=1024

freqaxis=linspace(856,1712,num=fftlen/2);

F=abs(fft.fft(D[start+1e8:start+1e8+fftlen*fftnum].reshape(fftnum,fftlen),axis=1))[:,:fftlen/2].mean(axis=0);   F[0]=F[1]
plot(10*log10(F),label='noise off')

F=abs(fft.fft(D[start-1e6-fftlen*fftnum:start-1e6].reshape(fftnum,fftlen),axis=1))[:,:fftlen/2].mean(axis=0);   F[0]=F[1]
plot(10*log10(F),label='noise on')

plot([0,4,NaN,9,29],[24,24,NaN,24,24],lw=3,color='r',label='chosen channels')

legend(loc=0);  ylabel('Power, arbitrary dB');  title('Noise on and off power spectra');   grid(True); xlabel('Channel');  #xlabel('Frequency, MHz'); xlim([856,1712]);

```

```python
# average the power in a bunch of samples, fft for power, select channels.   Plot this for each event.

averagesize = 128                            # samples, ~= 70 ns
channelselect = range(0,5)+range(9,30)       # list of fft channels to include in the analysis
plotlength = 65536                           # samples

figure(figsize=(12,6))

# make a time vector
T = np.linspace(0, plotlength/samplerate*1e6, num=plotlength/averagesize)       # time vector in microseconds

# make a total power vector
totalpower=T[:]*0.0;

for pulsenum in range(4):                # plot each pulse seen
    Pow = (abs(fft.fft((D[(pulsenum*samplerate+start):(pulsenum*samplerate+start+plotlength)]).reshape(-1,averagesize),axis=1))**2.0)[:,channelselect].mean(axis=1)/(averagesize**2.0)
    totalpower = totalpower + Pow/4.0
    plot(T,10*log10(Pow),label='transition %i' % pulsenum)

plot(T,10*log10(totalpower),lw=2,color='k',label='Average transition')

High=totalpower[:128].mean();  Low=totalpower[-128:].mean();  Thresh=(High*Low*Low)**0.33
FineOffset = T[list(totalpower>Thresh).index(False)]
plot(FineOffset,10*log10(Thresh),marker='*',color='c',markersize=20,label='detection')

title('Filtered for GTR noise band only. Note the clear double-step.  Transition time = %5.1f us' % FineOffset);
xlabel('Time after true 1PPS, us');    ylabel('Power, dB');  grid(True);   xlim([14,16]);  legend(loc=0);
```

```python
def rolling_window(a, window,axis=-1,pad=False,mode='reflect',**kargs):
    """
        This function produces a rolling window shaped data with the rolled data in the last col
        a      :  n-D array of data
        window : integer is the window size
        axis   : integer, axis to move the window over
        default is the last axis.
        pad    : {Boolean} Pad the array to the origanal size
        mode : {str, function} from the function numpy.pad
        One of the following string values or a user supplied function.
        'constant'      Pads with a constant value.
        'edge'          Pads with the edge values of array.
        'linear_ramp'   Pads with the linear ramp between end_value and the
        array edge value.
        'maximum'       Pads with the maximum value of all or part of the
        vector along each axis.
        'mean'          Pads with the mean value of all or part of the
        con  vector along each axis.
        'median'        Pads with the median value of all or part of the
        vector along each axis.
        'minimum'       Pads with the minimum value of all or part of the
        vector along each axis.
        'reflect'       Pads with the reflection of the vector mirrored on
        the first and last values of the vector along each
        axis.
        'symmetric'     Pads with the reflection of the vector mirrored
        along the edge of the array.
        'wrap'          Pads with the wrap of the vector along the axis.
        The first values are used to pad the end and the
        end values are used to pad the beginning.
        <function>      of the form padding_func(vector, iaxis_pad_width, iaxis, **kwargs)
        see numpy.pad notes
        **kargs are passed to the function numpy.pad

        Returns:
        an array with shape = np.array(a.shape+(window,))
        and the rolled data on the last axis

        Example:
        import numpy as np
        data = np.random.normal(loc=1,scale=np.sin(5*np.pi*np.arange(10000).astype(float)/10000.)+1.1, size=10000)
        stddata = rolling_window(data, 400).std(axis=-1)
        """
    if axis == -1 : axis = len(a.shape)-1
    if pad :
        pad_width = []
        for i in xrange(len(a.shape)):
            if i == axis:
                pad_width += [(window//2,window//2 -1 +np.mod(window,2))]
            else :
                pad_width += [(0,0)]
        a = np.pad(a,pad_width=pad_width,mode=mode,**kargs)
    a1 = np.swapaxes(a,axis,-1) # Move target axis to last axis in array
    shape = a1.shape[:-1] + (a1.shape[-1] - window + 1, window)
    strides = a1.strides + (a1.strides[-1],)
    return np.lib.stride_tricks.as_strided(a1, shape=shape, strides=strides).swapaxes(-2,axis) # Move original axis to
```

```python
# average the power in a bunch of samples, fft for power, select channels.   Plot this for each event.

averagesize = 128                            # samples, ~= 70 ns
channelselect = range(0,5)+range(9,30)       # list of fft channels to include in the analysis
plotlength = 65536                           # samples

figure(figsize=(12,6))

# make a time vector
T = np.linspace(0, plotlength/samplerate*1e6, num=plotlength)           # time vector, full length, values in microseconds

# make a total power vector
totalpower=T[:]*0.0;

for pulsenum in range(4):                # plot each pulse seen
    datanow=D[(pulsenum*samplerate+start):(pulsenum*samplerate+start+plotlength)]
    Pow=(np.abs(np.fft.fft(rolling_window(datanow, averagesize,pad=True),axis=1))**2)[:,channelselect].mean(axis=1)/(averagesize**2.0)
#    Pow = (abs(fft.fft((datanow).reshape(-1,averagesize),axis=1))**2.0)[:,channelselect].mean(axis=1)/(averagesize**2.0)
    totalpower = totalpower + Pow/4.0
    plot(T,10*log10(Pow))

plot(T,10*log10(totalpower),lw=2,color='k',label='Average transition')

High=totalpower[:64*averagesize].mean();  Low=totalpower[-64*averagesize:].mean();  Thresh=(High*Low*Low)**0.33
FineOffset = T[list(totalpower>Thresh).index(False)]
plot(FineOffset,10*log10(Thresh),marker='*',color='k',markersize=20,label='detection')
plot([0,99],[10*log10(Thresh),10*log10(Thresh)],lw=1,color='k',linestyle='-.',alpha=0.6)
title('Filtered for GTR noise band only. Note the clear double-step.  Transition time = %5.1f us' % FineOffset);
xlabel('Time after true 1PPS, us');    ylabel('Power, dB');  grid(True);   xlim([14,16]);  legend(loc=0);
```
