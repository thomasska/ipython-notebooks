UHF spectrum on site
==============
Measurements on AP3 in December 2015.

* UHF receiver, in the focus
* Fake-UHF digitiser, which has a reasonably flat response from 200-700 MHz, and some roll-off beyond this
* Comb generator on.
* Antenna moved in a complex pattern, elevation from 15 to 85, azimuth from 0 to 60.
* Comb channels have been removed from the graphs below.

```python
import katdal, numpy, matplotlib
fileName='/var/kat/archive/data/MeerKATAR1/telescope_products/2015/12/09/1449665735.h5'    # nice scan of the panel gaps, comb generator on

D=katdal.open(fileName);
D.select(corrprods=[0,1],dumps=range(2000))    # initial filter to reduce the amount of data to process

# find and exclude channels which contain the integer multiples of 10 MHz
F=1712-D.channel_freqs/1e6;                                                        # whole frequency axis
noncombchannels = [i for i, x in enumerate(abs((F*1e6+1.0*D.channel_width) % 10e6)-D.channel_width/2) if (x > 1.5*D.channel_width)];    # wide list of non-comb channels
D.select(channels=noncombchannels,freqrange=(1712e6-750e6,1712e6-400e6));         # select only non-comb-channels in the chosen frequency range

# current frequency axis
F = 1712 - D.channel_freqs[:]/1e6;    

# retrieve visibilities, flatten the time axis
Pwr = 10*log10(real(squeeze(np.mean(D.vis,axis=0))));

```

```python
# plot which channels are being plotted, to be sure there are no comb channels
figure(figsize=(8,2))
hist(F % 10.0,bins=np.arange(0,0.6,0.03));  xlabel('Frequency, MHz, modulo 10 MHz');  title('Occurrance of frequencies, modulo 10 MHz, in selected No-Comb frequency list')
figure(figsize=(8,2))
hist(F % 10.0,bins=np.arange(9.3,10.0,0.03));  xlabel('Frequency, MHz, modulo 10 MHz');
```

```python
# plot the UHF power seen on site
figure(figsize=(16,7))
plot(F,Pwr[:,0],lw=2,alpha=0.6,label='%s%s' % (D.corr_products[0][0][-1:],D.corr_products[0][1][-1:]));
plot(F,Pwr[:,1],lw=2,alpha=0.6,label='%s%s' % (D.corr_products[1][0][-1:],D.corr_products[1][1][-1:]));
legend(loc=0);  grid(True);
title('December 2015  UHF receiver power, measured by fake-UHF digitiser')
xlabel('Frequency, MHz');  ylabel('Correlator power, arbitrary dB');
annotate('UHF Horn is below cutoff here', xy=(535, 8),  xycoords='data', xytext=(-20, 30), textcoords='offset points', arrowprops=dict(arrowstyle="->"));
annotate('Carnarvon Analogue TV', xy=(620, 16),  xycoords='data', xytext=(-120, 20), textcoords='offset points', arrowprops=dict(arrowstyle="->"));
annotate('Internal to Receiver / Digitiser?\nNo variation over time at all.\nMiddle of SAPS bands, no TV', xy=(421, 8),  xycoords='data', xytext=(-48, 60), textcoords='offset points', arrowprops=dict(arrowstyle="->"));
```

```python
# close-up of 500-550 MHz
figure(figsize=(16,7))
plot(F,Pwr[:,0],lw=1,marker='.',ms=7,label='%s%s' % (D.corr_products[0][0][-1:],D.corr_products[0][1][-1:]));
plot(F,Pwr[:,1],lw=1,marker='.',ms=7,label='%s%s' % (D.corr_products[1][0][-1:],D.corr_products[1][1][-1:]));
legend(loc=0);  grid(True);  xlim([620,630])
title('December 2015  UHF receiver powermeasured by fake-UHF digitiser')
xlabel('Frequency, MHz');  ylabel('Correlator power, arbitrary dB');
```
