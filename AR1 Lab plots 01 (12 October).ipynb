1. Steady power test
===================

```python
import katdal, numpy as np, matplotlib
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/13/1444729458.h5');     # steady power. Signal and noise. (Replace with 120 second one when it's ready)
D.select()
D.select(compscans='~transition')
```

```python
# autocorrelations
t=10;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[t,:,0]))),label=D.corr_products[0])
plot(10*log10(abs(squeeze(D.vis[t,:,1]))),label=D.corr_products[1])
title('Autocorrelation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); xlim([0,4095]);  legend(loc=0);

ch=1360;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[:,ch,0]))),label=D.corr_products[0])
plot(10*log10(abs(squeeze(D.vis[:,ch,1]))),label=D.corr_products[1])
title('Autocorrelation baselines of %s over time at channel=%i'% (D.experiment_id,ch)); xlabel('Dump');  ylabel('Correlator power, dBarbitrary');  legend(loc=0);
```

```python
# cross correlations
t=10;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[t,:,2]))),label=D.corr_products[2])
title('Cross-correlation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); xlim([0,4095]);  legend(loc=0);
figure(figsize=(12,2))
plot(angle(squeeze(D.vis[t,:,2])),label=D.corr_products[2])
title('Cross-correlation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); xlim([0,4095]);  legend(loc=0);

ch=1360;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[:,ch,2]))),label=D.corr_products[2])
title('Cross-correlation baselines of %s over time at channel=%i'% (D.experiment_id,ch)); xlabel('Dump');  ylabel('Correlator power, dBarbitrary');  legend(loc=0);
```

```python
# statistics in a single channel
ch=[2650,3650];
t=10;

P = abs(squeeze(D.vis[:,range(ch[0],ch[1]),0]));

figure(figsize=(12,2))
plot(range(ch[0],ch[1]),P.T,lw=2,alpha=0.3)
title('Autocorrelation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, linear'); # xlim(ch); ylim([70,90]);

figure(figsize=(12,4))
P = P/mean(P,axis=0)
P=reshape(P,-1)
print(shape(P))

hist(P,100);
title('Histogram of normalised powers over the above channels. Total of %i points' % shape(P))
```

```python
# phase slope in the noise band
ch=[2600,3700];
ph=squeeze(angle(D.vis[3:-3,ch[0]:ch[1],3])).T;
ph=unwrap(ph.T).T

figure(figsize=(12,3));
plot(range(ch[0],ch[1]),ph,color='m');
xlim(ch);  ylabel('Unwrapped Phase');  xlabel('Channel');  title('AR1 Lab cross correlation unwrapped phase, noise band, all times overlaid');

ph=mean(ph,axis=1);  # average before finding slope
[m,c] = np.polyfit(range(ch[0],ch[1]),ph,1)
delay=1/(.856 / (m/(2*pi)*4096));
delaysamples=(1/(.856 / (m/(2*pi)*4096)))/(1/1.7120);

print('Phase slope in noise band of %s  %s is as follows: ' %  (D.experiment_id,D.description));
print('  %f radians/channel  = %6.2f ns  = %6.2f samples' % (-m,-delay,-delaysamples))
```

2. Digitiser Phase Slope comparisons before/after firmware update
===================


```python
import katdal, numpy as np, matplotlib
ch=[2900,3700];
figure(figsize=(10,4));
print('Henno:  Switch from 12 to 12a happened at 2015-10-14 08:00:00 UTC')
print('Thomas: Cables were not equal length until 2015-10-01 10:30:40 UTC or so')

filenames=['/var/kat/archive/data/MeerKATAR1/telescope_products/2015/09/30/1443625623.h5',
           '/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/01/1443695415.h5',
           '/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/13/1444729458.h5','/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/13/1444729458.h5','/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/14/1444836653.h5','/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/15/1444892523.h5','/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/15/1444894863.h5',
           '/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/15/1444898108.h5', '/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/15/1444900989.h5','/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/15/1444901478.h5']
for fn in filenames:
    D=katdal.open(fn);
    D.select()
    D.select(compscans='~transition')
    ph=squeeze(angle(D.vis[3:-3,ch[0]:ch[1],3])).T;
    ph=unwrap(ph.T).T
    ph=mean(ph,axis=1);  # average before finding slope
    plot(range(ch[0],ch[1]),ph,label=D.start_time);
    [m,c] = np.polyfit(range(ch[0],ch[1]),ph,1)
    delay=1/(.856 / (m/(2*pi)*4096));
    delaysamples=(1/(.856 / (m/(2*pi)*4096)))/(1/1.7120);
    delaysamples=abs(delaysamples)
    print('abs(Delay)= %4.1f samples  (%s UTC  %s)' % (delaysamples, D.start_time, D.description))

legend(loc=0)
```

3. Correlator F-engine non-linearity
================

```python
import katdal, numpy as np, matplotlib
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/14/1444838000.h5');     # fine power ramp

#plot correlator power vs signal generator power

D.select()
powerbase=squeeze(abs(D.vis[0,1362,0]))  # get a base power number from a side channel in the first dump

D.select(compscans='~transition')

Freq=np.zeros(D.shape[0]);                # initialise Frequency and Power vectors
Power=np.zeros(D.shape[0]);
ch1360=np.zeros(D.shape[0]);
ch1370=np.zeros(D.shape[0]);
ch1371=np.zeros(D.shape[0])
count=0;

for cs in D.compscans():
    num=D.shape[0];
    freq=float(cs[1].split(',')[1].split('=')[1])/1.0e6;
    pwr=float(cs[1].split(',')[2].split('=')[1]);
    Freq[count] = freq;
    Power[count] = pwr;
    ch1360[count] = squeeze(abs(D.vis[0,1360,0]))
    ch1370[count] = squeeze(abs(D.vis[0,1371,0]))
    ch1371[count] = squeeze(abs(D.vis[0,1372,0]))
    count=count+1;
    #if count>57: break;
    #print('shape %s  %s  Frequency %f MHz   Power %s dBm  startdump=%i'  % (D.shape, shape(D.vis), freq, pwr, D.dumps[0] ) );


figure(figsize=(12,6))
plot(Power,label='Signal Generator')
plot(10*log10(ch1360),label='Correlator power')
legend(loc=0)

figure(figsize=(12,6))
plot(Power,10*log10(ch1360),marker='.',label='channel 1360');
grid(True); legend(loc=0);
xlabel('Signal generator power, dBm');  ylabel('Total correlator power, dB CU');  title('Meas %s -  %s  Correlator Linearity plot' % (D.name[-13:-3],D.start_time));


figure(figsize=(12,6))
plot(Power,10*log10(ch1360),marker='.',label='channel 1360');
plot(Power,10*log10(ch1370),marker='.',label='channel 1370');
plot(Power,10*log10(ch1371),marker='.',label='channel 1371');
grid(True); legend(loc=0);
xlabel('Signal generator power, dBm');  ylabel('Total correlator power, dB CU');  title('Meas %s -  %s  Correlator Linearity plot' % (D.name[-13:-3],D.start_time));


D.select()

print(1712e6-D.channel_freqs[1360])
print(Freq[0]*1e6)
print(Power)
```
