Check of the absolute power scale
---------------------
The cell below tests the absolute power scale. A test was done with the LPDA directly facing the feed horn, 7 m away. This gives a direct measurement of the power received due to the source, without considering the details of the source. I attenuate this for the increased distance between subreflector and LPDA in the actual test, and compare it to the calculated free space path loss. This should confirm the absolute power scale.

In the first graph, the two polarisations are clear. The total power is constant in each measurement, even as I moved the antenna up and down, you can see the variation in the individual components

```python
#Cross-check of my calculations - the direct-horn-radiation scheme
#F,direct=loadMeas('05/27/1432731158.h5',atten=36,Tsys=300);                               # 7 m from feed horn, 36 dB attenuation, H and then V. Assume Tsys is now fairly warm
F,direct=loadMeas('07/29/1438163010.h5',atten=50,Tsys=300);                               # 5 m from feed horn, 36 dB attenuation, H and then V. Assume Tsys is now fairly warm
lossV=10.0*log10(mean(sum(direct[140:205,:,:],axis=2),axis=0));                           # time-selection for the different polarisations of the sources
lossH=10.0*log10(mean(sum(direct[225:325,:,:],axis=2),axis=0));                           # time-selection for the different polarisations of the sources
directLoss=10.0*log10(  (10.0**(lossV/10.0)+10.0**(lossH/10.0))/2.0  );
W = np.ones(2);  W = W / sum(W);
directLoss = convolve(W,directLoss, 'same');  directLoss[0]=directLoss[1];  directLoss[-2]=directLoss[-1];

# expected free space loss (excluding lpda etc)
freeSpaceHorn = -20.0*log10(7)-20.0*log10(F)+27.55 + GainHorn;                                # 7 m from horn

# reflected waves
deltapathlength=sqrt(3**2+1.7**2)+sqrt(4**2+3.5**2)-sqrt(7**2+(3.5-1.7)**2);              # 7 m away, source 1.7m high, horn 3.5 m high
freeSpaceReflected=freeSpaceHorn + 20*log10(abs(1+exp(deltapathlength/(300/F)*2*pi*1j)));     # for a good reflector...

figure(figsize=(12,5));
plot(10.0*log10(direct[:,:,0]),lw=1,alpha=0.15,color='red');  plot(10.0*log10(direct[:,:,1]),lw=1,alpha=0.15,color='green');
plot(10.0*log10(mean(direct[:,:,0],axis=1)),lw=3,color='red',label='hh');  plot(10.0*log10(mean(direct[:,:,1],axis=1)),lw=3,color='green',label='vv');
#xlim([100,400]); ylim([-50,20]); legend(loc=0);   xlabel('time, dumps'); ylabel('path, dB');
title('hand-held direct path measurements, 7 m from horn');

figure(figsize=(12,5));
plot(F,lossV,label='Source VP',lw=2);  plot(F,lossH,label='Source HP',lw=2);   plot(F,directLoss,label='mean',lw=1,color='black',ls=':');
plot(F,freeSpace,color='black',label='free space');  plot(F,freeSpaceReflected,color='black',linestyle='--',label='fs + reflection');
grid(True);  ylim([-45,-25]);  legend(loc=0);  xlabel('Frequency, MHz'); ylabel('path, dB');
```

The second graph shows the path loss calculated from the known radiated power of the LPDA/comb source and the system noise figure, and compares it to the expected free-space path loss.

There is quite a large difference, particuarly at middle frequencies?!

Comparing the absolute leakage levels
-------------------------
of the two methods

```python
# compare  abs path loss by erp,  to abs path loss by horn-to-lpda
addLossDistance=20.0*log10(Dsubreflpda/7.0);                           # a positive number for Loss
freeSpace = -20.0*log10(Dsubreflpda)-20.0*log10(F)+27.55 + Gsubref;    # expected free space path gain
AP1direct=AP1-(directLoss-addLossDistance)-Gsubref+GainHorn;
AP2direct=AP2-(directLoss-addLossDistance)-Gsubref+GainHorn;
figure(figsize=(12,5));
plot(F,AP1-freeSpace,color='blue',lw=2,label='AP1, compared to expected free space');
plot(F,AP2-freeSpace,color='green',lw=2,label='AP2,  (Gsubref, distance, freq)');
plot(F,AP1direct,color='blue',lw=2,linestyle='--',label='AP1, by comparison to direct wave');
plot(F,AP2direct,color='green',lw=2,linestyle='--',label='AP2,  (using the 7 m direct wave test)');
ylim([-50,-10]);  legend(loc=0); grid(True); ylabel('dB transmission');
title('Leakage through dish, by comparison to measured and expected free-space path loss, dB');
grid(True);
```

It is also good to see in the second graph above, that the two methods of calibrating the absolute leakage come to the same result. This is a comparison of

* Path loss calculated from known transmitter power (Comb power, attenuator, LPDA gain) and known received power (System noise temperature, adjacent channel noise power),
* Path loss calculated from a free-space measurement from LPDA to Receiver, adjusted for the change in distance between the free test and the full LPDA test.

These two methods are quite independent, and give the same result, with some ripple from various unmodelled effects.

Attempt to calculate absolute leakage
----------------
By assuming that the gap covers on AP1 work perfectly, I can calculate the reflector efficiency of AP2.  
$Transmission_{AP2} = \frac{TotalLeakage_{AP2}-TotalLeakage_{AP1}}{PathLoss_{FreeSpace}}$  

```python
freeSpace = -20.0*log10(Dsubreflpda)-20.0*log10(F)+27.55 + Gsubref;
AP2minus1 = 10*log10(10.0**(AP2/10.0)-10.0**(AP1/10.0));
AP2minus1direct = 10*log10(10.0**((AP2direct+freeSpace)/10.0)-10.0**((AP1direct+freeSpace)/10.0));

# subtract the power and see how it looks:
figure(figsize=(12,4));
plot(F,AP1,lw=2,label='AP1',ls='-',color='blue');
plot(F,AP2,lw=2,label='AP2',ls='-',color='green');
plot(F,AP2minus1,lw=2,label='AP2-AP1, i.e. subtracting diffraction etc',color='black');

plot(F,AP1direct+freeSpace,lw=2,ls='--',color='blue');
plot(F,AP2direct+freeSpace,lw=2,ls='--',color='green');
plot(F,AP2minus1direct,lw=2,ls='--',color='black');
plot(F,freeSpace,lw=1,ls='--',color='black',label='free space');
legend(loc=9);  ylabel('Path loss, dB');  ylim([-90,-30])
title('Path Losses, dB');

# plot the percentage transmission
figure(figsize=(12,4));
plot(F,100.0*10.0**((AP2minus1-freeSpace)/10.0),lw=2,label='AP2-AP1 compared to free space',color='black');
plot(F,100.0*10.0**((AP1-freeSpace)/10.0),lw=1,label='AP1 assuming it\'s all leakage, compared to free space',color='blue');
plot(F,100.0*10.0**((AP2-freeSpace)/10.0),lw=1,label='AP2 assuming it\'s all leakage',color='green');
plot(F,100.0*10.0**((AP2minus1direct-freeSpace)/10.0),lw=2,color='black',ls='--');
plot(F,100.0*10.0**((AP1direct)/10.0),lw=1,color='blue',ls='--');
plot(F,100.0*10.0**((AP2direct)/10.0),lw=1,color='green',ls='--');
legend(loc=0);   xlabel('Frequency, MHz');   ylabel('% transmission');
title('Transmission through the dish, %');  ylim([0,1.4]);

# plot the percentage transmission, final nice summary graph
figure(figsize=(12,5));
plot(F,100.0*10.0**((AP2minus1-freeSpace)/10.0),lw=2,label='AP2-AP1 compared to free space',color='black');
plot(F,100.0*10.0**((AP2minus1direct-freeSpace)/10.0),lw=2,label='AP2-AP1 compared to direct measurement',color='black',ls='--');
legend(loc=0);   xlabel('Frequency, MHz');   ylabel('% transmission');
title('Transmission through the dish, %');  ylim([0,1.3]);
plot([500,850],[1.0,1.0],lw=1,ls=':')
```

Simple code check of all the above
----------------------
This section tries to produce the same result as the dotted lines above (comparison of direct and through-dish), without all of the Tsys and Comb generator details getting worked in.

```python
# sanity check - just subtract AP1, AP2, and the direct measurement, no comb generator or source LPDA

F,tempAP1v=loadRaw('05/13/1431517785.h5',atten=10,azLimit=[-1,5],elLimit=[17,80]);                     # AP1, 5m test
F,tempAP1h=loadRaw('05/13/1431514867.h5',atten=10,azLimit=[-1,5],elLimit=[17,80]);                     # AP1, 5m test
rawAP1 = (np.mean(np.sum(tempAP1v,axis=2),axis=0) + np.mean(np.sum(tempAP1h,axis=2),axis=0)) / 2.0;     # average of the two polarisations
rawAP1 = convolve(np.ones(2)/2.0,rawAP1, 'same');  rawAP1[0]=rawAP1[1];  rawAP1[-2]=rawAP1[-1];         # size 2 smoothing. Not a good idea, the comb spikes are still in here

F,tempAP2v=loadRaw('05/27/1432732570.h5',atten=16,azLimit=[-1,5],elLimit=[17,80]);                     # AP2, 5m test
F,tempAP2h=loadRaw('05/27/1432733258.h5',atten=16,azLimit=[-1,5],elLimit=[17,80]);
rawAP2 = (np.mean(np.sum(tempAP2v,axis=2),axis=0) + np.mean(np.sum(tempAP2h,axis=2),axis=0)) / 2.0;     # average of the two polarisations
rawAP2 = convolve(np.ones(2)/2.0,rawAP2, 'same');  rawAP2[0]=rawAP2[1];  rawAP2[-2]=rawAP2[-1];         # size 2 smoothing. Not a good idea, the comb spikes are still in here

F,tempDirect=loadRaw('05/27/1432731158.h5',atten=36);                                                  # 7 m from feed horn, 36 dB attenuation, H and then V. Assume Tsys is now fairly warm
rawDirect=(mean(sum(tempDirect[140:205,:,:],axis=2),axis=0) + mean(sum(tempDirect[225:325,:,:],axis=2),axis=0) )/2.0;  # time-selection for the different polarisations of the sources
rawDirect = convolve(np.ones(2)/2.0,rawDirect, 'same');  rawDirect[0]=rawDirect[1];  rawDirect[-2]=rawDirect[-1];      # size 2 smoothing. Not a good idea, the comb spikes are still in here
rawDirect = rawDirect / (Dsubreflpda/7.0)**2.0;                                                         # reduce it so we are at the same distance. Focus-Dish-LPDA = 20 m.  Direct Horn-LPDA = 7 m. 
#rawDirect = rawDirect / 10.0**((GainHorn-Gsubref)/10.0);                                                # compensate for the higher horn gain, compared to subreflector. They're very similar in the region of interest

figure(figsize=(12,4));
plot(F,10*log10(rawAP1),F,10*log10(rawAP2),F,10*log10(rawDirect))
grid(True);
#plot(F,100.0*ratioDelta,color='k');    #,
#ylim([0,2]);

# find the power ratios
ratioAP1 = rawAP1 / rawDirect;            # compare the direct path with the shielded path
ratioAP2 = rawAP2 / rawDirect;            # In this division, Power_Transmit is removed

ratioDelta=ratioAP2-ratioAP1;             # subtract, which assumes that all the power received on AP1 was diffraction, not leakage

figure(figsize=(12,5));
plot(F,100.0*ratioAP1,label='AP1, bent gap covers')
plot(F,100.0*ratioAP2,label='AP2, no gap covers')
plot(F,100.0*ratioDelta,lw=2,color='k',label='AP2, assuming AP1 is all diffraction');
legend(loc=0); plot([500,850],[1.0,1.0],lw=1,ls=':',color='black'); ylim([0,2.0]);  xlim([550,800])
xlabel('Frequency, MHz');   ylabel('% transmission');
title('Measured average transmission through the dish, %'); 
```

```python
##############################################################################################################################3
# define some functions
import katdal, csv, numpy as np, matplotlib

# some global constants
Gsubref = 10*log10(42000.0/(50.0*50.0));   # estimate for the subreflector gain
Dsubreflpda=17;                            # distance in metres from subreflector (or prime focus?) to LPDA.
GainHorn = np.interp(F,[499,550,600,700,800,900],[2.0,9.8,12.2,12.5,12.2,10.5]);                           # horn gain,  from EA-MKU-1000-DREP-01_2   Figure 13 page 22, extrapolated

##############################################################################################################################3
def queryFile(fileName):
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);
    T = D.timestamps - D.timestamps[0];
    fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12,5));
    axes[0,0].plot(D.az,D.el,color='black',lw=2);
    axes[0,0].set_title('Az vs El');
    axes[0,1].plot(T,((D.az+0.2) % 360)-0.2,label='Az',color='blue',lw=4);
    axes[0,1].plot(T,D.el,label='El',color='red',lw=4);
    axes[0,1].set_title('Az and El');  axes[0,1].legend(loc=0);
    D.select(corrprods=[0,1]);
    axes[1,0].plot(D.channel_freqs/1e6,10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=0))));   axes[1,0].set_title('average spectrum');
    axes[1,1].plot(10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=1))));   axes[1,1].set_title('power over time');
    print('Contents of %s - %i dumps' % (D.name,shape(D.vis)[0]))
    #print(D)
    

    
##############################################################################################################################3
def loadTwo(fileName1,fileName2,atten=0,azLimit=[-190,190],elLimit=[1,99]):
    combFreq,temp1=loadMeas(fileName1,atten=atten,azLimit=azLimit,elLimit=elLimit);
    temp1 = np.mean(np.sum(temp1,axis=2),axis=0);            #  sum of the polarisations
    combFreq,temp2=loadMeas(fileName2,atten=atten,azLimit=azLimit,elLimit=elLimit);
    temp2 = np.mean(np.sum(temp2,axis=2),axis=0);           # linear-average over all the scans.
    combPower = (temp1 + temp2)*0.5;                        # average the two source polarisations
    
    combPower = 10*log10(combPower);
    return combFreq,combPower



##############################################################################################################################3
# load one file, return the comb frequencies and path loss
def loadMeas(fileName,atten=-10,Tsys=150,azLimit=[-190,190],elLimit=[-1,90]):
    Glpda = 5;  # dBi
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);
    #print(D)
    
    D.select(corrprods=[0,1]);

    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # pull out the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    noisePower = np.min(noisePower,axis=0);                                                      # find the minimum along the time axis, this is the true noise power, removes some spikes etc.
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=combchannels);             # the comb power
    combPower=real(D.vis[:]);
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis

    combPower = combPower - noisePower;
    
    # find the noise floor, and use it to convert the comb channels' power into watts
    systemNoise= Tsys * 1.3806488e-23 * D.channel_width;      # scalar - noise floor from kTB, watts
    combPower = combPower / noisePower * systemNoise;         # combPower is now in Watts
    
    # restrict the azimuth and elevation range
    limitsBool=np.logical_and(   np.logical_and(np.greater(D.az,azLimit[0]),np.less(D.az,azLimit[1])),    np.logical_and(np.greater(D.el,elLimit[0]),np.less(D.el,elLimit[1]))   );
    limitsBool=reshape(limitsBool,shape(limitsBool)[0]);             # in desparation
    combPower = combPower[limitsBool,:,:];
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);
    
    # compensate for the signal generator and the antenna gain
    radiatedPower= sigGenPower(combFreq) * 10.0**((Glpda-atten)/10.0);
    radiatedPower = np.transpose([radiatedPower,radiatedPower]);
    combPower = combPower / radiatedPower;
    
    print('Loaded %s, atten=%4.1fdB, returned %i dumps' % (D.name,atten,shape(combPower)[0]));
    
    return combFreq,combPower;





##############################################################################################################################3
# load one file, return the comb frequencies and additional noise  power in correlator power units
def loadRaw(fileName,atten=-10,azLimit=[-190,190],elLimit=[-1,90]):
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);

    D.select(corrprods=[0,1]);

    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # extract the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-499e6),channels=combchannels);             # the comb power
    combPower=real(D.vis[:]);
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis

    combPower = combPower - noisePower;                                                          # find the additional power on the comb channels
    
    # restrict the azimuth and elevation range
    limitsBool=np.logical_and(   np.logical_and(np.greater(D.az,azLimit[0]),np.less(D.az,azLimit[1])),    np.logical_and(np.greater(D.el,elLimit[0]),np.less(D.el,elLimit[1]))   );
    limitsBool=reshape(limitsBool,shape(limitsBool)[0]);             # in desparation
    combPower = combPower[limitsBool,:,:];
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);
    
    # Apply the attenuator... so Increase the power received by the attenuator value
    combPower = combPower * 10.0**(atten/10.0);
    
    print('RawLoaded %s, atten=%4.1fdB, returned %i dumps' % (D.name,atten,shape(combPower)[0]));
    
    return combFreq,combPower;
    

    

##############################################################################################################################3
# given the frequencies in MHz, returns the power of the relevant comb line, in watts, at the end of the cable, without any attenuators
def sigGenPower(FC,filename='150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv'):
    temp = np.genfromtxt('panelgap/'+filename, comments="#", delimiter=",", dtype=float)        # read CSV
    sgpower=np.zeros(shape(FC[:]));
    for a in range(size(FC)):
        sgpower[a]=temp[(  np.abs( temp[:,0] - np.round(FC[a]/10)*10 )  ).argmin(),1];
    sgpower = 10**((sgpower-30+10)/10.0);         # convert to watts, removing the 10 dB attenuator!
    return sgpower
```
