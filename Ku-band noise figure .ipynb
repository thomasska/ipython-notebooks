```python
#%reset
import math

import time
import katdal
import csv
import numpy as np
import matplotlib
import matplotlib.gridspec as gridspec
pi=math.pi
#constants
c=299792458  #speed of light
k=1.380622e-23 #Boltzmann's constant
sfu=1e-22      #1 SFU (solar flux unit) = 1e-22 W/m^2/Hz

#DSS 15 antenna worked example 
#input
S=246*sfu           #solar flux measurement for the observation
nabla=0.64          #antenna efficiency
D=34                #antenna diameter [m]
freq=8420e6         #measurement frequency
lbda=c/freq
#calculation
A=pi*(D/2)**2
print 'Area:', A, 'Wavelength:', lbda
#Correction  for sources larger than dish main beam
#solid angle  dish
t_3db=1.22*lbda/D/pi*180     #check theta3dB=1.22*lambda/D
print 'Check beamwidth:', t_3db
theta_3dB=0.061              #beamwitdh in degrees
Kp=1.03                      #correction factor pattern shape for uniform illumination as given in paper
sigma_dish=Kp*theta_3dB**2   #solid angle of main beam
Kbeam=1.14                   #correction due to solardisk energy into total beam of DSS-15
sigma_dishc=Kbeam*sigma_dish #1.14 correction for sun energy in sidelobes
#solid angle sun
sunDisk=0.533              #diameter of the solar disk in degrees
sigma_sun=pi/4*sunDisk**2  #solid angle of the sun
deltaTc=nabla*S*A/(2*k)*(sigma_dishc/sigma_sun)
print 'Calculated noise temperature of sun:', deltaTc, deltaTc*0.99

#M063 antenna  
#input
S=416.7*sfu           #solar flux measurement for the observation
nabla=0.75            #antenna efficiency
D=13.5                #antenna diameter [m]
freq=12.5e9           #measurement frequency
lbda=c/freq
#calculation
A=pi*(D/2)**2
print 'Area:', A, 'Wavelength:', lbda
#Correction  for sources larger than dish main beam
#solid angle  dish
t_3db=1.22*lbda/D/pi*180     #check theta3dB=1.22*lambda/D
t_3db=lbda/D/pi*180          #approx theta3dB=lambda/D
print 'Check beamwidth:', t_3db
#theta_3dB=0.1                #beamwitdh in degrees
theta_3dB=t_3db              #beamwitdh in degrees
Kp=1.3                      #correction factor pattern shape for gaussian illumination 
sigma_dish=Kp*pi/4*theta_3dB**2   #solid angle of main beam
#solid angle sun
sunDisk=0.533              #diameter of the solar disk in degrees
sigma_sun=pi/4*sunDisk**2  #solid angle of the sun
deltaTc=nabla*S*A/(2*k)*(sigma_dish/sigma_sun)

print 'Calculated noise temperature of sun:', deltaTc
L=1+0.38*(sunDisk/theta_3dB)**2
T=A*nabla*S/2/k

print(T)
```

### Spectral plots, showing H and V polarisation is mostly maintained through/around the dish

```python
import matplotlib.pyplot as plt
from pylab import *

# plot spectra of all the data, just to see
fig, axes = plt.subplots(nrows=3, ncols=1,figsize=(14,10));

Yhs=mean(real(ds_sunr.vis[330:340,:,0]),axis=0)/mean(real(ds_csp.vis[50:220,:,0]),axis=0)
Yvs=mean(real(ds_sunr.vis[330:340,:,1]),axis=0)/mean(real(ds_csp.vis[50:220,:,0]),axis=0)

Tsrc=deltaTc
Tsrc=11532

#L-band example from SEFD.xlsx spreadsheet
#Tsys = Tcmb (2.73) + Tgal (4.1) + Tatm (2.2)  + Tspill (6.8) + Tspill_noshroud (1-3) +  Treflector(ohmic and panelgaps 1) + Trx (90???)
# Ku-band
Tcmb=2.73
Tgal=0
Tatmh=4.56  #4.56  from RA calculator for centre F and ~ degree elevation for this observation #sun at 55 deg el, CSP at 30degree ish 7.22
Tatmc=7.22
Tspill=2 # from M0000-000V1-11 TM at Ku
Trefl=1.5

Thot= (Tsrc+Tcmb+Tatmh+Tspill+Trefl) # per polarisation 
Tcold=(     Tcmb+Tatmc+Tspill+Trefl) #

Tesh=(Thot-Yhs*Tcold)/(Yhs-1) #h; 
Tesv=(Thot-Yvs*Tcold)/(Yvs-1) #v
print 'Thot:', Thot, 'Tcold:', Tcold

axes[0].plot(ds_sun.channel_freqs/1e6,10*log10(mean(real(ds_sun.vis[:,:,0]),axis=0)),lw=2,alpha=1.0,color='r',label='sun h');
axes[0].plot(ds_sun.channel_freqs/1e6,10*log10(mean(real(ds_sun.vis[:,:,1]),axis=0)),lw=2,alpha=1.0,color='m',label='sun v');
axes[0].plot(ds_sun.channel_freqs/1e6,10*log10(mean(real(ds_csp.vis[:,:,0]),axis=0)),lw=3,alpha=0.6,color='g',label='csp h');
axes[0].plot(ds_sun.channel_freqs/1e6,10*log10(mean(real(ds_csp.vis[:,:,1]),axis=0)),lw=3,alpha=0.6,color='c',label='csp v');
axes[0].set_xlabel(ds_sun.description + '  Channel number');
axes[0].set_ylabel('Mean power over samples');
axes[0].legend(loc=1);
axes[0].set_title('Comparison of V and H source polarisations, 0-200 seconds only');axes[0].set_xlim([860,1650]); #axes[0].set_ylim([0, 4000]);
axes[0].grid();

axes[1].plot(ds_csp.channel_freqs/1e6,Yhs,lw=2,alpha=1.0,color='r',label='Yh');
axes[1].plot(ds_csp.channel_freqs/1e6,Yvs,lw=2,alpha=1.0,color='g',label='Yvv');
axes[1].set_xlabel(ds_csp.description + '  Channel number');
axes[1].set_ylabel('Y-factor');
axes[1].set_xlim([860,1650]); #axes[1].set_ylim([0, 50]);
axes[1].grid();

axes[2].plot(ds_sun.channel_freqs/1e6+11200,Tesh,lw=2,alpha=0.7,color='r',label='H');
axes[2].plot(ds_sun.channel_freqs/1e6+11200,Tesv,lw=2,alpha=0.7,color='g',label='V');
axes[2].legend(loc=1);
axes[2].set_xlabel('IF Frequency [MHz] ');
axes[2].set_ylabel('Rough Ku-RX noise figure [K]');
axes[2].set_xlim([860+11200,1650+11200]); 
axes[2].set_ylim([8, 180]);
axes[2].grid();
fig.tight_layout();
```
