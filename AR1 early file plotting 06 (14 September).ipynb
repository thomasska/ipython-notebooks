```python
import katdal, numpy as np, matplotlib, random
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/09/15/1442309778.h5');
```

```python
# autocorrelations
t=100;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[t,:,0]))),label='hh')
plot(10*log10(abs(squeeze(D.vis[t,:,1]))),label='vv')
title('Autocorrelation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); xlim([0,4095]);  legend(loc=0);

ch=1000;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[:,ch,0]))),label='hh')
plot(10*log10(abs(squeeze(D.vis[:,ch,1]))),label='vv')
title('Autocorrelation baselines of %s over time at channel=%i'% (D.experiment_id,ch)); xlabel('Dump');  ylabel('Correlator power, dBarbitrary');  legend(loc=0);

figure(figsize=(25,4));
imshow(squeeze(10*log10(abs(D.vis[:,:,0]))),aspect='auto',interpolation='bilinear', cmap=plt.get_cmap('CMRmap'),vmin=2,vmax=21,extent=[0,4095,(shape(D.timestamps)[0]),0]);
xlim([800,2700]);  xlabel('channel'); title('auto HH, amplitude');  ylabel('Dump');
```

Saturation effects
------------------
Plotting the average power in some even and odd channels, over time

```python
#Ooh look, saturation effects
sgp=range(-28,44,2);
sgp.extend([-99,-99,-99,-99,-99,-99,-99]);
ch=[1600,1800];
figure(figsize=(12,3))
plot(10*log10(abs(squeeze(mean(D.vis[1::2,range(ch[0],ch[1],2),0],axis=1)))),label='hh, even channels')
plot(10*log10(abs(squeeze(mean(D.vis[1::2,range(ch[0]+1,ch[1]+1,2),0],axis=1)))),label='hh, odd channels')
xlim([13,40]); ylim([2.43,2.54]); grid(True)
title('Autocorrelation baselines of %s over time, mean of channels=%i-%i'% (D.experiment_id,ch[0],ch[1])); xlabel('Dump');  ylabel('Correlator power, Units');  legend(loc=0);

ch=1835;
figure(figsize=(12,3))
plot(10*log10(abs(squeeze(D.vis[1::2,ch,0]))-abs(squeeze(D.vis[18,ch,0]))),label='hh, signal generator channel');
plot(sgp,label='(expected signal generator power +54 dB)',linestyle='--');
xlim([13,40]); ylim([-3,45]);  grid(True)
title('Autocorrelation baselines of %s over time at channel=%i'% (D.experiment_id,ch)); xlabel('Dump');  ylabel('Correlator power, dBarbitrary');  legend(loc=0);
```

A wide colour plot of the output, showing the broad spread of raised power as the correlator starts to saturate

```python
# Wide colour plot
figure(figsize=(12,4));
imshow(squeeze(10*log10(abs(D.vis[:,:,0]))),aspect='auto',interpolation='bilinear', cmap=plt.get_cmap('CMRmap'),vmin=2,vmax=4,extent=[0,4095,(shape(D.timestamps)[0]),0]);
xlim([1000,2000]);  ylim([80,40]); xlabel('channel'); title('auto HH, amplitude');  ylabel('Dump');
```

Some close-ups of the power over time, as the correlator saturates

```python
# detail colour plot showing ripple etc
figure(figsize=(12,4));
imshow(squeeze(10*log10(abs(D.vis[:,:,0]))),aspect='auto',interpolation='none', cmap=plt.get_cmap('CMRmap'),vmin=2.3,vmax=3.1,extent=[0,4095,(shape(D.timestamps)[0]),0]);
xlim([1700,1840]);  ylim([80,40]); xlabel('channel'); title('auto HH, amplitude');  ylabel('Dump');

figure(figsize=(12,3))
plot(10*log10(abs(squeeze(D.vis[32,:,0]))),label='hh, t=32',lw=2,alpha=0.9)
plot(10*log10(abs(squeeze(D.vis[67,:,0]))),label='hh, t=67',lw=1.5,alpha=0.8)
plot(10*log10(abs(squeeze(D.vis[72,:,0]))),label='hh, t=72',lw=2,alpha=0.7)
xlim([1700,1840]);  ylim([2.3,2.6]);
title('Autocorrelation baselines of %s over channel at t=72' % (D.experiment_id)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); legend(loc=0);
```

Find the additional power, as a function of frequency

```python
figure(figsize=(12,3))
plot(range(0,4096,2),(abs(squeeze(D.vis[72,0::2,0]))-abs(squeeze(D.vis[50,0::2,0]))),label='hh, even channels')
plot(range(1,4096,2),(abs(squeeze(D.vis[72,1::2,0]))-abs(squeeze(D.vis[50,1::2,0]))),label='hh, odd channels')
xlim([0,2300]); ylim([-0.03,0.25]); grid(True);  legend(loc=0);
title('Difference between Autocorrelation baselines of %s at t=72 and t=50'% (D.experiment_id)); xlabel('Dump');  ylabel('Correlator power increase, Units');  
```

**Conclusions**

1. There are duplicate channels coming out of the correlator.
1. There is clearly some power leakage from the active channel, into nearby channels
1. Power leaks into odd channels only


**Further tests to be done**

1. Try the correlator saturation test again

1. Dwell longer at each power level, at least 10 dumps
1. Slightly smaller steps might be useful
