```python
import numpy;
import matplotlib
from matplotlib import patches
import matplotlib.pyplot as plt

def plotConcentration(stimulus, hl=[900, 4800, 36000, 86400], dt = 30., tMax = 86400):
    # all units are in seconds
    
    t=6.75*3600;
    
    
    # duplicate the stimulus by a few days
    #stimulus = np.append(stimulus,np.array([stimulus[:,0]+24*3600,stimulus[:,1]]).T,axis=0);
    
    T = np.empty(0);  V=np.empty((0,4))             # records, for plotting
    
    Vcurrent = np.array([0.0,0.0,0.0,1.0]);         # initial state vector
    
    
    # run a simulation
    
    while t<=tMax:
        # stimulus - add 1 unit at t=5
        
        for a in stimulus:
            if (a[0]>=(t%86400.0)) and (a[0] < (t%86400.0)+dt):
                Vcurrent[0] = Vcurrent[0] + a[1];
        
        # delta calculations
        D=zeros(shape(Vcurrent)[0]);
        
        diff01 = Vcurrent[0] / hl[0] * dt * log(2);                     # this one uses a half-life, no dependence on the target value
        diff12 = Vcurrent[1] / hl[1] * dt * log(2);                     # this one uses a half-life, no dependence on the target value
        diff13 = (Vcurrent[1]-Vcurrent[3]) / hl[2] * dt * log(2);       # going both ways, depending on amount in each
    
    
        # apply the changes to the state vector
        Vcurrent = Vcurrent + [-diff01,  diff01-diff12-diff13,   diff12,   +diff13];
        
        for a in range(shape(Vcurrent)[0]):                 # prevent any of them going below zero
            if Vcurrent[a] < 0:
                Vcurrent[a] = 0;
        
        # store for plotting
        T = np.append(T,[t]);  V = np.append(V,[Vcurrent],axis=0)
    
        # time
        t = t + dt;
        
        #print('T=%5.1f, V1 = %5.3f   d1 = %4.4f     V2 = %5.3f  d2 = %4.4f' % (t,V1current,diffusion1,V2current,diffusion2))
    
    # limit levels
    limits=[1.0,1.7];
    
    
    # plot
    fig, ax = plt.subplots(figsize=(16,5))

    ax.plot(T/3600.,V[:,0],label='V0',lw=1,alpha=0.5)
    ax.plot(T/3600.,V[:,1],label='V1',lw=2)
    ax.plot(T/3600.,V[:,3],label='V3',lw=1)
    
    # coloured regions
    ax.add_patch(patches.Rectangle([6,0],tMax/3600,limits[0],facecolor='yellow',edgecolor='none',alpha=0.1))
    ax.add_patch(patches.Rectangle([6,limits[0]],tMax/3600,limits[1]-limits[0],facecolor='green',edgecolor='none',alpha=0.1))
    ax.add_patch(patches.Rectangle([6,limits[1]],tMax/3600,10,facecolor='red',edgecolor='none',alpha=0.1))

    ax.legend(loc=0); ax.grid(True); ax.set_xlabel('Time, hours');  ax.set_ylim([0,2]);
    ax.set_title('HL1 = %6.2f s   HL2 = %6.2f s   dt = %4.1f s   %i simulation points' % (hl[0],hl[1],dt,shape(T)[0]) );
    
    #print('final values = %f %f %f %f  sum= %f  expected = %f' % (Vcurrent[0],Vcurrent[1],Vcurrent[2],Vcurrent[3],np.sum(Vcurrent),np.sum(stimulus[:,1])/2.)) 

    # percentage of time in range
    inrange=100*np.float(((limits[0] < V[:,1]) & (V[:,1] < limits[1])).sum())/shape(V)[0]
    over=100*np.float((limits[1] < V[:,1]).sum())/shape(V)[0]
    under=100*np.float((V[:,1] < limits[0]).sum())/shape(V)[0]
    print('in range 3-6 for %4.1f %% of the time.  Over=%4.1f %%.  Under=%4.1f %%' % (inrange,over,under))
```
