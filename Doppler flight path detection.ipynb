Investigating a Doppler flight path detection system
========================

```python
# receiver is at the origin
receiver = [0,0,0]
#define some variables
transmitters=np.array([[27,0,0],[0,15,0],[-10,-15,0],[-26,4,0]])*1000.     # transmitter x,y,z   in m

start=[[-14000,-3000,100]]       # metres x y z
velocity=np.array([[70,30,0]])             # m/s x y x
timestep = 5                     # seconds
timeend = 500                    # seconds

# start calculating vectors

# time
t = np.reshape(np.arange(0,timeend,timestep),[-1,1])
# plane position
plane = start + velocity*t
# ranges to plane
totrange = np.empty([shape(plane)[0],shape(transmitters)[0]])
doppler = np.empty([shape(plane)[0],shape(transmitters)[0]])
signal = np.empty([shape(plane)[0],shape(transmitters)[0]])
for n,tx in enumerate(transmitters):
    totrange[:,n]=np.sum( (tx-plane)**2 + (plane-receiver)**2 ,axis=1)**0.5
    signal[:,n]=-10*log10(np.sum( (tx-plane)**2 ,axis=1) * np.sum( (plane-receiver)**2 ,axis=1))
    tempdelta = np.sum( (tx-plane+velocity*0.01)**2 + (plane-receiver+velocity*0.01)**2 ,axis=1)**0.5
    doppler[:,n] = (tempdelta-totrange[:,n])*100.;

# plot transmitters and flight path
figure(figsize=(10,5))
plot(t,doppler)
title('Doppler per transmitter'); xlabel('time'); ylabel('m/s');

figure(figsize=(10,5))
for tx in transmitters:
    plot(tx[0],tx[1],'o',lw=0,ms=15)
plot(receiver[0],receiver[1],'k+',ms=15)
plot(plane[:,0],plane[:,1],color='k',lw=1)
plot(plane[0,0],plane[0,1],'d',color='k',ms=10)
grid(True); title('Plan view of flight path'); axis('equal')

figure(figsize=(10,5))
plot(t,totrange)
title('Total range per transmitter'); xlabel('time'); ylabel('m');

figure(figsize=(10,5))
plot(t,signal)
title('Signal per transmitter'); xlabel('time'); ylabel('m');
```
