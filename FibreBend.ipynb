```python
# Define the function that plots things
def readphase(fname,fibres=2,reverse=0,N=2000):
    test = np.genfromtxt('fibre/' + fname + '.csv', comments="#", delimiter="\t", dtype=float, skip_header=3)[:,1]        # read CSV
    scale = 1.0 / 1500e6 / 360 * 1e12;            # Phase scale - to convert degrees @ 1500 MHz into ps
    test = 1.0 * test * scale / fibres;           # number of fibres
    
    if N > 1:
        test = convolve(ones(N),test,'valid')/N;  # use boxcar averaging (of size N) as a lowpass filter
    
    test = test -mean(test[1:1000]);              # normalise to first 1000 items
    return test;
```

### Data smoothing

The graphs from the network analyser are noisy, and can be smoothed to show the slow variation as the angle varies.


```python
# Compare three ways of smoothing
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(10,7));
axes.plot(readphase('9001',fibres=10,N=1)[2000:],lw=1,alpha=0.25,color='black',label='Original data, unsmoothed');
axes.plot(readphase('9001',fibres=10,N=200)[1800:],lw=2,alpha=0.6,color='red',label='filtered with a 200 point moving average');
axes.plot(readphase('9001',fibres=10,N=2000)-0.03,lw=2,alpha=1.0,color='blue',label='filtered with a 2000 point moving average');
axes.set_xlabel('Time, arbitrary - about 100 s');
axes.set_ylabel('Change in delay, picoseconds per fibre');
axes.legend(loc=3);
axes.set_title('Smoothing of network analyser data');
```

**Discussion:** The noise on the network analyser data does not contribute to the results.

We are interested in variations on the timescale of several seconds, the movement of the antenna takes about 1 minute.

Smoothing at 2000 points will be applied.

### Azimuth movements, 180-0 and 0-180

```python
# Compare the 180-0 movements
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(10,7));
axes.plot(readphase('9001',fibres=10),lw=2,alpha=1.0,color='blue',label='9001 - 180 to 0');
axes.plot(readphase('9003',fibres=10),lw=2,alpha=1.0,color='blue',label='9003 - 180 to 0');
axes.plot(readphase('9005',fibres=10),lw=2,alpha=1.0,color='blue',label='9005 - 180 to 0');
axes.plot(readphase('9007',fibres=2),lw=2,alpha=1.0,color='lightblue',label='9007 - 180 to 0, one loop only');
axes.set_xlabel('Time, arbitrary - about 100 s');
axes.set_ylabel('Change in delay, picoseconds per fibre');
axes.legend(loc=3);
axes.set_title('Fibre bending by rotating antenna. Elev @18.6, Azimuth moving');
#xes.set_xlim([0, 4095]); axes[0].set_ylim([0, 500]);
```

There is some common movement near the end of the graph here, a drop of 0.25 ps or so.

It is also encouraging to note that the single fibre matches the 10-fibre measurements quite well.

```python
# Compare the 0-180 movements
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(10,7));
axes.plot(readphase('9002',10),lw=2,alpha=1.0,color='black',label='9002 - 0 to 180, suspect');
axes.plot(readphase('9004',10),lw=2,alpha=1.0,color='green',label='9004 - 0 to 180');
axes.plot(readphase('9006',10),lw=2,alpha=1.0,color='green',label='9006 - 0 to 180');
axes.set_xlabel('Time, arbitrary - about 100 s');
axes.set_ylabel('Change in delay, picoseconds per fibre');
axes.set_title('Fibre bending by rotating antenna. Elev @18.6, Azimuth moving');
axes.legend(loc=3);
```

**Discussion:** It's very difficult to draw any conclusion here.
Very likely the azimuth wrap is slipping at slightly arbitrary times, when unwrapped, giving rise to a strange signal.

**Comment:** Strange that there is no consistency in the phase change over angle change.

## Conclusions

Data quality is good.

It is likely that the fibre bending is not completely tied to the movement of the AP.

Other effects like temperature drift over time, seem to dominate the results here. It is difficult to conclude that there is a clear signal from the bending.

The dela caused by any bending is bounded, certainly less than 0.2 ps per fibre. This is excellent, we can conclude there is no risk from the flexing of the fibre.

**Further tests:** We need to test for 20-minute variation caused by temperature changes, over a few days.
