UHF Horn Cover effectiveness
==============================

Covers have been made for the L-band and UHF receivers, to protect the LNAs when there is possible RFI work on the antenna.

A measurement was performed on 9 December 2015, to evaluate the effectiveness of the covers. Due to time constraints, only the UHF cover could be tested.
