```python
# funtion to return the number of samples by which the digitiser streams are offset

def dmcip(receptor):
    return {
        24: '10.103.254.2:7224',
        25: '10.103.254.2:7225',
        31: '10.103.254.2:7231',
        34: '10.103.254.2:7234',
        62: '10.103.254.2:7262',
        63: '10.103.254.2:7263',
        101: '192.168.4.36:7201'        
    }[receptor]
# lab DMC is 192.168.4.36.  AR1 DMC is 10.103.254.2,  RTS one is .253.2


def digitiserCommand(commandString,receptor=63,main=False):
    subIP=dmcip(receptor)
    mainIP=subIP[:-3] + '147'
    
    import commands
    if main: ip=mainIP
    else:  ip=subIP
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 100 -s " + ip + " ?" + commandString);
    if result[0]>0:
        print('Error performing digitiser command: %s' % commandString)
        print(result[1])
    return result


def getSamples(receptor):
    result=digitiserCommand("adc-snap-shot vh | grep '#adc-snap-shot'",receptor=receptor)
    if result[0]!=0: print('Failure retrieving samples');
    result=result[1]
    snapshot=result.replace('#adc-snap-shot v','').replace('#adc-snap-shot h','').replace('\n','');
    samples=np.array(  map(int, snapshot.split(' ')[1:] )  ).reshape(2,-1)
    return samples


def sigGen(onOff):
    # Connect and set up.
    import socket, time, datetime, array, numpy
    sg = socket.socket(socket.AF_INET, socket.SOCK_STREAM);  sg.setblocking(0); sg.settimeout(1.0)
    sg.connect(('192.168.4.33', 10001));  # Lantronix TCP-RS232 converter     In the lab: 192.168.4.33
    time.sleep(0.5)
    if onOff:
        sg.send('OUTP ON\n');
        print(' Signal generator On, wait 1 second');
    else:
        sg.send('OUTP OFF\n');
        print(' Signal generator Off, wait 1 second');
    # Disconnect
    time.sleep(1)
    sg.close();
    
    
    
def doGlobalSync(receptor=101):
    result=digitiserCommand("global-synchronise",receptor=receptor,main=True);
    if result[0]!=0: print('Error performing global sync');

        
# the repaired plotSpectrum with correct FFT stacking, no mixing up of channels!
def plotSpectrum(numSnapshots=4,fftLength=1024,receptor=63):
    print('Spectrum from receptor %i.  %i snapshots, fftLength=%i.     Getting snapshots...' % (receptor,numSnapshots,fftLength))
    allSamples=np.zeros([2,0,fftLength]);
    for a in range(numSnapshots):
        #print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples(receptor);
        allSamples=np.concatenate(  ( allSamples, samples.reshape(2,-1,fftLength) ),  axis=1  )
    
    print('Samples max=%i, min=%i, mean=%f, std=%f' % (np.max(allSamples),np.min(allSamples),np.mean(allSamples),np.std(allSamples)))
    
    Fsamples=fft.fft(  allSamples,  fftLength,  axis=2) / fftLength
    freq=np.linspace(856,1712,num=fftLength/2);
    
    Corr = Fsamples[0,:,:] * conj(Fsamples[1,:,:]);
    Corr = np.mean(Corr,axis=0)[:fftLength/2];
    
    Auto = Fsamples[:,:,:] * conj(Fsamples[:,:,:]);
    Auto = np.mean(Auto,axis=1)[:,:fftLength/2].T;
    
    flatSamples=np.concatenate((squeeze(allSamples[0,:,:]).reshape(1,-1),squeeze(allSamples[1,:,:]).reshape(1,-1)),0);
    
    figure(figsize=(16,6))
    plot(freq,10*log10(abs(Auto)),alpha=0.7); title('Auto and Cross - Magnitude dB');
    plot(freq,10*log10(abs(Corr)),alpha=0.7);
    xlim([856,1712]);  ylim([-40,10]); grid(True);

    #meanPwrCross=10*log10(  np.mean( abs( Corr[int((1300.0-856)/1712.0*fftLength):int((1500.0-856)/1712.0*fftLength)] ) )  );
    #meanPwrAuto=10*log10(    np.mean(   np.mean( abs( Auto[int((1300.0-856)/1712.0*fftLength):int((1500.0-856)/1712.0*fftLength),:] ),axis=1 ),axis=0   )    );
    
    #plot([freq[int((1300.0-856.0)/1712.0*fftLength)],freq[int((1500.0-856.0)/1712.0*fftLength)]],[meanPwrCross,meanPwrCross],'r-',lw=2,alpha=0.5) 
    #plot([freq[int((1300.0-856.0)/1712.0*fftLength)],freq[int((1500.0-856.0)/1712.0*fftLength)]],[meanPwrAuto,meanPwrAuto],'k-',lw=2,alpha=0.5) 
    #print('Average power 1300-1440 MHz, cross-correlation:  %f dB,  autocorrelation  %f dB' % (meanPwrCross,meanPwrAuto) );
    
    figure(figsize=(16,2))
    plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,700]);
    figure(figsize=(16,2))
    hist(flatSamples.T,np.arange(-100,100,2)); title('voltage histogram');

        
        
        
    
def getDigitiserOffset(numSnapshots=4,globalSync=False,fftLength=256,verbose=0,receptor=63):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync(receptor=receptor)
        time.sleep(2)
        
    selectedRange=[0.63,0.9]        # channels to us to find sample offset
    selectedRange = np.round(np.multiply(selectedRange,fftLength/2))

    allSamples=zeros([2,0])
    for a in range(numSnapshots):
        if verbose>0: print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples(receptor=receptor);
        allSamples=np.concatenate((allSamples,samples),1)
    
    Fsamples=fft.fft(  allSamples.reshape(2,-1,fftLength),  fftLength,  2)
    
    Corr = Fsamples[0,:,0:fftLength] * conj(Fsamples[1,:,0:fftLength]);
    Corr=squeeze(np.sum(Corr,axis=0));
    CorrAngle=unwrap(angle(Corr))
    Pwr=squeeze(np.sum(abs(Fsamples**2),axis=1)).T;

    offsetSamples = (CorrAngle[selectedRange[1]]-CorrAngle[selectedRange[0]]) / (2.*pi*(selectedRange[1]-selectedRange[0])/fftLength)
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(allSamples.T,alpha=0.6); title('Voltages over time'); xlim([0,1000]);
        figure(figsize=(16,3))
        hist(allSamples.T,np.arange(-40,40)); title('voltage histogram');
        
    if verbose>1:
        figure(figsize=(16,5))
        plot(10*log10(Pwr)); title('Auto and Cross - Magnitude dB'); xlim([0,fftLength/2]);
        plot(10*log10(abs(Corr)));
        figure(figsize=(16,3))
        plot(unwrap(angle(Corr[0:fftLength/2]))); title('Cross - Unwrapped Angle'); xlim([0,fftLength/2]);
        plot(selectedRange,[0,0],'b-',lw=3)
        
    if verbose>0:
        print('Samples offset:  %5.2f' % offsetSamples)
        print('Voltage mean=%5.2f stdev=%5.2f max=%3i min=%3i' % (np.mean(allSamples),np.std(allSamples),np.max(allSamples),np.min(allSamples) ))
    
    return offsetSamples



def getDigitiserOffsetTD(numSnapshots=1,globalSync=False,verbose=0,receptor=63,limitRange=False):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync()
        time.sleep(2)

    for a in range(numSnapshots):
        if verbose>3: print('Getting snapshot %i' % a)
        samples=getSamples(receptor=receptor);
        if a==0: corr=np.zeros((shape(samples)[1]));
        corr=corr+np.correlate(samples[0,:],samples[1,:],mode='same')
    corr=np.abs(corr)
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,100]);
        figure(figsize=(16,3))
        hist(samples.T,np.arange(-40,40)); title('voltage histogram (first snapshot only)');
        
    if verbose>1:
        figure(figsize=(16,3))
        plot(range(-8192,8192),corr); title('Correlation magnitude');
        xlim([-200,200]); grid(True);

    centreIndex=size(corr)/2;
    if limitRange:
        corr=corr[(centreIndex-8):(centreIndex+9)]
        print(shape(corr))
        centreIndex=size(corr)/2;
        
    offsetSamples = numpy.argmax(corr) - centreIndex;
    peakMag=corr[numpy.argmax(corr)];
    if verbose>0:
        print('Receptor %02i: %i snapshots. Samples offset: %4i.  Peak Ratio = %6.2f.  Voltage mean=%6.2f stdev=%5.2f max=%3i min=%3i.  AveragePower=%6.2f dB' % (receptor,numSnapshots,offsetSamples,peakMag/np.mean(corr),np.mean(samples),np.std(samples),np.max(samples),np.min(samples),10*log10(np.mean(samples**2)/261632.50) ))
    
    return offsetSamples





def powerCycleDigitiser(globalSync=True,verbose=0,receptor=101,digitiser=999):
    import time
    if (verbose>0)&~globalSync: print('  Power cycling digitiser (m%i) with absent / ready' % receptor);
    if (verbose>0)&globalSync: print('  Power cycling digitiser (m%i) with absent / ready, and Global Sync' % receptor);
    time.sleep(5)
    if verbose>1: print('   Selecting band 0');
    result=digitiserCommand("select-band 0",main=False,receptor=receptor)
    if result[0]: print('   Error selecting band 0')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Marking absent');
    result=digitiserCommand("digitiser-status l absent",main=False,receptor=receptor)
    if result[0]: print('   Error marking absent!')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Marking present');
    result=digitiserCommand("digitiser-status l ready %i" % digitiser,main=False,receptor=receptor)
    if result[0]: print('   Error marking present!')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Selecting band L');
    result=digitiserCommand("select-band l",main=False,receptor=receptor)
    if result[0]: print('   Error selecting band L!')
    if verbose>2: print(result[1])
    time.sleep(3)
    if globalSync:
        if verbose>1: print('   Global Sync');
        result=digitiserCommand("global-synchronise",main=True,receptor=receptor)
        if verbose>2: print(result[1])
    
```
