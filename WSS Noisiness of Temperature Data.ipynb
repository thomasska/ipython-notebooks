WSS Temperature - Noisy readings in March 2016
------------------
Plotting some temperature data from March 2016, looking at the noise

```python
import csv, numpy as np

temp = np.genfromtxt('airtemp2.csv', delimiter=",", dtype=float)        # read CSV
T=temp[:,0];  T=(T-1457560800e6)/86400e6;  # days after local midnight
D=temp[:,1];


temp = np.genfromtxt('aptemp2.csv', delimiter=",", dtype=float)        # read CSV
Tap=temp[:,0];  Tap=(Tap-1457560800e6)/86400e6;  # days after local midnight
Dap=temp[:,1];


temp = np.genfromtxt('m031_ap_struct_tilt_temp.csv', delimiter=",", dtype=float)        # read CSV
Ttilt=temp[:,0];  Ttilt=(Ttilt-1457560800e6)/86400e6;  # days after local midnight
Dtilt=temp[:,1];

temp = np.genfromtxt('m031_rsc_rsc_he_compressor_temperature2.csv', delimiter=",", dtype=float)        # read CSV
Tcomp=temp[:,0];  Tcomp=(Tcomp-1457560800e6)/86400e6;  # days after local midnight
Dcomp=temp[:,1];

temp = np.genfromtxt('m031_rsc_rxl_cryostat_body_temperature.csv', delimiter=",", dtype=float)        # read CSV
Tcryo=temp[:,0];  Tcryo=(Tcryo-1457560800e6)/86400e6;  # days after local midnight
Dcryo=temp[:,1];

temp = np.genfromtxt('anc.air_temperature.csv', delimiter=",", dtype=float)        # read RTS CSV
Trts=temp[:,0];  Trts=(Trts-1457560800e6)/86400e6;  # days after local midnight
Trts=Trts[Drts>0];
Drts=temp[:,1];
Drts=Drts[Drts>0];

# smooth the WSS readings
ws=100;
w=hamming(ws);   w=w/sum(w)
Dsm = convolve(D,w,mode='same')
Dsmooth=D-Dsm
Dsmooth=Dsmooth[ws:-ws];
T=T[ws:-ws]
D=D[ws:-ws]
Dsm = Dsm[ws:-ws]
```

```python
figure(figsize=(28,10))
plot(T,D,alpha=0.5,label='WSS, raw');
plot(T,Dsm,label='WSS, smoothed');
plot(Tap,Dap,label='AP (slowest)');
plot(Ttilt,Dtilt,lw=1,alpha=0.4,label='Tilt sensor',color='g');
plot(Tcomp,Dcomp,lw=1,alpha=0.4,label='Compressor',color='r');
plot(Tcryo,Dcryo,lw=1,alpha=0.4,label='Cryo body',color='b');
plot(Trts,Drts,lw=1,alpha=0.4,label='RTS/KAT7 Wx',color='k');
xlabel('days after local midnight on 10 March'); ylabel('Temperature, deg C');
grid(True); xlim([1,6]); ylim([15,45]); legend(loc=0)



figure(figsize=(28,3))
plot(T,Dsmooth);
xlabel('days'); ylabel('High Pass Temperature, hamming %i, deg C' % ws);
grid(True); xlim([0,13])


wd=500;
w=hamming(wd);  w=w/sum(w)
Dpower=sqrt(convolve(Dsmooth**2.0,w,mode='same'));


figure(figsize=(28,3))
plot(T,Dpower);
plot(T[Dpower>0.4],Dpower[Dpower>0.4],color='r',marker='.',lw=0,ms=2);
xlabel('days'); ylabel('Temperature HF power');
grid(True); xlim([0,13])

figure(figsize=(12,5))
hist(Dsmooth[(Dpower>0.5)],100,color='r');
title('Histogram of high-pass-temperature, high power times only');
```

Variation of the noise power with time and reported air temperature
---------------------------------------------------------------------

There's almost no correlation with its own reported temperature:

```python
figure(figsize=(8,5)); title('Temperature dependence of the temperature noise power');
plot(D,Dpower,lw=0,marker='.',alpha=0.01,ms=10);  xlabel('Temperature');  ylabel('High Pass Temperature Power'); xlim([15,40]);
```

```python
figure(figsize=(8,5)); title('Temperature dependence of the temperature noise power, using AP temperature');
plot(np.interp(T,Tap,Dap),Dpower,lw=0,marker='.',alpha=0.01,ms=10,color='r');  xlabel('AP measured Temperature(slow)');  ylabel('High Pass Temperature Power'); xlim([15,40]);
```

```python
figure(figsize=(8,5)); title('Temperature dependence of the temperature noise power, using Cryo temperature');
plot(np.interp(T,Tcryo,Dcryo),Dpower,lw=0,marker='.',alpha=0.01,ms=10,color='m');  xlabel('Cryo body measured Temperature(slow)');  ylabel('High Pass Temperature Power'); xlim([15,40]);
```

```python
figure(figsize=(8,5)); title('Temperature dependence of the temperature noise power, using KAT-7 RTS WSS temperature');
plot(np.interp(T,Trts,Drts),Dpower,lw=0,marker='.',alpha=0.01,ms=10,color='k');  xlabel('KAT-7 air Temperature');  ylabel('High Pass Temperature Power'); xlim([15,40]);
```

Fairly well correlated with site temperature, it's more noisy when the real temperature is low.

Noise power over the day
-------------------------

```python
figure(figsize=(8,5)); title('Time dependence of the temperature noise power');
plot((T*24.0)%24,Dpower,lw=0,marker='.',alpha=0.005,ms=15,color='g');  xlabel('Hour of day');  ylabel('High Pass Temperature Power');  xlim([0,24]);
```

The high frequency noise power is strongly time dependent. Depending on the temperature I suppose.
