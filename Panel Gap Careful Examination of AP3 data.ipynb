```python
def loadFile(fileName,cax=[10,40],dr=[1000,2000],lineFreq=620):
    #global D, combPower, combFreq, noisePower, F
    
    # define some functions
    import katdal, csv, numpy as np, matplotlib, random
    
    
    D=katdal.open(fileName);
    
    D.select(corrprods=[0,1]);
    
    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # extract the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    
    noisePowerMin = np.min(noisePower,axis=0);                                                      # find the minimum along the time axis, this is the true noise power, removes some spikes etc.
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=combchannels);             # the comb power
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis
    
    combPower=real(D.vis[:]);
    
    combPower = combPower - noisePowerMin;                                                          # find the additional power on the comb channels
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);

    
    
    #Detailed examination of just one comb line, selected by hand

    D.select();
    D.select(freqrange=(1712e6-(lineFreq+2)*1.e6,1712e6-(lineFreq-2)*1.e6), corrprods=[0], dumps=range(dr[0],dr[1]));
    
    F=1712-D.channel_freqs/1.e6;
    V=squeeze(np.abs(D.vis)) #                     * 10.0**(atten/10.0);     # including attenuation so figures will be comparable later   - CAREFUL what you do with attenuation, it doesn't apply to the noise!
    
    VdB = 10*log10(V)
    
    fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(25,12));
    axes[0,0].imshow(VdB.T,aspect='auto',interpolation='none', cmap=plt.get_cmap('CMRmap'),vmin=cax[0],vmax=cax[1],extent=[0,shape(noisePower)[0],F[-1],F[0]]);
    
    axes[0,1].step(VdB.T,F,color='k',alpha=0.1,lw=1.5);  axes[0,1].set_xlim(cax);  axes[0,1].grid(True)
    
    axes[1,0].plot(VdB[:,8],color='k',alpha=0.5,lw=1,label='comb channel');
    axes[1,0].plot(VdB[:,7],color='b',alpha=0.2,lw=2,label='+1 ch');
    axes[1,0].plot(VdB[:,6],color='b',alpha=0.2,lw=2,label='+1 ch');
    axes[1,0].plot(VdB[:,9],color='g',alpha=0.2,lw=2,label='-1 ch');
    axes[1,0].plot(VdB[:,10],color='g',alpha=0.2,lw=2,label='-2 ch');
    axes[1,0].legend(loc=0);  axes[1,0].set_ylim(cax);  axes[1,0].grid(True)
    print('H pol only - %s  -  %s' % (D.experiment_id,D.description))
    
    mnoise = np.mean(V[:,[6,7,10,11]]);
    totcomb = np.mean(np.sum(V[:,[8,9]],axis=1),axis=0) - 2.0*mnoise;
    
    print('Mean noise power, channels -3, -2, 1, 2, is %f ' %  mnoise)
    print('Total comb power, channels -1, 0, +1, with two channels of noise removed, is  %f' %  totcomb)

    allnoise = np.mean(V[:,[6,7,10,11]],axis=1);
    allcomb0 = np.sum(V[:,[9]],axis=1);   # comb signal, which is in two channels, plus some noise
    allcomb1 = np.sum(V[:,[8]],axis=1);   # comb signal, which is in two channels, plus some noise
    allcomb = allcomb0+allcomb1;
    allcombnonoise = allcomb - 2.0*allnoise;
    
    figure(figsize=(12,8)); plot(allnoise,label='noise');  plot(allcomb0,label='comb0'); plot(allcomb1,label='comb1');  plot(allcomb,label='comb tot'); plot(allcombnonoise,label='comb tot, no noise'); legend(loc=0); #ylim([0,10])
    
    print('Total comb power, channels -1, 0, +1, with noise over time removed, is  %f' %  np.mean(allcombnonoise))
```
