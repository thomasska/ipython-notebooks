## Loading the data

```python
# IMPORT MODULES AND LOAD THE DATA
# preliminary stuff
import katdal
import csv
import numpy as np
import matplotlib
import matplotlib.gridspec as gridspec

# load the measurement files
# these are with no attenuation, short RG223 cable. Source power TBD. Source position: 1 m off ground, 5 m behind back of telescope. (directly on the axis of the shaft when at az0,el15.

# measurements 3 and 4, source is further away, on the edge of the concrete circle
#dsh=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/03/11/1426077546.h5');   # first attempt, strange movements.

DV=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/03/11/1426077938.h5');   # big scan 1 source Vertical Attenuator 0 Cable xxx
DH=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/03/11/1426079592.h5');   # big scan 2 source Horizontal Attenuator 0 Cable xxx

#dsh=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/03/11/1426081186.h5');   # big scan 3 source Horizontal Attenuator 0 Cable xxx
#dsh=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/03/11/1426082374.h5');   # big scan 4 source Vertical Attenuator 0 Cable xxx

print 'Loaded two files, ',DV.name,' and ',DH.name
```

```python
# Create a time axis for both channels, with a small offset to allow synchronised plotting
TV = DV.timestamps - DV.timestamps[0] + 6;
TH = DH.timestamps - DH.timestamps[0];


# Create a flipped frequency axis for plotting purposes
F=1712-dsh.channel_freqs/1e6;

# create a cropped frequency axis being only those allowed through the horn and the filters
freqrange=[450,850];  # one neat elevation cut
sanechannels = [i for i, v in enumerate(F) if ((v<=freqrange[1])&(v>=freqrange[0])) ];
FS = F[sanechannels];

# find the channels which would contain the integer multiples of 10 MHz
combchannels = [i for i, x in enumerate(abs((F*1e6+DV.channel_width/2) % 10e6)-DV.channel_width/2) if (x < DV.channel_width/2)];
# remove combchannels which are outside the chosen frequency range, not pretty but it works
temp = combchannels[:];
for a in temp:
    if (not(a in sanechannels)):
        combchannels.remove(a);
FC = F[combchannels];

# create a frequency axis which does not contain the comb channels
sanechannelsnocomb = sanechannels[:];
temp = sanechannelsnocomb[:];
for a in temp:
    if (a in combchannels):
        sanechannelsnocomb.remove(a);
FSNC = F[sanechannelsnocomb];


print 'Dump period = ',DV.dump_period,'seconds'
print 'Start Time = V:',DV.start_time,'Z','  H:',DH.start_time,'Z'
print 'End Time = V:',DV.end_time,'Z','  H:',DH.end_time,'Z'
print 'Az range = V',np.min(DV.az),' to ',np.max(DV.az),'  H:',np.min(DH.az),' to ',np.max(DH.az)
print 'El range = V',np.min(DV.el),' to ',np.max(DV.el),'  H:',np.min(DH.el),' to ',np.max(DH.el)

print 'Found ', size(combchannels), ' comb channels'
print '  Channels ',combchannels[0],combchannels[1],combchannels[2], '...',combchannels[-1]
print '  Freqs ',FC[0],FC[1],FC[2], '...',FC[-1], ' MHz'
```

```python
# load the CSV of comb output powers and interpolate
attenuator = 0;    # actual attenuator used in the experiments
temp = np.genfromtxt('panelgap/comb10MHz.csv', comments="#", delimiter=",", dtype=float)        # read CSV

combpower=zeros(shape(FC));
for i in range(shape(FC)[0]):
    idx = abs(temp[:,0]-FC[i]).argmin();
    combpower[i]=temp[idx,1];
    #print('number %01i, MeerKAT freq %6.2f MHz, Index %i, Comb freq %5.1f MHz, Comb power %4.1f' % (i,FC[i],idx,temp[idx,0],combpower[i]))

#combpower = np.power(10.0,((combpower-30-attenuator)/10.0));    # convert to W, add attenuator

figure();
scatter(FC,combpower); plot(FC,combpower);
#xlim([450,850]);  ylim([-46,-25]);
xlabel('Frequency, MHz');  ylabel('Comb line power, dBm');  title('Comb generator, 10 MHz setting, at SMA connector of shielded box');

combpower=np.power(10.0,(combpower-30-attenuator)/10.0);

print 'Loaded comb generator data, in Watts, aligned to your frequencies'
#for i in range(20): print('at f=%6.2f, comb power=%6.2g W' % (FC[i],combpower[i]))
```

The conclusion here, is sadly that there is no reliable way to detect the strange clipping behaviour of the ADC.  
The effect will happen during a cycle, in the time domain, and slowly add power to the signal in a fairly analogue way.  
Even at the fringes of the areas of strangeness, there is a gentle ramp up.
