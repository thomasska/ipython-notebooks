```python
# some functions you'll need - run this first!

from IPython.core.display import Image
import katdal, numpy, matplotlib
##############################################################################################################################3
# load one file, return the comb frequencies and additional noise  power in correlator power units
def loadRaw(fileName,atten=-10,azLimit=[-190,190],elLimit=[-1,90]):
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);

    D.select(corrprods=[0,1]);

    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # extract the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    noisePower = np.min(noisePower,axis=0);                                                      # find the minimum along the time axis, this is the true noise power, removes some spikes etc.
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=combchannels);             # the comb power
    combPower=real(D.vis[:]);
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis

    combPower = combPower - noisePower;                                                          # find the additional power on the comb channels
    
    # restrict the azimuth and elevation range
    limitsBool=np.logical_and(   np.logical_and(np.greater(D.az,azLimit[0]),np.less(D.az,azLimit[1])),    np.logical_and(np.greater(D.el,elLimit[0]),np.less(D.el,elLimit[1]))   );
    limitsBool=reshape(limitsBool,shape(limitsBool)[0]);             # in desparation
    combPower = combPower[limitsBool,:,:];
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);
    
    # Apply the attenuator... so Increase the power received by the attenuator value
    combPower = combPower * 10.0**(atten/10.0);
    
    print('RawLoaded %s, atten=%4.1fdB, returned %i dumps' % (D.name,atten,shape(combPower)[0]));
    
    return combFreq,combPower;




##############################################################################################################################3
# given the frequencies in MHz, returns the power of the relevant comb line, in watts, at the end of the cable, without any attenuators
def sigGenPower(FC,filename='150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv'):
    temp = np.genfromtxt('panelgap/'+filename, comments="#", delimiter=",", dtype=float)        # read CSV
    sgpower=np.zeros(shape(FC[:]));
    for a in range(size(FC)):
        sgpower[a]=temp[(  np.abs( temp[:,0] - np.round(FC[a]/10)*10 )  ).argmin(),1];
    sgpower = 10**((sgpower-30+10)/10.0);         # convert to watts, removing the 10 dB attenuator!
    return sgpower
```

Finding and fixing the correlator saturation

```python
# load the files I need
F,tempDirect1=loadRaw('05/27/1432731158.h5',atten=36);                                                  # AP2 old direct meas
F,tempDirect3=loadRaw('07/29/1438163010.h5',atten=50);                                                  # AP2 July direct meas

# comb power levels, dBm
C1=10*log10(sigGenPower(F,'150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv'));
C2=10*log10(sigGenPower(F,'150610 Comb-10dB-BothCables-Orientation1-tight.csv'));
C3=10*log10(sigGenPower(F,'150610 Comb-10dB-BothCables-Orientation2-tight.csv'));
C=(C1+C2+C3)/3.0     
```

```python
figure(figsize=(10,2.5))
plot(F,C1,label='May measurement');
plot(F,C2,label='June 1');
plot(F,C3,label='June 2');
legend(loc=0); xlabel('Frequency, MHz');  ylabel('dBm');   grid(True);
title('Comb generator output power, dBm');
```

Saturation can be seen here - some channels go up to > 80 dB and stay there.  
(80 dB is the saturated correlator accumulator, plus some compensation for the attenuator used on the signal generator).

So I plot a histogram of every channel, showing the range of values seen over time

```python
peakLimit=80.36;   # super-careful choice of this number traps just the right list of frequencies

# plot a histogram for each frequency, showing saturation?
f=range(32);
fig,ax=subplots(nrows=4,ncols=8,figsize=(25,14))
for a,f in enumerate(F[0:32]):
    ax[a/8,a%8].hist(10*log10(abs(tempDirect1[t0,a,0])),range=[72,81],bins=36,color='green',alpha=0.4);
    ax[a/8,a%8].hist(10*log10(abs(tempDirect1[t1,a,1])),range=[72,81],bins=36,color='red',alpha=0.4);
    
    if (max(10*log10(abs(tempDirect1[t0,a,0])))>80.3)|(max(10*log10(abs(tempDirect1[t1,a,1])))>peakLimit):   # was there saturation in either channel over their time range
        ax[a/8,a%8].plot(75,30,color='red',marker='*',markersize=25);
    
    ax[a/8,a%8].set_title('%i: %iM, %3.0fdBm' % (a, f,C3[a]));  ax[a/8,a%8].set_xlim([72,81]);  ax[a/8,a%8].set_ylim([0,40]);  
```

```python
# Try to fix the faulty frequencies by interpolation, using the known signal generator power

print('*** Fixing data with a peak power exceeding %4.1f dBCU ***' % peakLimit);

Direct1=np.dstack((tempDirect1[140:210,:,0],tempDirect1[240:310,:,1]));    # find the direct path, stack each pol into a new dual-pol vector

for a,f in enumerate(F):
    if (max(10*log10(abs(Direct1[:,a,0])))>80.3) | (max(10*log10(abs(Direct1[:,a,1])))>peakLimit):       # detect saturation
        oldDirect1=10*log10(mean(Direct1[:,a,:]));
        Direct1[:,a,:] = 0.5 * ( Direct1[:,a-1,:]*(10.0**(-C[a-1]/10)) + Direct1[:,a+1,:]*(10.0**(-C[a+1]/10)) )*(10.0**(C[a]/10));  #  (interpolate from the two sides, all corrected for signal generator power C)
        print('Fixing number %i at freq %i MHz, avg was %4.1f, now %4.1f' % (a,f,oldDirect1,10*log10(mean(Direct1[:,a,:]))))
```

```python
# plot a histogram for each frequency, showing saturation again after repairs
f=range(32);
fig,ax=subplots(nrows=4,ncols=8,figsize=(25,14))
for a,f in enumerate(F[0:32]):
    ax[a/8,a%8].hist(10*log10(abs(Direct1[:,a,0])),range=[72,91],bins=76,color='green',alpha=0.4);
    ax[a/8,a%8].hist(10*log10(abs(Direct1[:,a,1])),range=[72,91],bins=76,color='red',alpha=0.4);
    
    ax[a/8,a%8].set_title('%i: %iM, %3.0fdBm' % (a, f,C3[a]));  ax[a/8,a%8].set_xlim([72,91]);  ax[a/8,a%8].set_ylim([0,40]);  
```

It seems that this could fix the saturation!
