Off-axis gain of MeerKAT at UHF
-----------

Gain measured by placing a transmitter on the ground, 100 m away from the antenna, and rotating around 360 degrees.

```python
# May measurement - direct path measurement
Direct1=np.dstack((tempDirect1[140:210,:,0],tempDirect1[240:310,:,1]))     # time-selection for the different polarisations of the sources, choose appropriate channel, re-stack again

# fix the saturation mess
for a,f in enumerate(F):
    if (max(10*log10(abs(Direct1[:,a,0])))>80.3) | (max(10*log10(abs(Direct1[:,a,1])))>80.3):       # detect saturation
        oldDirect1=10*log10(mean(Direct1[:,a,:]));
        Direct1[:,a,:] = 0.5 * ( Direct1[:,a-1,:]*(10.0**(-C[a-1]/10)) + Direct1[:,a+1,:]*(10.0**(-C[a+1]/10)) )*(10.0**(C[a]/10));  #  (interpolate from the two sides, all corrected for signal generator power C)
        print('%i Fixing Freq %i MHz, avg was %4.1f, now %4.1f dB' % (a,f,oldDirect1,10*log10(mean(Direct1[:,a,:]))))

Direct1 = Direct1 / (Dsubreflpda/7.0)**2.0;                                # reduce it so we are at the same distance. Focus-Dish-LPDA = 20 m.  Direct Horn-LPDA = 5 m. 

# average down Direct1 to make it only Nchannels x Npol
Direct1 = mean(Direct1,axis=0);

# normalise V and H to direct path measurements
V = V / Direct1;
H = H / Direct1;
```

```python
# define some functions
import katdal, csv, numpy as np, matplotlib, random

# some global constants
Dsubreflpda=100;                            # distance in metres from subreflector (or prime focus?) to LPDA.
#Gsubref = 10*log10(42000.0/(50.0*50.0));   # estimate for the subreflector gain
#GainHorn = np.interp(F,[499,550,600,700,800,900],[2.0,9.8,12.2,12.5,12.2,10.5]);    # horn gain,  from EA-MKU-1000-DREP-01_2   Figure 13 page 22, extrapolated


##############################################################################################################################3
def queryFile(fileName,azLimit=[10,10],elLimit=[20,20]):
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);
    T = D.timestamps - D.timestamps[0];
    fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12,7));
    axes[0,0].plot(D.az,D.el,color='black',lw=2,label='Measurement');
    axes[0,0].plot([azLimit[0],azLimit[0],azLimit[1],azLimit[1],azLimit[0]],[elLimit[0],elLimit[1],elLimit[1],elLimit[0],elLimit[0]],color='blue',ls='--',lw=1,label='Range Used')
    axes[0,0].set_title('Az vs El');
    axes[0,0].legend(loc=7);
    axes[0,1].plot(T,D.az,label='Az',color='blue',lw=4);
    axes[0,1].plot(T,D.el,label='El',color='red',lw=4);
    axes[0,1].set_title('Az and El');  axes[0,1].legend(loc=6);
    D.select(corrprods=[0,1]);
    axes[1,0].plot(1712-D.channel_freqs/1e6,10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=0))));   axes[1,0].set_title('average spectrum');  axes[1,0].set_xlim([500,850]); axes[1,0].set_ylim([10,50]);
    axes[1,1].plot(10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=1))));   axes[1,1].set_title('power over time');
    print('Contents of %s - %i dumps' % (D.name,shape(D.vis)[0]))
    #print(D)
    


##############################################################################################################################3
# load one file, return the comb frequencies and additional noise  power in correlator power units
def loadRaw(fileName,atten=-10):
    D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + fileName);

    D.select(corrprods=[0,1]);

    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # extract the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    noisePower = np.min(noisePower,axis=0);                                                      # find the minimum along the time axis, this is the true noise power, removes some spikes etc.
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=combchannels);             # the comb power
    combPower=real(D.vis[:]);
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis

    combPower = combPower - noisePower;                                                          # find the additional power on the comb channels
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);
    
    # Apply the attenuator... so Increase the power received by the attenuator value
    combPower = combPower * 10.0**(atten/10.0);
    
    print('RawLoaded %s, atten=%4.1fdB, returned %i dumps' % (D.name,atten,shape(combPower)[0]));
    
    return combFreq,combPower,D.az,D.el;




##############################################################################################################################3
# given the frequencies in MHz, returns the power of the relevant comb line, in watts, at the end of the cable, without any attenuators
def sigGenPower(FC,filename='150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv'):
    temp = np.genfromtxt('panelgap/'+filename, comments="#", delimiter=",", dtype=float)        # read CSV
    sgpower=np.zeros(shape(FC[:]));
    for a in range(size(FC)):
        sgpower[a]=temp[(  np.abs( temp[:,0] - np.round(FC[a]/10)*10 )  ).argmin(),1];
    sgpower = 10**((sgpower-30+10)/10.0);         # convert to watts, removing the 10 dB attenuator!
    return sgpower
```
