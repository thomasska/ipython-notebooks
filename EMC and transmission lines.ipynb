## Input impedance of a simple transmission line

```python
import matplotlib
import numpy

Length=3;  # metres
Zload=10;  # Ohms
Z0 = 100;  # Ohms
c = 3e8;

# signal generator parameters
Vsource = 10;
Zsource=50;

F=logspace(3,7.5,101);
Lambda = c/F;
Beta=2*pi/Lambda;

# transmission line equation
Zin = Z0 * (Zload + 1j*Z0*tan(Beta*Length))/(Z0 + 1j*Zload*tan(Beta*Length))


fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(10,7));
axes.semilogx(F,real(Zin),lw=1,alpha=0.7,color='green',label='Real');
axes.semilogx(F,imag(Zin),lw=1,alpha=0.7,color='blue',label='Imag');
axes.semilogx(F,abs(Zin),lw=2,alpha=0.7,color='red',label='Magnitude');
axes.legend(loc=2);
axes.set_xlabel('Frequency, Hz');
axes.set_ylabel('Impedance, Ohms');
#axes.set_xlim([925,960]); axes.set_ylim([15, 45]);
axes.set_title('Input impedance of a length of 2-core flex, terminated with a resistor');
fig.tight_layout();

```

```python
# voltage divider
Vin = Vsource * Zin / (Zsource+Zin);

fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(10,7));
axes.semilogx(F,abs(Vin),lw=2,alpha=0.7,color='red',label='Vin');
axes.legend(loc=2);
axes.set_xlabel('Frequency, Hz');
axes.set_ylabel('Voltage, V');
axes.set_title('Voltage seen on the input to the experiment');
fig.tight_layout();
```
