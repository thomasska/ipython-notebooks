Measurement Method
--------------

Measurements were taken of the transmission through AP1 (with bent gap covers fitted) and AP2 (no gap covers).

The source used was an LPDA connected to a comb generator, producing a signal every 10 MHz.  
The actual MeerKAT UHF receiver was used, connected to a modified L-band receiver, sampling in the 1st Nyquist zone.  
The receiver was in the warm state as the helium pipe was not yet connected, but the comb power was still large, many times the system noise.

Calibration measurements
--------------
The comb generator an antenna gain are removed from the calculation by a direct measurement.  
The indexer was turned to face the UHF receiver out of the antenna focus, and the source held 5 m from the horn.

Analysis of a single measurement
-----------------
Each measurement was performed by scanning the dish up and down, at several azimuth angles. The measurement was repeated with the source mounted in each polarisation.  
The correlator dump rate was set to 4 Hz; with a slew speed of 1&deg;/s this provides points about 0.25&deg; apart in elevation, which is sufficient to fully sample the interference pattern from the many slot radiators.  
An example of the scan pattern and raw results is show below.

Source antenna was mounted 5 m behind the dish, on a tripod, 1 m from the ground.

Calculation of leakage through dish
----------------------
The transmission through the dish surface is both compared to a direct LPDA->Horn measurement (with no dish in the way).  
After compensation for the change in range, this yields the "transmission coefficient" of the dish, including transmission and diffraction around the sides.  
Diffraction, a component of spill-over, is not part of the AP specfication.

Then, assuming the gap covers on AP1 work perfectly, we can assume this is all diffraction *around* the dish, and use it to remove any diffraction on AP2.  
This yields a final result for the leakage **through AP2 with no gap covers**.

A further measurement was performed on AP2 after installation of the flat gap covers, in July.

```python
# May measurement - direct path measurement
Direct1=np.dstack((tempDirect1[140:210,:,0],tempDirect1[240:310,:,1]))     # time-selection for the different polarisations of the sources, choose appropriate channel, re-stack again

# fix the saturation mess
peakLimit = 80.36;
for a,f in enumerate(F):
    if (max(10*log10(abs(Direct1[:,a,0])))>80.3) | (max(10*log10(abs(Direct1[:,a,1])))>peakLimit):       # detect saturation
        oldDirect1=10*log10(mean(Direct1[:,a,:]));
        Direct1[:,a,:] = 0.5 * ( Direct1[:,a-1,:]*(10.0**(-C[a-1]/10)) + Direct1[:,a+1,:]*(10.0**(-C[a+1]/10)) )*(10.0**(C[a]/10));  #  (interpolate from the two sides, all corrected for signal generator power C)
        print('%i Fixing Freq %i MHz, avg was %4.1f, now %4.1f dB' % (a,f,oldDirect1,10*log10(mean(Direct1[:,a,:]))))

Direct1 = Direct1 / (Dsubreflpda/7.0)**2.0;                                # reduce it so we are at the same distance. Focus-Dish-LPDA = 20 m.  Direct Horn-LPDA = 5 m. 

# July measurement - direct path measurement
Direct22=np.dstack((tempDirect22[260:400,:,0],tempDirect22[490:630,:,1]))     # time-selection for the different polarisations of the sources, choose appropriate channel, re-stack again
Direct22 = Direct22 / (Dsubreflpda/5.0)**2.0;                                # reduce it so we are at the same distance. Focus-Dish-LPDA = 20 m.  Direct Horn-LPDA = 7 m. 

# July measurement - direct path measurement
Direct3=np.dstack((tempDirect3[260:400,:,0],tempDirect3[490:630,:,1]))     # time-selection for the different polarisations of the sources, choose appropriate channel, re-stack again
Direct3 = Direct3 / (Dsubreflpda/6.0)**2.0;                                # reduce it so we are at the same distance. Focus-Dish-LPDA = 20 m.  Direct Horn-LPDA = 7 m. 

# find the average transmission through the three dishes
AP1v=np.mean(tempAP1v,axis=0);  AP1h=np.mean(tempAP1h,axis=0);
AP2v=np.mean(tempAP2v,axis=0);  AP2h=np.mean(tempAP2h,axis=0);
AP22v=np.mean(tempAP22v,axis=0);  AP22h=np.mean(tempAP22h,axis=0);
AP3v=np.mean(tempAP3v,axis=0);  AP3h=np.mean(tempAP3h,axis=0);

# find the ratio of Through-Panel to DirectPath. ( Broadcasting enlarges this to  (60 x 36channels x 2pol) )
AP1v = AP1v / Direct1;   AP1h = AP1h / Direct1;
AP2v = AP2v / Direct1;   AP2h = AP2h / Direct1;
AP22v = AP22v / Direct22;   AP22h = AP22h / Direct22;
AP3v = AP3v / Direct3;   AP3h = AP3h / Direct3;

# now add the polarisations, and take the mean of the two
AP1 = ( sum(AP1v,axis=2) + sum(AP1h,axis=2) )/2.0
AP2 = ( sum(AP2v,axis=2) + sum(AP2h,axis=2) )/2.0
AP22 = ( sum(AP22v,axis=2) + sum(AP22h,axis=2) )/2.0
AP2 = ( sum(AP3v,axis=2) + sum(AP3h,axis=2) )/2.0

# multiply in the contribution of the signal generator error
AP1=AP1*DR;

# 1/lambda^2 line
fineFreq=np.linspace(400,1000,101)
linePower=6.0;
fineLine=0.8*550**linePower * fineFreq**-linePower;


# plot
alp=0.08;   # plot alpha
figure(figsize=(12,5));
plot(0,0,color='blue',label='AP1 bent covers',lw=3);
plot(0,0,color='green',label='AP2 no covers',lw=3);
plot(0,0,color='red',label='AP2 July with flat covers',lw=3);
plot(0,0,color='black',label='AP2-AP1, (assuming AP1 is all diffraction)',lw=3);
#plot(0,0,color='m',label='$\lambda^%i$' % linePower,lw=1,linestyle='--');
plot(F,100.0*AP1.T,color='blue',alpha=alp,lw=3);
plot(F,100.0*AP2.T,color='green',alpha=alp,lw=3);
plot(F,100.0*AP22.T,color='red',alpha=alp,lw=3);
plot(F,100.0*(AP2.T-AP1.T),color='black',alpha=alp,lw=2);
plot([580,580,900],[10,0.5,0.5],lw=1,ls=':',color='black');
#plot(fineFreq,fineLine,color='m',lw=1,linestyle='--');
ylim([0,1.3]);  xlim([550,800]);  xlabel('Frequency, MHz');   ylabel('% transmission');
text(720,0.55,'R.AP.C.77: Reflector efficiency \n >= 99.5 % 580 MHz - 14.5 GHz.');
title('Measured transmission through and around the dish, %');
legend(loc=0);
```

Conclusions
-------------
1. The reflector on AP2, without gap covers, fails on the requirement to reflect 99.5% of the wave. At lower frequencies, as much as 1.5% of the power is transmitted.
1. Above 700 MHz, both antennas appear to meet the 99.5% requirement, from a reflector efficiency point of view
1. A more careful repeat of the measurement, with slightly better technique, yields a slightly lower leakage result.
1. The flat gap covers work as well as the bent gap covers

Error Analysis
===============

Three significant sources of uncertainty were noted and have been included in the modelling for the results above.

1) Correlator saturation from direct comb measurement
------------

It has been discovered later that the correlator saturated when the measurements of the direct path were taken.  
This affected only odd channels, and could be somewhat corrected by interpolating from the adjacent channels. This has been done for the data above, but the process was not perfect and there may still be some saturation in the data.

The effect of the saturation error is to increase the calculated leakage.

2) Direct path variations due to reflections from ground and structure
-----------
During the 7 m direct path measurements, the source antenna was moved slowly up and down over about 0.5 m. This caused some variation in the transmission from LPDA to Horn, depending on the environmental reflections.

```python
lossV=10.0*log10(sum(tempDirect1[480:640,:,:],axis=2));                           # time-selection for the different polarisations of the sources
lossH=10.0*log10(sum(tempDirect1[240:420,:,:],axis=2));                           # time-selection for the different polarisations of the sources

figure(figsize=(10,2.5));
plot(F,np.mean(lossH,axis=0),color='red',lw=2,label='HH');
plot(F,np.mean(lossV,axis=0),color='green',lw=2,label='VV');
ylim([65,85]);  title('Mean calibration power, direct measurement 7 m from horn, over frequency');  ylabel('dB');  xlabel('Frequency, MHz');   grid(True);
figure(figsize=(10,2.5));
plot(F,(lossH-np.mean(lossH,axis=0)).T,color='red',alpha=0.15,lw=4);
ylim([-4,4]); title('HH power, deviation around the mean at each frequency');  ylabel('dB'); grid(True);
annotate('Horn is below cutoff here', xy=(527, 0.8),  xycoords='data', xytext=(-20, 30), textcoords='offset points', arrowprops=dict(arrowstyle="->"))
figure(figsize=(10,2.5));
plot(F,(lossV-np.mean(lossV,axis=0)).T,color='green',alpha=0.15,lw=4);
ylim([-4,4]); title('VV power, deviation around the mean at each frequency');  ylabel('dB');  xlabel('Frequency, MHz');   grid(True);
```

This variation has been included in the graph above. Each of the 60 direct-path calibration measurements has been used for calculation, which causes the spreading of the lines in the final graph.

3) Comb generator power output uncertainty
-----------
The output of the comb generator box is slightly variable, depending on how tightly the shielded box is assembled.  
This matters because the comb generator was re-assembled between the AP1 and AP2 measurements, and the calibration was only done during the AP2 tests.

The maximum variation, per frequency, has been added to the AP1 data, and included in the final graph above.

```python
# define some functions
import katdal, csv, numpy as np, matplotlib, random

# some global constants
Dsubreflpda=17;                            # distance in metres from subreflector (or prime focus?) to LPDA.
#Gsubref = 10*log10(42000.0/(50.0*50.0));   # estimate for the subreflector gain
#GainHorn = np.interp(F,[499,550,600,700,800,900],[2.0,9.8,12.2,12.5,12.2,10.5]);    # horn gain,  from EA-MKU-1000-DREP-01_2   Figure 13 page 22, extrapolated


##############################################################################################################################3
def queryFile(fileName,azLimit=[10,10],elLimit=[20,20]):
    D=katdal.open(fileName);
    T = D.timestamps - D.timestamps[0];
    fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12,7));
    axes[0,0].plot(D.az,D.el,color='black',lw=2,label='Measurement');
    axes[0,0].plot([azLimit[0],azLimit[0],azLimit[1],azLimit[1],azLimit[0]],[elLimit[0],elLimit[1],elLimit[1],elLimit[0],elLimit[0]],color='blue',ls='--',lw=1,label='Range Used')
    axes[0,0].set_title('Az vs El');
    axes[0,0].legend(loc=7);
    axes[0,1].plot(T,((D.az+0.2) % 360)-0.2,label='Az',color='blue',lw=4);
    axes[0,1].plot(T,D.el,label='El',color='red',lw=4);
    axes[0,1].set_title('Az and El');  axes[0,1].legend(loc=6);
    D.select(corrprods=[0,1]);
    axes[1,0].plot(1712-D.channel_freqs/1e6,10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=0))));   axes[1,0].set_title('average spectrum');  axes[1,0].set_xlim([500,850]); axes[1,0].set_ylim([10,50]);
    axes[1,1].plot(10*log10(squeeze(mean(real(D.vis[:,:,:]),axis=1))));   axes[1,1].set_title('power over time');
    print('Contents of %s - %i dumps' % (D.name,shape(D.vis)[0]))
    #print(D)
    


##############################################################################################################################3
# load one file, return the comb frequencies and additional noise  power in correlator power units
def loadRaw(fileName,atten=-10,azLimit=[-190,190],elLimit=[-1,90]):
    D=katdal.open(fileName);

    D.select(corrprods=[0,1]);

    # Create a flipped frequency axis for plotting purposes
    F=1712-D.channel_freqs/1e6;                                                                  # whole frequency axis
    
    # find the channels which contain the integer multiples of 10 MHz
    combchannels = [i for i, x in enumerate(abs((F*1e6+D.channel_width/2) % 10e6)-D.channel_width/2) if (x < D.channel_width/2)];
    
    # extract the signal power and noise power
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,-2));  # the noise power, use the channel below.
    noisePower=real(D.vis[:])*0.5;
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=np.add(combchannels,2));   # the noise power, use the channel above.
    noisePower=noisePower + real(D.vis[:])*0.5;
    noisePower = np.min(noisePower,axis=0);                                                      # find the minimum along the time axis, this is the true noise power, removes some spikes etc.
    
    D.select(reset='F',freqrange=(1712e6-851e6,1712e6-498e6),channels=combchannels);             # the comb power
    combPower=real(D.vis[:]);
    combFreq = np.round(1712-D.channel_freqs[:]/1e6);                                            # corresponding frequency axis

    combPower = combPower - noisePower;                                                          # find the additional power on the comb channels
    
    # restrict the azimuth and elevation range
    limitsBool=np.logical_and(   np.logical_and(np.greater(D.az,azLimit[0]),np.less(D.az,azLimit[1])),    np.logical_and(np.greater(D.el,elLimit[0]),np.less(D.el,elLimit[1]))   );
    limitsBool=reshape(limitsBool,shape(limitsBool)[0]);             # in desparation
    combPower = combPower[limitsBool,:,:];
    
    # flip combPower and combFreq to get them the right way around
    combFreq=combFreq[::-1];
    combPower=np.fliplr(combPower);
    
    # Apply the attenuator... so Increase the power received by the attenuator value
    combPower = combPower * 10.0**(atten/10.0);
    
    print('RawLoaded %s, atten=%4.1fdB, returned %i dumps' % (D.name,atten,shape(combPower)[0]));
    
    return combFreq,combPower;




##############################################################################################################################3
# given the frequencies in MHz, returns the power of the relevant comb line, in watts, at the end of the cable, without any attenuators
def sigGenPower(FC,filename='150515 Comb10MHz-Box-Shortcable-Attenuator10dB-LongerCable.csv'):
    temp = np.genfromtxt('panelgap/'+filename, comments="#", delimiter=",", dtype=float)        # read CSV
    sgpower=np.zeros(shape(FC[:]));
    for a in range(size(FC)):
        sgpower[a]=temp[(  np.abs( temp[:,0] - np.round(FC[a]/10)*10 )  ).argmin(),1];
    sgpower = 10**((sgpower-30+10)/10.0);         # convert to watts, removing the 10 dB attenuator!
    return sgpower
```
