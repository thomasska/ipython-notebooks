This file, manually captured on 11 August, was made with one digitiser connected to a signal generator (switched on halfway through the file) and a noise source (filtered for 70-300 MHz or so).

```python
# load the results from the h5 file directly
import h5py, numpy as np
#f = h5py.File('/var/kat/archive/data/ftp/failed/1439289300.h5','r')
#vis = f['Data/correlator_data'].value
#v= vis[2:,:,:,0] + 1.0j*vis[2:,:,:,1];  # discard the first few dumps in the file

# load the results saved as a numpy array
v=np.load('ar1/1439289300h5.npy')

# Notes about the file:
# Noise source attenuator: 10 dB.
# Signal generator: on and off, at channel 3000, frequency 626.953 MHz, power of -25 dBm
```

Plots of autocorrelations
------------

```python
#print('(time, frequency, baseline, real+imag)',(shape(v)))
# because of oversubscription, there are only be two types of vis - auto and cross. nothing else is possible.  Auto is baseline 5, Cross is baseline 15, among others.

figure(figsize=(12,2));
plot(10*log10(squeeze(abs(v[:,1000,5]))),color='r',label='#1000 (floor)');
plot(10*log10(squeeze(abs(v[:,1096,5]))),color='b',label='#1096 (signal generator)');
plot(10*log10(squeeze(abs(v[:,3000,5]))),color='g',label='#3000 (noise band)');
ylim([0,35]);  legend(loc=0); xlabel('time'); title('Autocorrelation power, over time, at a few channels');  ylabel('dB correlator counts');

figure(figsize=(12,3));
plot(10*log10(squeeze(abs(v[10,:,5].T))),color='b',lw=2,label='t=10');
plot(10*log10(squeeze(abs(v[60,:,5].T)))+1.0,color='g',lw=2,alpha=0.7,label='t=60 (offset by 1 dB)');
ylim([0,35]); xlim([0,4095]);  legend(loc=0)
xlabel('channel'); title('Autocorrelation power, over channel, at two times');  ylabel('dB correlator counts');
annotate('Signal Generator', xy=(1050,30),  xycoords='data',xytext=(-30, -20), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='right', verticalalignment='bottom',fontsize=10)
annotate('Noise Source, filtered', xy=(2500,20),  xycoords='data',xytext=(-30, +15), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='right', verticalalignment='bottom',fontsize=10)
#annotate('2nd harmonic of signal generator?', xy=(3960,33),  xycoords='data',xytext=(-25, -25), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='right', verticalalignment='bottom',fontsize=10)
```

These show sensible values for autocorrelation and signal generator powers. No harmonics are seen (2nd harmonic was out of band, and power was kept quite low)
