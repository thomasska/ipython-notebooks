```python

def dmcip(receptor):
    return {
        24: '10.103.254.2:7224',
        25: '10.103.254.2:7225',
        31: '10.103.254.2:7231',
        34: '10.103.254.2:7234',
        62: '10.103.254.2:7262',
        63: '10.103.254.2:7263',
        101: '192.168.4.36:7201'        
    }[receptor]
# lab DMC is 192.168.4.36.  AR1 DMC is 10.103.254.2,  RTS one is .253.2


def digitiserCommand(commandString,receptor=63,main=False):
    subIP=dmcip(receptor)
    mainIP=subIP[:-3] + '147'
    
    import commands
    if main: ip=mainIP
    else:  ip=subIP
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 100 -s " + ip + " ?" + commandString);
    if result[0]>0:
        print('Error performing digitiser command: %s' % commandString)
        print(result[1])
    return result


def getSamples(receptor):
    result=digitiserCommand("adc-snap-shot vh | grep '#adc-snap-shot'",receptor=receptor)
    if result[0]!=0: print('Failure retrieving samples');
    result=result[1]
    snapshot=result.replace('#adc-snap-shot v','').replace('#adc-snap-shot h','').replace('\n','');
    samples=np.array(  map(int, snapshot.split(' ')[1:] )  ).reshape(2,-1)
    return samples


def getManySamples(receptor=63,numSnapshots=4,samples=[]):
    print('Snapshots from receptor %i.  %i snapshots.   Getting snapshots...' % (receptor,numSnapshots))
    if size(samples)>0:
        allSamples=samples;
    else:
        allSamples=np.zeros([2,0]);
    for a in range(numSnapshots):
        #print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples(receptor);
        allSamples=np.concatenate(  ( allSamples, samples ), axis=1  )
    return allSamples

        
def plotSpectrum(allSamples,fftLength=1024,verbose=1,freqRange=[856,1712]):
    print('Plotting %i samples (%i snapshots). max=%i, min=%i, mean=%f, std=%f' % (shape(allSamples)[1],shape(allSamples)[1]/16384,np.max(allSamples),np.min(allSamples),np.mean(allSamples),np.std(allSamples)))
    
    allSamples = allSamples.reshape(2,-1,fftLength)
    
    Fsamples=fft.fft(  allSamples,  fftLength,  axis=2) 
    freq=np.linspace(856,1712,num=fftLength/2);
    
    Corr = Fsamples[0,:,:] * conj(Fsamples[1,:,:]);
    CorrFlat = np.mean(Corr,axis=0)[:fftLength/2];
    
    print('mean corr = %f' % np.mean(Corr))
    
    Auto = Fsamples[:,:,:] * conj(Fsamples[:,:,:]);
    AutoFlat = np.mean(Auto,axis=1)[:,:fftLength/2].T;
    
    print('mean auto = %f' % np.mean(Auto))
    
    flatSamples=np.concatenate((squeeze(allSamples[0,:,:]).reshape(1,-1),squeeze(allSamples[1,:,:]).reshape(1,-1)),0);
    
    figure(figsize=(16,6))
    plot(freq,10*log10(abs(AutoFlat)),alpha=0.7);
    plot(freq,10*log10(abs(CorrFlat)),alpha=0.6);
    xlim(freqRange);  grid(True);  xlabel('Frequency, MHz');  ylabel('Power, dB');  title('Auto and Cross - Magnitude in dB');
    ylim([20,100]); 
    if (verbose>1) and (shape(Auto)[1]>1):
        figure(figsize=(16,4));
        imshow(10*log10(abs(squeeze(Auto[0,:,:fftLength/2]))),extent=(856,1712,shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=40,vmax=70,cmap=plt.get_cmap('CMRmap'));
        xlim(freqRange); ylabel('Snapshot number'); title('V pol spectrogram');
        figure(figsize=(16,4));
        imshow(10*log10(abs(squeeze(Auto[1,:,:fftLength/2]))),extent=(856,1712,shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=40,vmax=70,cmap=plt.get_cmap('CMRmap'));
        xlim(freqRange);  xlabel('Frequency, MHz'); ylabel('Snapshot number'); title('H pol spectrogram');
        
    if (verbose>2) and (shape(Auto)[1]>1):
        figure(figsize=(16,4));
        imshow(10*log10(abs(squeeze(Corr[:,:fftLength/2]))),extent=(856,1712,shape(Corr)[1],0),aspect='auto',interpolation='None',vmin=40,vmax=70,cmap=plt.get_cmap('CMRmap'));
        xlim(freqRange); ylabel('Snapshot number'); title('Spectrogram of V-H correlation power');
        figure(figsize=(16,4));
        imshow(angle(squeeze(Corr[:,:fftLength/2])),extent=(856,1712,shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=-3.14,vmax=3.14,cmap=plt.get_cmap('hsv'));
        xlim(freqRange);  xlabel('Frequency, MHz'); ylabel('Snapshot number'); title('Spectrogram of V-H correlation phase');
        
        
    if verbose>3:
        figure(figsize=(16,2))
        plot(flatSamples.T,alpha=0.6); title('Voltages over time, first 700 samples'); xlim([0,700]);
        figure(figsize=(16,2));
        plot(flatSamples.T,alpha=0.6);   title('Voltages over all time');
        figure(figsize=(16,2))
        hist(flatSamples.T,np.arange(-100,100,2)); title('voltage histogram');

        




def getDigitiserOffsetTD(numSnapshots=1,globalSync=False,verbose=0,receptor=63):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync()
        time.sleep(2)

    for a in range(numSnapshots):
        if verbose>3: print('Getting snapshot %i' % a)
        samples=getSamples(receptor=receptor);
        if a==0: corr=np.zeros((shape(samples)[1]));
        corr=corr+np.correlate(samples[0,:],samples[1,:],mode='same')
    corr=abs(corr)
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,100]);
        figure(figsize=(16,3))
        hist(samples.T,np.arange(-40,40)); title('voltage histogram (first snapshot only)');
        
    if verbose>1:
        figure(figsize=(16,3))
        plot(range(-8192,8192),corr); title('Correlation magnitude');
        xlim([-200,200]); grid(True);

    offsetSamples = numpy.argmax(corr) - size(corr)/2;
    peakMag=corr[numpy.argmax(corr)];
    if verbose>0:
        print('Receptor %02i: %i snapshots. Samples offset: %4i.  Peak Ratio = %6.2f.  Voltage mean=%6.2f stdev=%5.2f max=%3i min=%3i.  AveragePower=%6.2f dB' % (receptor,numSnapshots,offsetSamples,peakMag/np.mean(corr),np.mean(samples),np.std(samples),np.max(samples),np.min(samples),10*log10(np.mean(samples**2)/261632.50) ))
    
    return offsetSamples
```
