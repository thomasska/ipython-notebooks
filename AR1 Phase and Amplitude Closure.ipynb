AR1 Phase and Amplitude Closure
========================

Load and check out the file
----------------

```python
#Load the file in question
import katdal
#D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2016/03/04/1457079587.h5')
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2016/03/04/1457089334.h5');
for n,cp in enumerate(D.corr_products):   print('BL %02i   CP %s' % (n,cp))
```

```python
# plot power over time, to help choose dumps
D.select(); D.select(freqrange=[1330e6,1440e6])# dumps=range(95,110)

figure(figsize=(16,6))
for n,cp in enumerate(D.corr_products):
    plot(  D.dumps,10*log10( np.abs(squeeze(np.mean(D.vis[:,:,n],axis=1))) ), alpha=0.5, label='%i %s%s' % (n,cp[0][2:],cp[1][2:]))
legend(loc=0); ylim([-10,40]);
title('%s %s' % (D.experiment_id,D.catalogue)); ylabel('Magnitude, dB');  xlabel('Frequency, MHz');
```

```python
# plot all baselines on top of each other
D.select(); D.select(dumps=range(30,150),freqrange=[1330e6,1440e6])

figure(figsize=(16,8))
cmap=plt.get_cmap('jet',lut=D.corr_products.shape[0])
for n,cp in enumerate(D.corr_products):
    plot(  D.channel_freqs/1e6,10*log10( np.abs(np.mean(D.vis[:,:,n],axis=0)) ), alpha=0.5, label='%i %s%s' % (n,cp[0][2:],cp[1][2:]),color=cmap(n))
legend(loc=0); xlim([D.channel_freqs[0]/1e6,D.channel_freqs[-1]/1e6]);  ylim([-10,40]);
title('%s %s' % (D.experiment_id,D.catalogue)); ylabel('Magnitude, dB');  xlabel('Frequency, MHz');

figure(figsize=(16,8))
for n,cp in enumerate(D.corr_products):
    plot(  D.channel_freqs/1e6,np.angle(np.mean(D.vis[:,:,n],axis=0)), alpha=0.5, label='%i %s%s' % (n,cp[0][2:],cp[1][2:]),color=cmap(n))
xlim([D.channel_freqs[0]/1e6,D.channel_freqs[-1]/1e6]);  #ylim([-10,40]);
title('%s %s' % (D.experiment_id,D.catalogue)); ylabel('Phase');  xlabel('Frequency, MHz');
```

```python
# plot all baselines neatly on separate graphs

D.select(); D.select(dumps=range(30,150),freqrange=[1400e6,1440e6])

fig,ax=subplots(nrows=2,ncols=D.corr_products.shape[0],figsize=(30,7))
cmap=plt.get_cmap('jet',lut=D.corr_products.shape[0])

for n,cp in enumerate(D.corr_products):
    ax[0,n].plot(  D.channel_freqs/1e6,10*log10( np.abs(np.mean(D.vis[:,:,n],axis=0)) ), lw=2, color=cmap(n))
    ax[0,n].set_title('%i %s%s' % (n,cp[0][2:],cp[1][2:]));
    ax[0,n].set_xlim([D.channel_freqs[0]/1e6,D.channel_freqs[-1]/1e6]);  ax[0,n].set_ylim([-10,40]);

    ax[1,n].plot(  D.channel_freqs/1e6,np.angle(np.mean(D.vis[:,:,n],axis=0)), lw=2, color=cmap(n))
    ax[1,n].set_xlim([D.channel_freqs[0]/1e6,D.channel_freqs[-1]/1e6]);  ax[1,n].set_ylim([-3.5,3.5]);
        
    if n==0:
        ax[0,n].set_ylabel('Magnitude, dB');
        ax[1,n].set_ylabel('Phase, rad');   
    
print('Description   : %s' % D.description)
print('Experiment ID : %s' % D.experiment_id)
print('Catalogue     : %s' % D.catalogue)
print('Start time    : %s' % D.start_time.local())
```

Phase closure
----------------

```python
#baselines=[16,17,18];
baselines=[18,19,20];
#baselines=[17,20,21];

D.select(); D.select(dumps=range(30,150),freqrange=[1320e6,1460e6],corrprods=baselines)

phc = squeeze( np.angle( D.vis[:,:,0] * D.vis[:,:,2] / D.vis[:,:,1] ) );  # depends on the order of the baselines...

figure(figsize=(12,5))
plot(D.channel_freqs/1e6,phc[:20,:].T,alpha=0.1,lw=2,color='b')
title('Phase of %s, all readings, over frequency' % D.corr_products.flatten() )

figure(figsize=(12,5))
hist(reshape(phc,[-1]),bins=np.arange(-2,2,0.02),histtype='stepfilled',lw=0);
title('Phase closure of %s ' % D.corr_products.flatten() )
```
