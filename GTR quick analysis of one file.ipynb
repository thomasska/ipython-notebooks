Quick plotting of some results from site 1 June
===========================


```python
# load the file
S=int(numpy.genfromtxt('SiteGTR_noise_m063_test1-V.npy.timestamp'))              # the timestamp
D=numpy.load('SiteGTR_noise_m063_test1-V.npy', mmap_mode='r')                    # the data
samplerate = 1712e6
start=samplerate-S%samplerate;
```

```python
# quick look at the data
print 'First 10 numbers in the file ' , D[:10]
print 'First 10 powers in the file ' , D[:10]**2
print 'Lumped mean power of groups of 10  ' , (D[:100]**2).reshape(10,-1).mean(axis=0)
print 'Average power in the first 1000 samples  ', (D[:1000]**2).mean()
```

```python
# Plot some samples, near the event
plotlength = 50e3            # samples

figure(figsize=(5,4))
timevector=np.linspace(0, plotlength/samplerate*1e6, num=plotlength)
plot(timevector, D[(start):(start+plotlength)])
xlabel('Time after true 1PPS, us');    ylabel('Sample voltage');  grid(True);   #xlim([12,16]);
```

```python
# average the power in a bunch of samples, plot it over time.  Plot this for each event.

averagesize = 200      # samples, ~= 100 ns
plotlength = 50e3            # samples

figure(figsize=(16,6))

# make a time vector
T = np.linspace(0, plotlength/samplerate*1e6, num=plotlength/averagesize)       # time vector in microseconds

# make a total power vector
totalpower=T*0.0;

for pulsenum in range(4):                # plot each pulse seen
    selectedsamples=D[(pulsenum*samplerate+start):(pulsenum*samplerate+start+plotlength)]                # choose a few samples to save RAM
    Pow = (selectedsamples**2)                                                                           # square to find power
    Pow=Pow.reshape(-1,averagesize)                                                                      # reshape to make a shorter matrix with (averagesize) columns
    Pow=Pow.mean(axis=1)                                                                                 # average every column
    totalpower = totalpower + Pow
    plot(T,10*log10(Pow))

plot(T,10*log10(totalpower/4),lw=2,color='k')

xlabel('Time after true 1PPS, us');    ylabel('Power, dB');  grid(True);   xlim([12,16]);
```
