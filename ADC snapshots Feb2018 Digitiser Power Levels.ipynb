```python
import time

def dmcip(receptor):
    if receptor==101: return '192.168.4.36:7201'
    if receptor==90: return '192.168.4.36:7290'
    if receptor==17: return '10.103.253.2:72%02i' % receptor
    if receptor==18: return '10.103.253.2:72%02i' % receptor
    return '10.103.254.2:72%02i' % receptor
# lab DMC is 192.168.4.36.  AR1 DMC is 10.103.254.2,  RTS one is .253.2


def digitiserCommand(commandString,receptor=63,main=False):
    subIP=dmcip(receptor)
    mainIP=subIP[:-3] + '147'
    
    import commands
    if main: ip=mainIP
    else:  ip=subIP
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 100 -s " + ip + " ?" + commandString);
    if result[0]>0:
        print('Error performing digitiser command: %s' % commandString)
        print(result[1])
    return result


def getSamples(receptor):
    result=digitiserCommand("adc-snap-shot vh | grep '#adc-snap-shot'",receptor=receptor)
    if result[0]!=0: print('Failure retrieving samples');
    result=result[1]
    snapshot=result.replace('#adc-snap-shot v','').replace('#adc-snap-shot h','').replace('\n','');
    samples=np.array(  map(int, snapshot.split(' ')[1:] )  ).reshape(2,-1)
    return samples


def getManySamples(receptor=63,numSnapshots=4,samples=[]):
    #print('Snapshots from receptor %i.  %i snapshots.   Getting snapshots...' % (receptor,numSnapshots))
    if size(samples)>0:
        allSamples=samples;
    else:
        allSamples=np.zeros([2,0]);
    for a in range(numSnapshots):
        #print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples(receptor);
        allSamples=np.concatenate(  ( allSamples, samples ), axis=1  )
    return allSamples

        
def plotSpectrum(allSamples,fftLength=1024,verbose=1,freqRange=[856,1712],headingtext='',band='l',yRange=[40,80]):
    # if you request a frequency below 856, it will flip the band for you, assuming it's a UHF digitiser
    if verbose>1:
        print('Plotting %i samples (%i snapshots). max=%i, min=%i, mean=%f, std=%f' % (shape(allSamples)[1],shape(allSamples)[1]/16384,np.max(allSamples),np.min(allSamples),np.mean(allSamples),np.std(allSamples)))
    
    allSamples = allSamples.reshape(2,-1,fftLength)
    
    Fsamples=fft.fft(  allSamples,  fftLength,  axis=2) 
    
    if band=='l':  freq=np.linspace(856,1712,num=fftLength/2)
    if band=='u':  freq=np.linspace(544,1088,num=fftLength/2)
    
    #if freqRange[0]<856:                                    # if it's a UHF measurement,
    #    freq=np.linspace(0,856,num=fftLength/2)             # new axis, not flipped
    #    Fsamples=np.copy(Fsamples[:,:,fftLength/2::-1])                  # flip the data instead
    
    Corr = Fsamples[0,:,:] * conj(Fsamples[1,:,:]);
    CorrFlat = np.mean(Corr,axis=0)[:fftLength/2];
    
    if verbose>1:
        print('mean corr = %f' % np.mean(Corr))
    
    Auto = Fsamples[:,:,:] * conj(Fsamples[:,:,:]);
    AutoFlat = np.mean(Auto,axis=1)[:,:fftLength/2].T;
    
    if verbose>1:
        print('mean auto = %f' % np.mean(Auto))
    
    flatSamples=np.concatenate((squeeze(allSamples[0,:,:]).reshape(1,-1),squeeze(allSamples[1,:,:]).reshape(1,-1)),0);
    
    if verbose>0:
        figure(figsize=(16,4))
        plot(freq,10*log10(abs(AutoFlat)),alpha=0.7);
        plot(freq,10*log10(abs(CorrFlat)),alpha=0.6);
        grid(True);  xlabel('Frequency, MHz');  ylabel('Power, dB');  title('%s  Auto and Cross - Magnitude in dB' % headingtext);
        xlim(freqRange);
        ylim(yRange); 
    if (verbose>2) and (shape(Auto)[1]>1):
        figure(figsize=(16,4));
        if shape(D)[1]/fftLength<fftLength*2:
            imshow(10*log10(abs(squeeze(Auto[0,:,:fftLength/2]))),extent=(freq[0],freq[-1],shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            xlim(freqRange); ylabel('Snapshot number'); xlabel('Frequency, MHz');
        else:
            imshow(10*log10(abs(squeeze(Auto[0,:,:fftLength/2]))).T,extent=(shape(Auto)[1],0,freq[0],freq[-1]),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            ylim(freqRange); xlabel('Snapshot number'); ylabel('Frequency, MHz');
        title('V pol spectrogram');
        figure(figsize=(16,4));
        if shape(D)[1]/fftLength<fftLength*2:
            imshow(10*log10(abs(squeeze(Auto[1,:,:fftLength/2]))),extent=(freq[0],freq[-1],shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            xlim(freqRange);  xlabel('Frequency, MHz'); ylabel('Snapshot number');
        else:
            imshow(10*log10(abs(squeeze(Auto[1,:,:fftLength/2]))).T,extent=(shape(Auto)[1],0,freq[0],freq[-1]),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            ylim(freqRange);  ylabel('Frequency, MHz'); xlabel('Snapshot number');
        title('H pol spectrogram');
        
    if (verbose>3) and (shape(Auto)[1]>1):
        figure(figsize=(16,4));
        if shape(D)[1]/fftLength<fftLength*2:
            imshow(10*log10(abs(squeeze(Corr[:,:fftLength/2]))),extent=(freq[0],freq[-1],shape(Corr)[1],0),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            xlim(freqRange); ylabel('Snapshot number');
        else:
            imshow(10*log10(abs(squeeze(Corr[:,:fftLength/2]))).T,extent=(shape(Corr)[1],0,freq[0],freq[-1]),aspect='auto',interpolation='None',vmin=20,vmax=60,cmap=plt.get_cmap('CMRmap'));
            ylim(freqRange); xlabel('Snapshot number'); ylabel('Frequency, MHz'); 
        title('Spectrogram of V-H correlation power');
        figure(figsize=(16,4));
        if shape(D)[1]/fftLength<fftLength*2:
            imshow(angle(squeeze(Corr[:,:fftLength/2])),extent=(freq[0],freq[-1],shape(Auto)[1],0),aspect='auto',interpolation='None',vmin=-3.14,vmax=3.14,cmap=plt.get_cmap('hsv'));
            xlim(freqRange);  xlabel('Frequency, MHz'); ylabel('Snapshot number');
        else:
            imshow(angle(squeeze(Corr[:,:fftLength/2])).T,extent=(shape(Auto)[1],0,freq[0],freq[-1]),aspect='auto',interpolation='None',vmin=-3.14,vmax=3.14,cmap=plt.get_cmap('hsv'));
            ylim(freqRange);  ylabel('Frequency, MHz'); xlabel('Snapshot number');
        title('Spectrogram of V-H correlation phase');
        
        
    if verbose>4:
        figure(figsize=(16,3))
        plot(flatSamples.T,alpha=0.6); title('Voltages over time, first 700 samples'); xlim([0,700]);
        figure(figsize=(16,3));
        plot(flatSamples.T,alpha=0.6);   title('Voltages over all time');
        #figure(figsize=(16,2))
        #hist(flatSamples.T,np.arange(-100,100,2)); title('voltage histogram');

    if (verbose>5):
        figure(figsize=(16,3));
        plot(np.arange(0,np.shape(flatSamples)[1])[::128],10*log10(np.sum(flatSamples.reshape([2,-1,128])**2,axis=2).T),alpha=0.6);
        title('Signal power (128-sample bins) over all time. dB');



def plotHistogram(D,receptor,headingtext=''):
    # do the +-1 flipping
    D=D * (2*(np.arange(np.shape(D)[1])%2)-1)
    
    s=digitiserCommand('sensor-value l-band.rfcu.hpol.attenuation',receptor=receptor)[1]
    p=str.find(s,'nominal')
    q=str.find(s,'\n',p)
    ha=int(s[p+7:q])
    
    s=digitiserCommand('sensor-value l-band.rfcu.vpol.attenuation',receptor=receptor)[1]
    p=str.find(s,'nominal')
    q=str.find(s,'\n',p)
    va=int(s[p+7:q])
    
    s=digitiserCommand('sensor-value l-band.adc.hpol.rf.power.in',receptor=receptor)[1]
    p=str.find(s,'nominal')
    if p==-1: p=str.find(s,'warn')
    if p==-1: p=str.find(s,'error')
    q=str.find(s,' ',p)
    r=str.find(s,'\n',q)
    hp=float(s[q:r])
    
    s=digitiserCommand('sensor-value l-band.adc.vpol.rf.power.in',receptor=receptor)[1]
    p=str.find(s,'nominal')
    if p==-1: p=str.find(s,'warn')
    if p==-1: p=str.find(s,'error')
    q=str.find(s,' ',p)
    r=str.find(s,'\n',q)
    vp=float(s[q:r])
    
    vstdpower=20*log10(np.std(D[0,:]))-54;
    hstdpower=20*log10(np.std(D[1,:]))-54;
    
    figure(figsize=(16,4))
    x=histogram(D[0,:],bins=np.arange(-512,512)); 
    step(x[1][:-1],x[0],alpha=0.6,label='V, outliers=%i, att=%i dB, std=%.1f=%.1f dBFS,  ADCpwr=%.1f dBm, err=%.1f dB' % (np.sum(x[0][:(512-64)])+np.sum(x[0][(512+64):]),va,np.std(D[0,:]),vstdpower,vp,vp-vstdpower))
    
    x=histogram(D[1,:],bins=np.arange(-512,512));
    step(x[1][:-1],x[0],alpha=0.6,label='H, outliers=%i, att=%i dB, std=%.1f=%.1f dBFS,  ADCpwr=%.1f dBm, err=%.1f dB' % (np.sum(x[0][:(512-64)])+np.sum(x[0][(512+64):]),ha,np.std(D[1,:]),hstdpower,hp,hp-hstdpower))
    xlim([-64,64])
    legend(loc=0,fontsize=10)
    
    grid(True);  xlabel('Voltage bin');  ylabel('Frequency');  title('%s  Histogram' % headingtext);
    
    
def getDigitiserOffsetTD(numSnapshots=1,globalSync=False,verbose=0,receptor=63):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync()
        time.sleep(2)

    for a in range(numSnapshots):
        if verbose>3: print('Getting snapshot %i' % a)
        samples=getSamples(receptor=receptor);
        if a==0: corr=np.zeros((shape(samples)[1]));
        corr=corr+np.correlate(samples[0,:],samples[1,:],mode='same')
    corr=abs(corr)
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,100]);
        figure(figsize=(16,3))
        hist(samples.T,np.arange(-40,40)); title('voltage histogram (first snapshot only)');
        
    if verbose>1:
        figure(figsize=(16,3))
        plot(range(-8192,8192),corr); title('Correlation magnitude');
        xlim([-200,200]); grid(True);

    offsetSamples = numpy.argmax(corr) - size(corr)/2;
    peakMag=corr[numpy.argmax(corr)];
    if verbose>0:
        print('Receptor %02i: %i snapshots. Samples offset: %4i.  Peak Ratio = %6.2f.  Voltage mean=%6.2f stdev=%5.2f max=%3i min=%3i.  AveragePower=%6.2f dB' % (receptor,numSnapshots,offsetSamples,peakMag/np.mean(corr),np.mean(samples),np.std(samples),np.max(samples),np.min(samples),10*log10(np.mean(samples**2)/261632.50) ))
    
    return offsetSamples
```
