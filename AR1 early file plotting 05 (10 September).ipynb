Made with one digitiser connected to:

* Noise source. (filtered for 70-300 MHz or so).  Attenuator: 10 dB.
* Signal generator, controlled by the auto-random-channel script.  Channel range (500,2200). Power range (-53,-5 dBm). dwell time 4 s


```python
import h5py, numpy as np
try:
    #f = h5py.File('ar1/1441902063.h5','r')  # 25 seconds, Friday 11 Sept
    f = h5py.File('ar1/1442234889.h5','r')  # 15 minutes, Monday the 14th
except Exception as err_msg: raise SystemExit('An error has occured:\n%s' % err_msg)
vis = f['Data/correlator_data'].value
v = vis[2:,:,:,0] + 1.0j*vis[2:,:,:,1];  # discard the first few dumps in the file
```

```python
# my random-channel decoder
import random

chRange=(500,2200)    # channel ranges from 500 to 2200. Integer, half-open. NB, this affects the (random) ordering of the list.
pwrRange=(-53,-5)     # power ranges from   -5 to -53 dBm. 0.1 dB rounding.

# map  [0,N) to [start,end] in a nice non-repeating way.
def pickchannel(testnumber,channelstart,channelend):
    numchannels = channelend-channelstart;
    testnumber = testnumber%numchannels          # crop the test numbe so it works with an indefinite list of tests

    random.seed(42);
    channellist = random.sample(range(channelstart,channelend),numchannels);

    return channellist[testnumber]             #   return the indexed item

# map the channel number to the signal generator power.  Using seed to ensure this is always repeatable.
def findpower(channel,powerstart,powerend):
    random.seed(channel)
    return random.randrange(powerstart*10,powerend*10)/10.0



print(' #    Ch  Pwr');
for timestep in range(5):
    ch=pickchannel(timestep,chRange[0],chRange[1]);                # deterministically choose a channel from the range chosen
    pwr=findpower(ch,pwrRange[0],pwrRange[1]);                    # calculate the power from the channel number

    print('%3i  %3i %4.1f' % (timestep,ch,pwr))
# Step 164, Channel= 1820,  Freq= 475.65 MHz (1236.35 MHz),  Power= -45.9 dBm       
```

Plot of power against signal generator power - system linearity
------------------

```python
# first, generate the list of signal generator frequencies and powers, to correspond to the data

t=222;   # which time to plot. Even numbers only please

timestep=1292+(t-2)/2;
ch=pickchannel(timestep,chRange[0],chRange[1]);                # deterministically choose a channel from the range chosen
pwr=findpower(ch,pwrRange[0],pwrRange[1]);                    # calculate the power from the channel number

figure(figsize=(12,3));
plot(10*log10(squeeze(abs(v[:,ch,0]))),color='r',label='ch %i, siggenpower=%4.1f' % (ch,pwr));
ylim([0,40]); xlim([t-3,t+3]);  legend(loc=1)
xlabel('time'); title('Autocorrelation power, over time, just to check the channel-finder algorithm');  ylabel('dB correlator counts');  grid(True);
print('Time %3i.  Ch %3i, power %4.1f. Corr Power over BG: %4.2f dB' % ( t, ch, pwr, 10*log10(  squeeze(  abs(v[t,ch,0])-abs(v[t+3,ch,0])) )  )  ) 
```

```python
# now process the whole file, to a list   Channel, SigGen, CorrelatorPowerExcess
chlist=[];  siglist=[];  corrlist=[];
for a in range(2,250,2):
    timestep=1292+(a-2)/2;
    ch=pickchannel(timestep,chRange[0],chRange[1]);                # deterministically choose a channel from the range chosen
    pwr=findpower(ch,pwrRange[0],pwrRange[1]);                    # calculate the power from the channel number
    chlist.extend([ch]);
    siglist.extend([pwr]);
    corrlist.extend([10*log10(  squeeze(  abs(v[a,ch,0])-abs(v[a+3,ch,0])) )]);
    
figure(figsize=(12,8))
plot(siglist,corrlist,lw=0,marker='o');
grid(True);  ylim([-5,45]); xlim([-55,-3]);
xlabel('signal generator power'); title('Correlator+digitiser linearity');  ylabel('Correlator power over noise, dB arbitrary');  grid(True);
```

Comparison of all the baselines

A few graphs comparing all 40 baselines.

Problems seen:

1. Cross correlation baseline 0 shows some strange spikes
1. The imaginary part of the autocorrelation power is not zero - in some baselines only


```python
# All of the baselines
t=60;
bls=range(40);

figure(figsize=(12,5));
for a in bls:
   if abs(v[60,1100,a])>1:
        if abs(v[60,1100,a])>2:
            plot(10*log10(squeeze(abs(v[t,:,a].T))),lw=1,color='b',label='bl%i' % a);
        else:
            plot(10*log10(squeeze(abs(v[t,:,a].T))),lw=1,color='r',label='bl%i' % a);
ylim([0,35]); xlim([0,4095]); legend(loc=0)
xlabel('channel'); title('Autocorrelation power, all baselines. Shows the two kinds of autocorrelation baseline.');  ylabel('dB correlator counts');

figure(figsize=(12,5));
for a in bls:
   if abs(v[60,1100,a])<=1:
        if a==0:
          plot(10*log10(squeeze(abs(v[t,:,a].T))),lw=1,color='red',label='bl%i' % a);
        else:
          plot(10*log10(squeeze(abs(v[t,:,a].T))),lw=1,color='green',label='bl%i' % a);
ylim([-20,30]); xlim([0,4095]); legend(loc=0)
xlabel('channel'); title('Cross-correlation magnitude, all baselines. Shows the spikes in one baseline.');  ylabel('dB correlator counts');


figure(figsize=(12,5));
for a in bls:
   if abs(v[60,1100,a])>=1:
        if sum(abs(squeeze(angle(v[t,:,a].T))))>1e-5:
            plot(1e3*squeeze(angle(v[t,:,a].T)),lw=1,alpha=0.5,color='r',label='bl%i' % a);
        else:
            plot(1e3*squeeze(angle(v[t,:,a].T)),lw=2,alpha=0.5,color='b',label='bl%i' % a);
        
ylim([-0.4,0.4]); xlim([0,4095]); legend(loc=2)
xlabel('channel'); title('Auto-correlation power, phase angle. Shows the non-zero imaginary part.');  ylabel('angle, milliradians');
```

Noise statistics

```python
figure(figsize=(12,5));
nch=range(2200,3200);
D=abs(v[:,nch,5]/mean(v[:,nch,5],axis=0));
E=reshape(D,product(shape(D)));
hist(E,200,range=[0.941,1.059],log=True);
grid(True)
title('Histogram of normalised powers of channels 2200 to 3199.  (%i samples)  Stdev = %f' % (product(shape(D)),std(E)))

print('standard deviation of %i points is %6.4f' % (product(shape(D)),std(E)) );
n=10.0;
print('%i points, or %6.4f percent, exceed %i standard deviations!' % (sum(E>(1+n*std(E))), sum(E>(1+n*std(E)))/(1.0*product(shape(D)))*100.0, n))
```
