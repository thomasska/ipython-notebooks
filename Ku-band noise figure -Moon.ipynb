### Spectral plots, showing H and V polarisation is mostly maintained through/around the dish

```python
import matplotlib.pyplot as plt
from pylab import *

# plot spectra of all the data, just to see
fig, axes = plt.subplots(nrows=3, ncols=1,figsize=(14,10));

Yh=mean(real(ds_moonr.vis[345:355,:,0]),axis=0)/mean(real(ds_csp.vis[30:110,:,0]),axis=0)
Yv=mean(real(ds_moonr.vis[345:355,:,1]),axis=0)/mean(real(ds_csp.vis[30:110,:,1]),axis=0)

#Tsrc=190 #Kraus Fig.8-41 @ 3.15cm/9.5GHz, also Fig9  Marina Battaglia: Moon disc temperature as a function of moon phase (phase 73% and waxing)
#Thot=192 # Christian Monstein @ 2.77cm or 13.2GHz
#Tsrc=213 #Hey , p80 Radio straling uit het Heelal
Tsrc=215

#L-band example from SEFD.xlsx spreadsheet
#Tsys = Tcmb (2.73) + Tgal (4.1) + Tatm (2.2)  + Tspill (6.8) + Tspill_noshroud (1-3) +  Treflector(ohmic and panelgaps 1) + Trx (90???)
# Ku-band
Tcmb=2.73

Tgal=0
Tatm=7.02 #7.02 from RA calculator for centre F and ~ 32 degree elevation for this observation
Tspill=2 # from M0000-000V1-11 TM at Ku
Trefl=1.5

Thot= (Tsrc+    +Tatm+Tspill+Trefl) # per polarisation /2
Tcold=(     Tcmb+Tatm+Tspill+Trefl) #

Teh1=(Thot-Yh*Tcold)/(Yh-1) #h; 
Tev1=(Thot-Yv*Tcold)/(Yv-1) #v
print 'Thot:', Thot, 'Tcold:', Tcold

#To tidy BM input
#Thotra= (Tsrc+          Tatm)/2 #Tsrc+Tatm?
#Tcoldra=(     Tcmb+Tgal+Tatm)/2 #Tcmb +Tgal +Tatm
#Tsysh=(Thotra-Tcoldra)/(Yh-1)
#Tsysv=(Thotra-Tcoldra)/(Yv-1)
#Tsys=Tcold+Trx  , Trx = Tsys - Tcold
#Trx=Tspill+Trefl + Trxe + Tdig

#h5 vis order is t, f, pol

axes[0].plot(ds_moonr.channel_freqs/1e6,10*log10(mean(real(ds_moonr.vis[345:355,:,0]),axis=0)),lw=2,alpha=1.0,color='red',label=ds_moonr.description +' h');
axes[0].plot(ds_moonr.channel_freqs/1e6,10*log10(mean(real(ds_moonr.vis[345:355,:,1]),axis=0)),lw=2,alpha=1.0,color='green',label=ds_moonr.description+' v');
axes[0].plot(ds_csp.channel_freqs/1e6,10*log10(mean(real(ds_csp.vis[30:110,:,0]),axis=0)),lw=2,alpha=1.0,color='blue',label=ds_csp.description+' h');
axes[0].plot(ds_csp.channel_freqs/1e6,10*log10(mean(real(ds_csp.vis[30:110,:,1]),axis=0)),lw=2,alpha=1.0,color='black',label=ds_csp.description+' v');

axes[0].set_xlabel(' IF Frequency [MHz] ');
axes[0].set_ylabel('Mean power across scan [dB]');
axes[0].legend(loc=1);
axes[0].set_title('Tracking moon and CSP for NF measurement ');
axes[0].set_xlim([857,1700]); 
axes[0].set_ylim([15, 30]);
axes[0].grid();

axes[1].plot(ds_csp.channel_freqs/1e6,Yh,lw=2,alpha=1.0,color='red',label='h');
axes[1].plot(ds_csp.channel_freqs/1e6,Yv,lw=2,alpha=1.0,color='green',label='v');
axes[1].legend(loc=1);
axes[1].set_xlabel(' IF Frequency [MHz] ');
axes[1].set_ylabel('Y-factor');
axes[1].set_xlim([857,1700]); 
#axes[1].set_xlim([860,1650]); #axes[1].set_ylim([0, 50]);
axes[1].grid();

axes[2].plot(ds_moonr.channel_freqs/1e6+Ku_lo ,Teh1, lw=1,alpha=0.7,color='red',label='h');
#axes[2].plot(ds_moon.channel_freqs/1e6+Ku_lo ,Tsysh-Tcoldra-Tspill-Trefl,lw=2,alpha=0.5,color='m',ls='--',label='h');
axes[2].plot(ds_moonr.channel_freqs/1e6+Ku_lo ,Tev1, lw=1,alpha=0.7,color='green',label='v');
#axes[2].plot(ds_moon.channel_freqs/1e6+Ku_lo ,Tsysv-Tcoldra-Tspill-Trefl,lw=2,alpha=0.5,color='c',ls='--',label='v');
axes[2].legend(loc=1);
axes[2].set_xlabel('IF Frequency [MHz] ');
axes[2].set_ylabel('Rough Ku-RX noise figure [K]');
axes[2].set_xlim([Ku_lo+900,Ku_lo+1670]); #axes[2].set_xlim([11000,14000]); 
axes[2].set_ylim([50, 200]);
axes[2].grid();
fig.tight_layout();
```
