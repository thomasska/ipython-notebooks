## Initial experimental analysis

```python
# LOAD THE DATA AND IMPORT MODULES
# preliminary stuff
import katdal
import csv
import numpy as np
import matplotlib
import matplotlib.gridspec as gridspec

# load the measurement files
dsh=katdal.open('/var/kat/archive/data/RTS/telescope_products/2014/12/03/1417613297.h5');
dsv=katdal.open('/var/kat/archive/data/RTS/telescope_products/2014/12/03/1417612515.h5');

# find the channels which would contain the integer multiples of 10 MHz
combchannels = [i for i, x in enumerate(abs((dsh.channel_freqs+dsh.channel_width/2) % 10e6)-dsh.channel_width/2) if x < dsh.channel_width/2];
combfreqs = dsh.channel_freqs[combchannels];

# load the CSV of comb output powers and interpolate
attenuator = 13;    # actual attenuator used in the experiments
comb_power = np.genfromtxt('panelgap/comb10MHz.csv', comments="#", delimiter=",", dtype=float)        # read CSV
idx1 = (np.abs(comb_power[:,0]-dsh.channel_freqs[0]/1e6)).argmin()                           # find index of start and stop array frequencies
idx2 = (np.abs(comb_power[:,0]-dsh.channel_freqs[(shape(dsh.channel_freqs)[0]-1)]/1e6)).argmin()
comb_power = comb_power[idx1:(idx2+1),1];              # throw away frequencies we don't need, and keep only the dBm values
comb_power = 10**((comb_power-30 - attenuator)/10);    # convert to W

print 'Loaded measurement data'

```

### Finding the absolute power - assuming noise floor is 20 Kelvin

Finds a smooth noise floor by taking elevation minimum of non-comb channels and smoothing.
Creates new autocorrelation matrices that are in W and not arbitrary units.

```python
# create a full-size visibility matrix with no comb channels (use neighbour to the left)
nocombh = squeeze(np.real(dsh.vis));
nocombv = squeeze(np.real(dsv.vis));
for i in combchannels:    # cloning of the neighbours
    nocombh[:,i,:] = 1.0*nocombh[:,i-1,:];
    nocombv[:,i,:] = 1.0*nocombv[:,i-1,:];

# find the noise floor using min over time - which mostly eliminates RFI etc
noisefloorh = squeeze(np.amin(  np.add( nocombh[:,:,0], nocombv[:,:,0] )/2,  axis=0 ));
noisefloorv = squeeze(np.amin(  np.add( nocombh[:,:,1], nocombv[:,:,1] )/2,  axis=0 ));

# filter this over frequency
N=300;  W = numpy.hamming(N);  WS = sum(W);
snfh = convolve(W,noisefloorh, 'same') / WS;
snfv = convolve(W,noisefloorv, 'same') / WS;

# how much power at 20 K in each channel bandwidth?
channelnoisepower = 1.38e-23 * 20 * dsh.channel_width;
print 'Noise power in a %i Hz channel at 20 K is %6.3g W' % (dsh.channel_width,channelnoisepower)

# multiply this by the smoothed noise power.  AbsPowerSignal = LevelSignal / Level20K * AbsPower20K
vish = (squeeze(np.real(dsh.vis[:,:,0])) + squeeze(np.real(dsh.vis[:,:,1]))) * channelnoisepower / snfh;
visv = (squeeze(np.real(dsv.vis[:,:,0])) + squeeze(np.real(dsv.vis[:,:,1]))) * channelnoisepower / snfh;

print shape(vish)
print 'Average power in a channel is %6.3g W, in a comb channel (1598) is %6.3g W' % (mean(mean(vish)),mean(vish[:,1598]));
print shape(np.real(dsh.vis[:,:,0]))
print shape(vish)
print 'Total power in both channels is %6.3g W' % (mean(sum(vish))/2);
print 'Created vish and visv with shape ',shape(vish),' which now contain the absolute power in W';


# plot, for interest
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(14,5));
axes.plot(dsh.channel_freqs/1e6,squeeze(mean(real(dsh.vis[0:199,:,0]),axis=0)), label='HP, mean power',alpha=0.3,lw=1);
axes.plot(dsh.channel_freqs/1e6,noisefloorh, label='HP',alpha=0.7);
axes.plot(dsh.channel_freqs/1e6,snfh, label='smoothed',alpha=0.7,lw=3);
#axes.plot(noisefloor, label='Smoothed average, assume 20 K');
axes.set_xlim([870, 1650]);  axes.set_ylim([0, 220]);
axes.legend(loc=8);  axes.set_title('Noise floor power over frequency, showing derivation of the power figures');
```

```python
# find pathloss from comb generator to antenna
pathlossh = (vish[:,combchannels] - vish[:,[i-1 for i in combchannels]])  / comb_power;     # first remove neighbours on the low side, then divide by the known comb generator power output
pathlossv = (visv[:,combchannels] - visv[:,[i-1 for i in combchannels]])  / comb_power;     # first remove neighbours on the low side, then divide by the known comb generator power output

# heat plot of signal variation?
from scipy.stats import gaussian_kde
y=10*log10(pathlossh);
y[numpy.isnan(y)] = -111;
y=y / 1000;   # bring the Y points closer together, so the adjacent frequencies have limited effect
y=reshape(y,shape(pathlossh)[0]*shape(pathlossh)[1],'C');   #   C or F changes the shape
x=np.repeat(np.linspace(combfreqs[0]/1e6,combfreqs[85]/1e6,num=86),shape(pathlossh)[0]);
xy = np.vstack([x,y]);    z = gaussian_kde(xy)(xy);  # Calculate the point density
y = y * 1000;     # undo the Y compression
idx = z.argsort();   x, y, z = x[idx], y[idx], z[idx];  # Sort the points by density, so that the densest points are plotted last

# plot the heat density plot
fig, ax = plt.subplots(figsize=(14,6))
ax.scatter(x, y, c=z, s=50, edgecolor='', cmap=plt.get_cmap('CMRmap'))
ax.set_xlim([860,1650]); ax.set_ylim([-105,-70]);
ax.set_xlabel('Frequency, MHz');  ax.set_ylabel('Pathloss, dB');  ax.set_title('Pathloss from comb generator connector to receiver, HP');
plt.show()

# Repeat for Vertical Polarisation
y=10*log10(pathlossv);
y[numpy.isnan(y)] = -111;
y=y / 1000;   # bring the Y points closer together, so the adjacent frequencies have limited effect
y=reshape(y,shape(pathlossh)[0]*shape(pathlossh)[1],'C');   #   C or F changes the shape
xy = np.vstack([x,y]);    z = gaussian_kde(xy)(xy);  # Calculate the point density
y = y * 1000;     # undo the Y compression
idx = z.argsort();   x, y, z = x[idx], y[idx], z[idx];  # Sort the points by density, so that the densest points are plotted last

# plot the heat density plot
fig2, ax = plt.subplots(figsize=(14,6))
ax.scatter(x, y, c=z, s=50, edgecolor='', cmap=plt.get_cmap('CMRmap'))
ax.set_xlim([860,1650]); ax.set_ylim([-105,-70]);
ax.set_xlabel('Frequency, MHz');  ax.set_ylabel('Pathloss, dB');  ax.set_title('Pathloss from comb generator connector to receiver, VP');
plt.show()
```

**Conclusion here:** There is no ripple effect. The path loss is somewhat spread out, but the typical range is from 80 to 90 dB, and this doesn't change over frequency.  
The ripple seen in the graphs is an artefact, caused by smoothing a large range of data. The linear-mean average used above causes the line to be dominated by a few outliers at the top end, which isn't useful or fair.

I'm surprised that the actual level is so low - that the coupling is generally around -85 dB. This indicates excellent shielding by the dish, more than expected.
