Looking at the data afterwards a strong wide band noise source was seen  This notebook analyses this source and investigates possible causes.

```python
import katdal, csv, numpy as np, matplotlib, matplotlib.gridspec as gridspec, string, scipy
D1=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/05/27/1432735967.h5');   # 360 scan from 100 m away, VP
D2=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/05/27/1432736479.h5');   # 360 scan from 100 m away, HP

F=1712-D1.channel_freqs/1e6;   # Create a flipped frequency axis for plotting purposes

# create a cropped frequency axis being only those allowed through the horn and the filters.  sanechannels  and  FS  the corresponding MHz frequency vector
freqrange=[400,850];
sanechannels = [i for i, v in enumerate(F) if ((v<=freqrange[1])&(v>=freqrange[0])) ];
FS = F[sanechannels];

# find the channels which would contain the integer multiples of 10 MHz
combchannels = [i for i, x in enumerate(abs((F*1e6+D1.channel_width/2) % 10e6)-D1.channel_width/2) if (x < D1.channel_width/2)];
# remove combchannels which are outside the chosen frequency range, not pretty but it works
temp = combchannels[:];
for a in temp:
    if (not(a in sanechannels)):
        combchannels.remove(a);
FC = F[combchannels];

# create a frequency axis which does not contain the comb channels:   sanechannelsnocomb  and  FSNC the corresponding MHz frequency vector
sanechannelsnocomb = sanechannels[:];
temp = sanechannelsnocomb[:];
for a in temp:
    if (a in combchannels):
        sanechannelsnocomb.remove(a);
#sanechannelsnocomb.remove(0);   # remove the DC
FSNC = F[sanechannelsnocomb];
print('Loaded %s and %s OK' % (D1.name,D2.name));
```

The left-hand graphs have been shifted in amplitude, to have their noise floor figures lie on top of each other, at 0 dB. Amplitude units are arbitrary.  
The right hand graph is a (dB) ratio of signal+noise power to noise power, at the peak angle in each case.

Observations:

1. This is a truly broadband signal, not a trace of features!
1. The second pass, about 10 minutes after the first, shows a much smaller noise source, and the peak is reduced in amplitude.
1. Some sidelobes are seen on the second peak, indicating that the source was scanned away from boresight.



Beamwidth and Angle of peak
-------------

The graphs below show beamwidth and angle of peak signal, over frequency.  Peak signal is found as the centre of the half-power points, as the second pass shows some double-peaking which makes the actual peak indistinct.
The beamwidth decreases with increasing frequency, as expected.  
It appears that in the first pass, the source of interest passed directly through the beam, through the main lobe. In the second pass, it moved through the beam, cutting the sidelobe slightly.



```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,6));

# find the beamwidth and peak angle, over frequency, and plot it
BW1=1.0*np.ones(shape(sanechannelsnocomb));   BW2=1.0*ones(shape(sanechannelsnocomb));
PK1=1.0*np.ones(shape(sanechannelsnocomb));   PK2=1.0*ones(shape(sanechannelsnocomb));
timeindex=range(1710,1800)
fineaz=arange(D1.az[timeindex[0]],D1.az[timeindex[-1]],0.1);

for a in range(size(sanechannelsnocomb)):
    pat=squeeze(real(D1.vis[timeindex,sanechannelsnocomb[a],0]+D1.vis[timeindex,sanechannelsnocomb[a],1] - D1.vis[timeindex[0],sanechannelsnocomb[a],0]-D1.vis[timeindex[0],sanechannelsnocomb[a],1]));
    patinterp=scipy.interpolate.interp1d(squeeze(D1.az[timeindex]), pat, kind='cubic')(fineaz);
    pk=np.max(pat);
    try:
        BW1[a]=fineaz[patinterp>(pk/2.)][-1]-fineaz[patinterp>(pk/2.)][0];
        PK1[a]=(fineaz[patinterp>(pk/2.)][-1]+fineaz[patinterp>(pk/2.)][0])/2.0
    except:
        BW1[a]=NaN;  PK1[a]=NaN;
    pat=squeeze(real(D2.vis[timeindex,sanechannelsnocomb[a],0]+D2.vis[timeindex,sanechannelsnocomb[a],1] - D2.vis[timeindex[0],sanechannelsnocomb[a],0]-D2.vis[timeindex[0],sanechannelsnocomb[a],1]));
    patinterp=scipy.interpolate.interp1d(squeeze(D2.az[timeindex]), pat, kind='cubic',bounds_error=False)(fineaz);
    pk=np.max(pat);
    try:
        BW2[a]=fineaz[patinterp>(pk/2.)][-1]-fineaz[patinterp>(pk/2.)][0];
        PK2[a]=(fineaz[patinterp>(pk/2.)][-1]+fineaz[patinterp>(pk/2.)][0])/2.0;
    except:
        BW2[a]=NaN;  PK2[a]=NaN;
    

axes[0].plot(FSNC,BW1,label='first pass',color='blue',linestyle='-');
axes[0].plot(FSNC,BW2,label='second pass',color='red',linestyle='--',alpha=0.4);
axes[0].legend(loc=1);    axes[0].set_title('Beamwidth  over  Frequency'); axes[0].set_ylim([0,10])
axes[0].set_xlabel('Frequency, MHz');  axes[0].set_ylabel('beamwidth, degrees');
axes[0].grid(True);
axes[1].plot(FSNC,PK1,label='first pass',color='blue',linestyle='-');
axes[1].plot(FSNC,PK2,label='second pass',color='red',linestyle='--',alpha=0.4);
axes[1].legend(loc=1);  axes[1].set_title('Peak Azimuth  over  Frequency');
axes[1].set_xlabel('Frequency, MHz');  axes[1].set_ylabel('peak azimuth, degrees');
axes[1].grid(True);
```

Location of the sky source
-------------

- At 16:20, the sun was at az=330, el=31.9  
- At 16:29, the sun was at az=328, el=30.9
