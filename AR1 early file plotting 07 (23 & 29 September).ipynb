Fit a slope to the delay and predict delay time and number of packets
------------------

```python
fn='/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/01/1443695415.h5'    # static test, no cable length difference
try:
    f = h5py.File(fn,'r')
except Exception as err_msg: raise SystemExit('An error has occured:\n%s' % err_msg)
vis = f['Data/correlator_data'].value
v = vis[2:,:,:,0] + 1.0j*vis[2:,:,:,1];  # discard the first few dumps in the file
ph=squeeze(angle(v[10,:,[3]])).T;

figure(figsize=(10,3));
plot(ph,lw=1,alpha=0.6,label='%s bl %i' % (fn[-13:-3],bl));
title('File:%s cross baseline, phase, radians' % fn[-13:-3]);  xlabel('channel');  ylabel('radians');
xlim([0,4095])

ph=unwrap(ph.T).T

figure(figsize=(10,3));
plot(ph,lw=1,alpha=0.6,label='%s bl %i' % (fn[-13:-3],bl));
title('File:%s cross baseline, phase, unwrapped radians' % fn[-13:-3]);  xlabel('channel');  ylabel('radians');
xlim([0,4095])

# find the slope of the last one in the file:
[m,c] = np.polyfit(range(800),ph[3000:3800],1)
delay=1/(.856 / (m/(2*pi)*4096));
delaysamples=(1/(.856 / (m/(2*pi)*4096)))/(1/1.7120);
print('slope is %f radians/channel  = %f ns  = %f samples' % (-m,-delay,-delaysamples))
annotate('Delay here = %5.2f ns' % -delay, xy=(3300,-215),  xycoords='data',xytext=(-100, 30), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='left', verticalalignment='bottom',fontsize=14)
```
