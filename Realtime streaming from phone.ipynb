```python

from IPython import display
import time
import numpy as np

plt.axis([0, 100, 9, 11]) # supoose I know what the limits are going to be
plt.figsize(16,6)
plt.ion()
plt.show()

import socket
host=''
port=8880

s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
s.setsockopt(socket.SOL_SOCKET, socket.SO_BROADCAST, 1)
s.bind((host, port))

n=0
times=[]
accs=[]

while True:
    try:
        message, address = s.recvfrom(8192)
        message = message.translate(None, ' ')
        floats = [float(x) for x in message.split(',')]
        times = append(times,floats[0])
        accs = append(accs,floats[4])
        if n%50==1:
            plt.gca().cla()
            plt.plot(times-times[0],accs,label='Z')
            plt.title('Total of %i data points' % shape(times)[0])
            #plt.legend()
            display.clear_output()
            display.display(plt.gcf()) 
        n=n+1
    except (KeyboardInterrupt, SystemExit):
        raise
#    except:
#        traceback.print_exc()
#        socket.socket()
```

```python
# plot some captured dat

from scipy import interpolate
import time

channels=1024
keepchannels=channels/4
dt = 0.002
timesnorm = times - times[0]
tmax = timesnorm[-1]
finetime=np.arange(0,tmax,dt);
dumptime=finetime[:-channels:channels]
freqs=np.fft.fftfreq(channels,dt)[1:keepchannels]

f = interpolate.interp1d(timesnorm,accs,kind='nearest')
interpaccs=f(finetime)
reshapeaccs = reshape(interpaccs[:(channels*int(shape(interpaccs)[0]/channels))],[-1,channels])

frsha=np.fft.fft(reshapeaccs,axis=1)[:,1:keepchannels]

accpower=sum(reshapeaccs**2.0,axis=1)**0.5


print 'Total of %i data points over %4.1f seconds' % (shape(timesnorm)[0],tmax)
print 'typical delta-t is %f s' % (tmax/shape(timesnorm)[0])


figure(figsize=(12,1.5))
plot(finetime,interpaccs)

figure(figsize=(12,1.5))
plot(dumptime,accpower)

figure(figsize=(12,4))
imshow(10*log10(abs(flipud(frsha.T))),aspect='auto',extent=[dumptime[0],dumptime[-1],freqs[0],freqs[-1]]);
xlabel('time'); ylabel('frequency , Hz');  title('Capture at %s' % time.ctime(times[0]));
```
