UHF far out sidelobes
=====================

Some UHF measurements were taken on 13 May 2015 with a UHF digitiser and warm receiver.

These can be analysed in two ways:

1. For the pass across the sun, looking for ripple in the passband, and variations in the ripple over angle,
2. For the comb generator, plotting Gain vs Angle

**This notebook concentrates on the wideband effects.**

```python
import katdal, csv, numpy as np, matplotlib, matplotlib.gridspec as gridspec, string, scipy
D1=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/05/27/1432735967.h5');   # 360 scan from 100 m away, VP
#D2=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/05/27/1432736479.h5');   # 360 scan from 100 m away, HP

F=1712-D1.channel_freqs/1e6;   # Create a flipped frequency axis for plotting purposes

# create a cropped frequency axis being only those allowed through the horn and the filters.  sanechannels  and  FS  the corresponding MHz frequency vector
freqrange=[400,850];
sanechannels = [i for i, v in enumerate(F) if ((v<=freqrange[1])&(v>=freqrange[0])&(i!=1105)&(i!=1114)) ];
FS = F[sanechannels];
#remove 1056, 1064

# find the channels which would contain the integer multiples of 10 MHz
combchannels = [i for i, x in enumerate(abs((F*1e6+D1.channel_width/2) % 10e6)-D1.channel_width/2) if (x < D1.channel_width/2)];
# remove combchannels which are outside the chosen frequency range, not pretty but it works
temp = combchannels[:];
for a in temp:
    if (not(a in sanechannels)):
        combchannels.remove(a);
FC = F[combchannels];

# create a frequency axis which does not contain the comb channels:   sanechannelsnocomb  and  FSNC the corresponding MHz frequency vector
sanechannelsnocomb = sanechannels[:];
temp = sanechannelsnocomb[:];
for a in temp:
    if (a in combchannels):
        sanechannelsnocomb.remove(a);
#sanechannelsnocomb.remove(0);   # remove the DC
FSNC = F[sanechannelsnocomb];
print('Loaded %s OK' % (D1.name));
```

```python
# plot sun over time, and over frequency
figure(figsize=(15,5));
timeindex=range(1400,1870);

addPower = squeeze(real(sum(D1.vis[:,sanechannelsnocomb,0:2],axis=2)))  -  squeeze(real(sum(D1.vis[1733,sanechannelsnocomb,0:2],axis=2)));

plot(D1.az[timeindex],10*log10(squeeze(real(sum(D1.vis[timeindex,600,0:2],axis=2)-np.min(sum(D1.vis[timeindex,600,0:2],axis=2),axis=0))))-19,label='%i MHz' % F[600],color='red',lw=1);
plot(D1.az[timeindex],10*log10(squeeze(real(sum(D1.vis[timeindex,1350,0:2],axis=2)-np.min(sum(D1.vis[timeindex,1350,0:2],axis=2),axis=0))))-20.8,label='%i MHz' % F[1350],color='blue',lw=1);
plot(D1.az[timeindex],10*log10(squeeze(real(mean(sum(D1.vis[timeindex,sanechannelsnocomb,0:2],axis=2),axis=1)-np.min(mean(sum(D1.vis[timeindex,sanechannelsnocomb,0:2],axis=2),axis=1),axis=0))))-19.9,label='mean of all channels',color='black',lw=2);
ylabel('Additional Power, dB_arbitrary');
title('UHF Noise Event, azimuth plot');
xlabel('Azimuth, degrees. Elevation=%4.1f' % D1.el[1750]);
xlim([-65,-45]); ylim([-20,30]);   legend(loc=2);  grid(True);

figure(figsize=(15,7));
for a in range(1752,1758):
    plot(F[sanechannelsnocomb],10*log10(addPower[a,:]),label='time %i' % a,lw=1+(a==1755)*2,alpha=0.7);
title('UHF Noise Event, Spectrum plot');
xlabel('Frequency, MHz');  axes[1].set_ylabel('Additional power in the source, dB_arbitrary');
xlim([500,850]);  ylim([20,42]);  legend(loc=3);  grid(True);


figure(figsize=(15,7));
for a in range(1753,1757):
    plot(F[sanechannelsnocomb],10*log10(addPower[a,:])-10*log10(addPower[1755,:]),label='time %i' % a,lw=1+(a==1755)*2,alpha=0.7);  #
title('UHF Noise Event, Spectrum plot, Normalised to time 1755');
xlabel('Frequency, MHz');  axes[1].set_ylabel('Additional power in the source, dB_peak');
xlim([500,850]);  ylim([-2,0.1]);  legend(loc=3);  grid(True);
```

Observations:

1. Beamwidth is narrower at high frequencies, causes the slope in the graphs
1. There is some very small ripple seen from 550 MHz to 750 MHz, not simply related to beamwidth. Approx magnitude is about 0.05 dB
