Cross-correlation experiment for sanity
-----------------
The magnitude of the cross-correlation product should reflect the power of the coherent part of the two signals

```python
import numpy as np;

couple=-6; # dB
x=256;      # number of points
n=10000.0;  # number of integrations
ofs=5;      # cable delay to channel B, in samples


T=zeros(x);
for y in range(int(n)):
  C=randn(x+ofs);         # the common part of the signal

  A=randn(x) + 10.0**(couple/20.0) * C[ofs:];
  B=randn(x) + 10.0**(couple/20.0) * C[:-ofs];

  FA=np.fft.fft(A) / sqrt(x);
  FB=np.fft.fft(B) / sqrt(x);

  AxB = FA * np.conj(FB);  # correlate
  T=T+AxB;  # integrate
    
T=T/n;   # divide out the number of integrations

figure(figsize=(10,3));
plot(10*log10(abs(T)),color='b');
xlim([0,100]);  title('Cross-correlation, dB magnitude, coherent noise = %i dB' % couple);

figure(figsize=(10,3));
plot(angle(T),color='r');
xlim([0,100]);  title('Cross-correlation, phase, coherent noise = %i dB' % couple);
```
