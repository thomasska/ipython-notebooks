```python
import katdal

#filename = '/var/kat/archive/data/MeerKATAR1/telescope_products/2016/02/01/1454344099.h5'
#filename = '/var/kat/archive/data/MeerKATAR1/telescope_products/2016/04/28/1461875338.h5'  # look for PKS 0741-063
filename = '/var/kat/archive/data/MeerKATAR1/telescope_products/2016/04/29/1461914733.h5';  tgt='PKS 1814-63'
filename = '/var/kat/archive/data/MeerKATAR1/telescope_products/2016/04/29/1461918321.h5'; tgt='PKS 1921-293'

data=katdal.open(filename)


ants = data.ants
```

Plot autocorrelations
----------------

```python
data.select(corrprods='auto',scans='track',targets=tgt,pol='H')
print(shape(data.vis))

ch=2500
figure(figsize=(10,3))
plot(10*log10(squeeze(data.vis[:,ch,:])))
title('Autocorrelations over time, ch=%i F=%6.1f MHz' % (ch,data.channel_freqs[ch]/1.0e6))

dump=40
figure(figsize=(10,3))
plot(10*log10(squeeze(data.vis[dump,:,:])))
title('Autocorrelations over channel, dump=%i' % (dump));
```

Plot cross-correlations
---------------

```python
data.select(corrprods='cross',scans='track',targets=tgt,pol='H')

print(shape(data.vis))

ch=2500
figure(figsize=(10,3))
plot(10*log10(squeeze(abs(data.vis[:,ch,:]))))
title('Cross-correlation magnitude over time, ch=%i F=%6.1f MHz' % (ch,data.channel_freqs[ch]/1.0e6))
figure(figsize=(10,3))
plot(np.unwrap(squeeze(angle(data.vis[:,ch,:])),axis=0))
title('Cross-correlation phase over time, ch=%i F=%6.1f MHz' % (ch,data.channel_freqs[ch]/1.0e6))

dump=40
figure(figsize=(10,3))
plot(10*log10(abs(squeeze(data.vis[dump,:,:]))))
title('Cross-correlation magnitude over channel, dump=%i' % (dump));
figure(figsize=(10,3))
plot(angle(squeeze(data.vis[dump,:,:]))+np.linspace(0,pi*10,num=6))
title('Cross-correlation phase over channel, dump=%i  (spread out for visibility)' % (dump));
```
