Plot the power ramp
===================

```python
import katdal, numpy as np, matplotlib
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/09/29/1443528933.h5');     # fine power ramp
```

```python
# autocorrelations
t=10;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[t,:,0]))),label='hh')
plot(10*log10(abs(squeeze(D.vis[t,:,1]))),label='vv')
title('Autocorrelation baselines of %s over channel at t=%i' % (D.experiment_id,t)); xlabel('Channel');  ylabel('Correlator power, dBarbitrary'); xlim([0,4095]);  legend(loc=0);

ch=1360;
figure(figsize=(12,2))
plot(10*log10(abs(squeeze(D.vis[:,ch,0]))),label='hh')
plot(10*log10(abs(squeeze(D.vis[:,ch,1]))),label='vv')
title('Autocorrelation baselines of %s over time at channel=%i'% (D.experiment_id,ch)); xlabel('Dump');  ylabel('Correlator power, dBarbitrary');  legend(loc=0);
```

```python
#plot correlator power vs signal generator power

D.select()
powerbase=squeeze(abs(D.vis[0,1362,0]))  # get a base power number from a side channel in the first dump

D.select(compscans='~transition')

Freq=np.zeros(D.shape[0]);                # initialise Frequency and Power vectors
Power=np.zeros(D.shape[0]);
ch1360=np.zeros(D.shape[0]);
ch1370=np.zeros(D.shape[0]);
ch1371=np.zeros(D.shape[0])
count=0;

for cs in D.compscans():
    num=D.shape[0];
    freq=float(cs[1].split(',')[1].split('=')[1])/1.0e6;
    pwr=float(cs[1].split(',')[2].split('=')[1]);
    Freq[count] = freq;
    Power[count] = pwr;
    ch1360[count] = squeeze(abs(D.vis[0,1361,0]))
    ch1370[count] = squeeze(abs(D.vis[0,1371,0]))
    ch1371[count] = squeeze(abs(D.vis[0,1372,0]))
    count=count+1;
    if count>57: break;
    #print('shape %s  %s  Frequency %f MHz   Power %s dBm  startdump=%i'  % (D.shape, shape(D.vis), freq, pwr, D.dumps[0] ) );

figure(figsize=(12,6))
plot(Power,10*log10(ch1360-powerbase),marker='.',label='channel 1360');
grid(True); legend(loc=0);
xlabel('Signal generator power, dBm');  ylabel('Excess correlator power, dB CU');  title('Meas %s -  %s  Correlator Linearity plot' % (D.name[-13:-3],D.start_time));


figure(figsize=(12,6))
plot(Power,10*log10(ch1360),marker='.',label='channel 1360');
plot(Power,10*log10(ch1370),marker='.',label='channel 1370');
plot(Power,10*log10(ch1371),marker='.',label='channel 1371');
grid(True); legend(loc=0);
xlabel('Signal generator power, dBm');  ylabel('total correlator power, dB CU');  title('Meas %s -  %s  Correlator Linearity plot' % (D.name[-13:-3],D.start_time));


D.select()

print(1712e6-D.channel_freqs[1360])
print(Freq[0]*1e6)
print(Power)
```

PFB shape
============

```python
import katdal, numpy as np, matplotlib
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/09/30/1443645086.h5');     # 1 kHz steps

D.select()
powerbase=squeeze(abs(D.vis[0,1362,0]))  # get a base power number from a side channel in the first dump

D.select(compscans='~transition')

Freq=np.zeros(D.shape[0]);                # initialise Frequency and Power vectors
Power=np.zeros(D.shape[0]);
ch1359=np.zeros(D.shape[0]);
ch1360=np.zeros(D.shape[0]);
ch1361=np.zeros(D.shape[0]);
count=0;

for cs in D.compscans():
    num=D.shape[0];
    freq=float(cs[1].split(',')[1].split('=')[1])/1.0e6;
    pwr=float(cs[1].split(',')[2].split('=')[1]);
    Freq[count] = freq;
    Power[count] = pwr;
    ch1359[count] = squeeze(abs(D.vis[0,1360,0]))
    ch1360[count] = squeeze(abs(D.vis[0,1361,0]))
    ch1361[count] = squeeze(abs(D.vis[0,1362,0]))
    count=count+1;
    #print('shape %s  %s  Frequency %f MHz   Power %s dBm  startdump=%i'  % (D.shape, shape(D.vis), freq, pwr, D.dumps[0] ) );

figure(figsize=(12,6))
plot(Freq,10*log10(ch1359-powerbase),alpha=0.4);
plot(Freq,10*log10(ch1360-powerbase),lw=2);
plot(Freq,10*log10(ch1361-powerbase),alpha=0.4);
grid(True);
xlabel('Signal generator frequency, MHz');  ylabel('Excess correlator power, dB CU');  title('Meas %s -  %s  Correlator PFB shape plot' % (D.name[-13:-3],D.start_time));
ylim([-25,30]);  xlim([571.4,572.2]);

D.select()

```
