Test for repetition of data
------------------

Two related bugs here:

1. Data is repeated, sometimes a few times in a row
1. Data is stored, as though there's a 1-snapshot buffer somewhere

This script toggles the signal generator power and observes the total powe in the snapshots.

```python
import md5, time
waitBetween = 0.1;

oldVDigest='';  oldHDigest='';  signal=False;

print('Taking a series of snapshots, comparing data. Delay=%3.1f seconds' % waitBetween)
for a in range(47):
    if mod(a,4)==0:
        signal=~signal;
        sigGen(signal)
        
    time.sleep(waitBetween)
    dupeString='';  v = md5.new();  h = md5.new();
    samples=getSamples();
    
    v.update(samples[0,:]);
    h.update(samples[1,:]);
    thisVDigest=v.hexdigest()
    thisHDigest=h.hexdigest()
    if thisVDigest==oldVDigest:  dupeString=dupeString + '  Duplicate V';
    if thisHDigest==oldHDigest:  dupeString=dupeString + '  Duplicate H';
    
    print('%2i - Vhash=%s Vpwr=%5.1f   Hhash=%s Hpwr=%5.1f   %s' % (a,thisVDigest[:6],np.mean(samples[0,:]**2),thisHDigest[:6],np.mean(samples[1,:]**2),dupeString))
    
    oldVDigest=thisVDigest;     oldHDigest=thisHDigest

```

Functions to do the work
==============================

```python
# funtion to return the number of samples by which the digitiser streams are offset

def digitiserCommand(commandString,main=False):
    #mainIP='192.168.4.36:7147'; subIP='192.168.4.36:7201' # DBElab03
    #mainIP='192.168.4.23:7247'; subIP='192.168.4.23:7262' # DBElab04 - Wesley
    #mainIP='10.103.254.2:7247'; subIP='10.103.254.2:7224' # M024 on site :-)
    mainIP='10.103.253.2:7247'; subIP='10.103.253.2:7262'  # M062 on RTS
    
    import commands
    if main: ip=mainIP
    else:  ip=subIP
    return commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 100 -s " + ip + " ?" + commandString);


def getSamples():
    result=digitiserCommand("adc-snap-shot vh | grep '#adc-snap-shot'")
    if result[0]!=0: print('Failure retrieving samples');
    result=result[1]
    snapshot=result.replace('#adc-snap-shot v','').replace('#adc-snap-shot h','').replace('\n','');
    samples=np.array(  map(int, snapshot.split(' ')[1:] )  ).reshape(2,-1)
    return samples


def sigGen(onOff):
    # Connect and set up.
    import socket, time, datetime, array, numpy
    sg = socket.socket(socket.AF_INET, socket.SOCK_STREAM);  sg.setblocking(0); sg.settimeout(1.0)
    sg.connect(('192.168.4.33', 10001));  # Lantronix TCP-RS232 converter     In the lab: 192.168.4.33
    time.sleep(0.5)
    if onOff:
        sg.send('OUTP ON\n');
        print(' Signal generator On, wait 1 second');
    else:
        sg.send('OUTP OFF\n');
        print(' Signal generator Off, wait 1 second');
    # Disconnect
    time.sleep(1)
    sg.close();
    
    
    
def doGlobalSync():
    import time
    result=digitiserCommand("global-synchronise",main=True);
    if result[0]!=0: print('Error performing global sync');


def plotSpectrum(numSnapshots=4,fftLength=1024):
    print('Spectrum with %i snapshots, fftLength=%i.     Getting snapshots...' % (numSnapshots,fftLength))
    allSamples=zeros([0,2,fftLength]);
    for a in range(numSnapshots):
        #print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples();
        allSamples=np.concatenate((allSamples,samples.reshape(16384/fftLength,2,-1)),0)
    
    print(shape(allSamples))
    
    Fsamples=fft.fft(  allSamples,  fftLength,  2) / fftLength
    freq=np.linspace(856,1712,num=fftLength/2);
    
    Corr = Fsamples[:,0,:] * conj(Fsamples[:,1,:]);
    Corr = np.mean(Corr,axis=0)[:fftLength/2];
    
    Auto = Fsamples[:,:,:] * conj(Fsamples[:,:,:]);
    Auto = np.mean(Auto,axis=0)[:,:fftLength/2].T;
    
    flatSamples=np.concatenate((squeeze(allSamples[:,0,:]).reshape(1,-1),squeeze(allSamples[:,1,:]).reshape(1,-1)),0);
    
    figure(figsize=(16,6))
    plot(freq,10*log10(abs(Auto))); title('Auto and Cross - Magnitude dB');
    plot(freq,10*log10(abs(Corr)));
    xlim([856,1712]);  ylim([-40,10]); grid(True);

    meanPwrCross=10*log10(  np.mean( abs( Corr[int((1300.0-856)/1712.0*fftLength):int((1500.0-856)/1712.0*fftLength)] ) )  );
    meanPwrAuto=10*log10(    np.mean(   np.mean( abs( Auto[int((1300.0-856)/1712.0*fftLength):int((1500.0-856)/1712.0*fftLength),:] ),axis=1 ),axis=0   )    );
    
    plot([freq[int((1300.0-856.0)/1712.0*fftLength)],freq[int((1500.0-856.0)/1712.0*fftLength)]],[meanPwrCross,meanPwrCross],'r-',lw=2,alpha=0.5) 
    plot([freq[int((1300.0-856.0)/1712.0*fftLength)],freq[int((1500.0-856.0)/1712.0*fftLength)]],[meanPwrAuto,meanPwrAuto],'k-',lw=2,alpha=0.5) 
    print('Average power 1300-1440 MHz, cross-correlation:  %f dB,  autocorrelation  %f dB' % (meanPwrCross,meanPwrAuto) );
    
    figure(figsize=(16,2))
    plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,700]);
    figure(figsize=(16,2))
    hist(flatSamples.T,np.arange(-100,100,2)); title('voltage histogram');
    
        
        
        
    
def getDigitiserOffset(numSnapshots=4,globalSync=False,fftLength=256,verbose=0):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync()
        time.sleep(2)
        
    selectedRange=[0.63,0.9]        # channels to us to find sample offset
    selectedRange = np.round(np.multiply(selectedRange,fftLength/2))

    allSamples=zeros([2,0])
    for a in range(numSnapshots):
        if verbose>0: print('Getting snapshot %i of %i' % (a+1,numSnapshots) )
        samples=getSamples();
        allSamples=np.concatenate((allSamples,samples),1)
    
    Fsamples=fft.fft(  allSamples.reshape(2,-1,fftLength),  fftLength,  2)
    
    Corr = Fsamples[0,:,0:fftLength] * conj(Fsamples[1,:,0:fftLength]);
    Corr=squeeze(np.sum(Corr,axis=0));
    CorrAngle=unwrap(angle(Corr))
    Pwr=squeeze(np.sum(abs(Fsamples**2),axis=1)).T;

    offsetSamples = (CorrAngle[selectedRange[1]]-CorrAngle[selectedRange[0]]) / (2.*pi*(selectedRange[1]-selectedRange[0])/fftLength)
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(allSamples.T,alpha=0.6); title('Voltages over time'); xlim([0,1000]);
        figure(figsize=(16,3))
        hist(allSamples.T,np.arange(-40,40)); title('voltage histogram');
        
    if verbose>1:
        figure(figsize=(16,5))
        plot(10*log10(Pwr)); title('Auto and Cross - Magnitude dB'); xlim([0,fftLength/2]);
        plot(10*log10(abs(Corr)));
        figure(figsize=(16,3))
        plot(unwrap(angle(Corr[0:fftLength/2]))); title('Cross - Unwrapped Angle'); xlim([0,fftLength/2]);
        plot(selectedRange,[0,0],'b-',lw=3)
        
    if verbose>0:
        print('Samples offset:  %5.2f' % offsetSamples)
        print('Voltage mean=%5.2f stdev=%5.2f max=%3i min=%3i' % (np.mean(allSamples),np.std(allSamples),np.max(allSamples),np.min(allSamples) ))
    
    return offsetSamples



def getDigitiserOffsetTD(numSnapshots=1,globalSync=False,verbose=0):
    import time
    
    if globalSync:
        if verbose>3: print('Performing Global Sync' )
        doGlobalSync()
        time.sleep(2)

    if verbose>3: print('Getting snapshot' )
    samples=getSamples();

    corr=abs(np.correlate(samples[0,:],samples[1,:],mode='same'))
    offsetSamples = size(corr)/2 - numpy.argmax(corr);
    
    if verbose>2:
        figure(figsize=(16,3))
        plot(samples.T,alpha=0.6); title('Voltages over time'); xlim([0,100]);
        figure(figsize=(16,3))
        hist(samples.T,np.arange(-40,40)); title('voltage histogram');
        
    if verbose>1:
        figure(figsize=(16,3))
        plot(range(-8192,8192),corr); title('Correlation magnitude');
        xlim([-200,200]); grid(True);

    if verbose>0:
        print('Samples offset: %i.   Voltage mean=%5.2f stdev=%5.2f max=%3i min=%3i.  AveragePower=%5.2f dB' % (offsetSamples,np.mean(samples),np.std(samples),np.max(samples),np.min(samples),10*log10(np.mean(samples**2)/261632.50) ))
    
    return offsetSamples





def powerCycleDigitiser(globalSync=True,verbose=0):
    #mainIP='192.168.4.36:7147'; subIP='192.168.4.36:7201' # DBElab03
    mainIP='192.168.4.23:7247'; subIP='192.168.4.23:7262' # DBElab04 - Wesley
    #mainIP='10.103.254.2:7247'; subIP='10.103.254.2:7224' # M024 on site :-)
    import commands,time
    if (verbose>0)&~globalSync: print('  Power cycling digitiser with absent / ready');
    if (verbose>0)&globalSync: print('  Power cycling digitiser with absent / ready, and Global Sync');
    time.sleep(5)
    if verbose>1: print('   Digitiser IP: %s (master) and %s (subordinate)' % (mainIP,subIP));
    if verbose>1: print('   Selecting band 0');
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 120 -s " + subIP + " ?select-band 0");
    if result[0]: print('   Error selecting band 0')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Marking absent');
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 120 -s " + subIP + " ?digitiser-status l absent");
    if result[0]: print('   Error marking absent!')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Marking present');
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 120 -s " + subIP + " ?digitiser-status l ready 1");  ############ Remember m024 is serial number 7
    if result[0]: print('   Error marking present!')
    if verbose>2: print(result[1])
    time.sleep(10)
    if verbose>1: print('   Selecting band L');
    result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 120 -s " + subIP + " ?select-band l");
    if result[0]: print('   Error selecting band L!')
    if verbose>2: print(result[1])
    time.sleep(3)
    if globalSync:
        if verbose>1: print('   Global Sync');
        result=commands.getstatusoutput("svn/katcp_devel/cmd/kcpcmd -t 120 -s " + mainIP + " ?global-synchronise");
        if verbose>2: print(result[1])
    
```
