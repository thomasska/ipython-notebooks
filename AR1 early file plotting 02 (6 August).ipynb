AR1 plotting - 6 August
=============
Scripts to plot the first results from AR1.

```python
# load the results from the h5 file directly
import h5py, numpy as np
f = h5py.File('ar1/1438865356.h5','r')
vis = f['Data/correlator_data'].value
v= vis[2:,:,:,0] + 1.0j*vis[2:,:,:,1];
#ts = f['Data/timestamps'].value
```

Plot some autocorrelations
------------

```python
print('(time, frequency, baseline, real+imag)',(shape(v)))
# because of oversubscription, there are only be two types of vis - auto and cross. nothing else is possible.  Auto is baseline 5, Cross is baseline 15, among others.

figure(figsize=(12,2));
plot(10*log10(squeeze(abs(v[:,500,5]))),color='r',label='channel 500');
plot(10*log10(squeeze(abs(v[:,502,5]))),color='b',label='channel 502');
ylim([0,45]);  legend(loc=0)
xlabel('time'); title('Auto, over time');  ylabel('dB correlator counts');

figure(figsize=(12,5));
plot(10*log10(squeeze(abs(v[10,:,5].T))),color='b',label='Autocorrelation power');
ylim([0,41]); xlim([0,4095]);  legend(loc=0)
xlabel('channel'); title('auto, over channel');  ylabel('dB correlator counts');
#annotate('Original signal', xy=(1910,40),  xycoords='data',xytext=(-25, -20), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='right', verticalalignment='bottom',fontsize=10)
#annotate('Strange duplicate', xy=(2050,40),  xycoords='data',xytext=(+25, -20), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='left', verticalalignment='bottom',fontsize=10)
#annotate('2nd harmonic of signal generator?', xy=(3960,33),  xycoords='data',xytext=(-25, -25), textcoords='offset points', arrowprops=dict(facecolor='black',width=1.5,shrink=0.05,frac=0.3,headwidth=10),horizontalalignment='right', verticalalignment='bottom',fontsize=10)

figure(figsize=(12,3));
plot(10*log10(squeeze(abs(v[10,:,5].T))),color='b',label='Autocorrelation power');
ylim([0,41]); xlim([450,550]);  legend(loc=0)
xlabel('channel'); title('auto, over channel');  ylabel('dB correlator counts');

figure(figsize=(12,3));
plot(10*log10(squeeze(abs(v[10,:,5].T))),color='b',label='Autocorrelation power');
ylim([0,41]); xlim([1400,1600]);  legend(loc=0)
xlabel('channel'); title('auto, over channel');  ylabel('dB correlator counts');
```
