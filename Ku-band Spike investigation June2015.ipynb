Looking at the periodic 11 second spikes seen in the Ku band data on 2 June 2015

First plot is of the spikes over time.

```python
filename='07/09/1436423000.h5'   # is this a 32k file?
#filename='07/09/1436428152.h5'      # measurement after update to correlator by JM    20150709-0002
#filename='06/02/1433255123.h5'      # most recent Ku problem measurement              20150602-0010
#filename='05/27/1432736479.h5'      # investigate some of the panel gap measurements  20150520-0005
#filename='05/13/1431514867.h5'      # even older                                      20150513-0003

import katdal, matplotlib, numpy as np, scipy as sp
D=katdal.open('/var/kat/archive/data/RTS/telescope_products/2015/' + filename); 
spikeChannels=[429, 1453, 2477, 3501];
print(shape(D.vis))
```

```python
# spike period investigation
fig, axes = plt.subplots(nrows=1, ncols=1,figsize=(12,5));
timeRange=range(0,200);                       # dumps, not seconds
tim=(D.mjd[timeRange]-D.mjd[0])*86400;        # make a time axis in seconds
pks=squeeze(real(D.vis[timeRange,spikeChannels,0]-D.vis[timeRange,np.add(spikeChannels,1),0]));  # subtract noise power
pks=np.divide(pks,np.max(pks,axis=0));        # normalise
pks=pks+range(4);                             # spread out for plotting
axes.plot(tim,pks);
pulseTime=((tim-7)%10.9955<D.dump_period);    # spike clock, by experiment
axes.plot(tim,pulseTime*0.5-1,label='11 s markers');
axes.set_xlabel('time, seconds');  axes.legend(loc=1);  axes.set_title('Experiment: ' + D.experiment_id + '  Time plot of the 11 second period spikes'); axes.set_ylabel('arbitrary channel numbers');
```

The period of the spikes looks like 11 seconds, but it's not absolutely locked to 11 correlator dumps and slips slightly over time.  
A period of 10.9955 s fits better.

```python
# spike amplitude investigation
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(12,9));
tim=(D.mjd-D.mjd[0])*86400;                                  # make a time axis in seconds
pks=squeeze(real(D.vis[:,spikeChannels,0]));                 # channels with spikes in them
noises=squeeze(real(D.vis[:,np.add(spikeChannels,1),0]+D.vis[:,np.add(spikeChannels,-1),0]))/2.0;     # adjacent channels for noise removal
# plot the absolute power
axes[0].step(tim[0:50],pks[0:50,0],label='945 MHz')
axes[0].legend(loc=1);  axes[0].set_title('Amplitude of a single spiky channel over a short time. Note the exact doubling of power when the spike occurs'); axes[0].set_ylabel('abs(Autocorrelation)');
axes[0].set_ylim([0,450]);

# plot the ratio
for a in range(4):
    spikeIndex=(pks[:,a]-noises[:,a])>30                     # find the index of spike channels
    spiketim=tim[spikeIndex];                                  # extract time of spikes
    spikemag=pks[spikeIndex,a];                                # extract spike amplitude
    noisemag=noises[spikeIndex,a];                             # extract noise amplitude
    axes[1].plot(spiketim,spikemag/noisemag,label=str(int(D.channel_freqs[spikeChannels[a]]/1e6)));
    #axes[2].hist(spikemag/noisemag,bins=300,normed=True,label=str(int(D.channel_freqs[spikeChannels[a]]/1e6)));
axes[1].set_xlabel('time, seconds');  axes[1].legend(loc=1);  axes[1].set_title('Experiment  ' + D.experiment_id + '   Amplitude of the periodic spikes, as a ratio of P_spike to P_notspike');
axes[1].set_ylim([1.95,2.05]);
#axes[2].set_xlim([1.95,2.05]);
```

Very interesting - the spike power is almost exactly doubled.  

**Conclusion so far**  
This can only be a correlator problem.  
Four channels, equally spaced across the band, seem to be accidentally doubled, possibly by not reliably being cleared after integration.

Very interesting. The probability of getting the spike (at the 11th dump) varies continuously over time, and is different for each channel.
