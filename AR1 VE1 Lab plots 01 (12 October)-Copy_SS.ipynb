```python
# phase slope in the noise band
ch=[2600,3700];
ph=squeeze(angle(D.vis[3:-3,ch[0]:ch[1],3])).T;
ph=unwrap(ph.T).T

figure(figsize=(12,3));
plot(range(ch[0],ch[1]),ph,color='m');
xlim(ch);  ylabel('Unwrapped Phase');  xlabel('Channel');  title('AR1 Lab cross correlation unwrapped phase, noise band, all times overlaid');

ph=mean(ph,axis=1);  # average before finding slope
[m,c] = np.polyfit(range(ch[0],ch[1]),ph,1)
delay=1/(.856 / (m/(2*pi)*4096));
delaysamples=(1/(.856 / (m/(2*pi)*4096)))/(1/1.7120);

print('Phase slope in noise band of %s  %s is as follows: ' %  (D.experiment_id,D.description));
print('  %f radians/channel  = %6.2f ns  = %6.2f samples' % (-m,-delay,-delaysamples))
```

```python
import katdal, numpy as np, matplotlib
D=katdal.open('/var/kat/archive/data/MeerKATAR1/telescope_products/2015/10/27/1445947672.h5');

D.select()

D.select(compscans='~transition')

Freq=[];  Chan=[];  Power=[];
target=[];  nottarget=[];
count=0;


for cs in D.compscans():
    num=D.shape[0];
    siggenfreq=float(cs[1].split(',')[1].split('=')[1])/1.0e6;
    siggench=int(4096-siggenfreq*4096/856);
    siggenpwr=float(cs[1].split(',')[2].split('=')[1]);
    Freq.extend([siggenfreq])
    Chan.extend([siggench])
    Power.extend([siggenpwr])
    target.extend([squeeze(abs(D.vis[0,siggench,0]))])
    nottarget.extend([squeeze(abs(D.vis[0,siggench+5,0]))])
    count=count+1;
    #print('shape %s  %s  Frequency %f MHz   Channel %i  Power %s dBm  startdump=%i'  % (D.shape, shape(D.vis), siggenfreq, siggench, siggenpwr, D.dumps[0] ) );

figure(figsize=(12,4));
plot(Chan,10*log10(target),'b-',label='Channel excited by signal generator')
plot(Chan,10*log10(nottarget),'k-',label='nearby channel at each time')
xlabel('Channel');  ylabel('dB Correlator Counts');  grid(True);  xlim([0,4096]);
title('Selected and adjacent channels - %s - %s' % (D.experiment_id,D.description))
```
