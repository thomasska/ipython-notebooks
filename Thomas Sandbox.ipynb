```python
# try one simulation and look at the correlations
import numpy, random, scipy
powerRandom = 0           # dB
powerCoherent = -3        # dB
lengthSnapshot = 512       # samples
xyOffset = 10              # samples
integrations = 5*16384/lengthSnapshot        # integrations

CORR=zeros(lengthSnapshot);
AX=zeros(lengthSnapshot);
AY=zeros(lengthSnapshot);

for a in range(integrations):
    CRS=randn(lengthSnapshot+100)

    voltageX = 10.0**(powerRandom/20.0)*(randn(lengthSnapshot)) + 10.0**(powerCoherent/20.0)*CRS[0:lengthSnapshot]
    voltageY = 10.0**(powerRandom/20.0)*(randn(lengthSnapshot)) + 10.0**(powerCoherent/20.0)*CRS[xyOffset:lengthSnapshot+xyOffset]

    FX = fft.fft(voltageX);  FY = fft.fft(voltageY);
    AX = AX + FX * np.conj(FX);
    AY = AY + FY * np.conj(FY);
    CORR = CORR + FX * np.conj(FY);

CORR=CORR/integrations;
AX=AX/integrations;
AY=AY/integrations;

figure(figsize=(12,5))
normLevel=10*log10( (np.mean(abs(AX))+np.mean(abs(AY)))/2.0 ) 
plot(10*log10(abs(AX))-normLevel,'b-',alpha=0.5)
plot(10*log10(abs(AY))-normLevel,'g-',alpha=0.5)
plot(10*log10(abs(CORR))-normLevel,'r-',alpha=0.5)
ylim([-25,2]); ylabel('dB'); grid(True);
```

Hailstone Patterns
==============
from [Pickover](sprott.physics.wisc.edu/pickover/hailstone.html)

$h(n) = n/2$ if n even, or
$h(n) = 3n+1$ if n odd

```python
import Image, numpy as np
xmax=500;
ymax=500;
iters=50000

mtx=np.zeros([ymax,xmax],dtype=integer);

currentvector=range(xmax)

for x in range(iters):
    for a,n in enumerate(currentvector):
        # do the transformation
        if (n%2):
            newn=3*n+1
        else:
            newn=n/2
        currentvector[a]=newn;
        # apply to the matrix, if it's small enough, otherwise ignore
        if newn<ymax:
            mtx[newn,a]=mtx[newn,a]+1
        
    #if (x%45)==0:
    #    print(currentvector)
    #    print(mtx)

#print(mtx)
figure(figsize=(12,12*xmax/(1.0*ymax)))
imshow(np.flipud(-mtx),aspect='auto',interpolation='none', cmap=plt.get_cmap('CMRmap'),vmin=-1,vmax=0,extent=[0,xmax,0,ymax])
title('Pickover hailstone numbers. %i iterations' % iters)

del mtx
```

PFB shape experiment

```python
# swept-stimulus test
import numpy

l=2048;          # length of each block
ch=50;           # channel to be investigated
numpts=2001;     # number of fine-frequency points
over=32;         # length of the FIR in block-lengths

T=linspace(0.0,l-1.0,l);
TT=linspace(0.0,l*over-1.0,l*over);
F=numpy.fft.fftfreq(l);
deltafreq=F[2]-F[1];

wind=np.hanning(over*l)*np.sinc((TT-over*l/2)/l);  # sinc*hanning window, looks pretty good
figure(figsize=(12,2))
plot(wind); grid(True);  title('PFB window function  -   sinc*hanning')
dchrange=np.linspace(-0.99,3.99,num=numpts);
DFV=1.0*np.arange(numpts);
DFVP=1.0*np.arange(numpts);

for a in range(numpts):
    freq = F[ch] + deltafreq*dchrange[a];
    # the FFT version
    V = np.sin(2.0*3.14159*freq*T);
    FV = numpy.fft.fft( V )/l*2.0;
    DFV[a] = 20.0*log10(np.abs(FV[ch]))
    
    # now the PFB version
    V = np.sin(2.0*3.14159*freq*TT)
    V = V * wind
    V = np.reshape(V,(over,l))
    V=np.sum(V,axis=0)
    FV = numpy.fft.fft( V )/l*2.0;
    DFVP[a] = 20.0*log10(np.abs(FV[ch]))
    
    # print('a=%i  dch=%f  freq=%f  FV[ch]=%f dB   DFV[a]=%f dB    DFVP[a]=%f dB' % (a,dch,freq,20.0*log10(abs(FV[ch])),DFV[a],DFVP[a]))

figure(figsize=(12,4))
plot(dchrange,DFV,label='FFT, %i points' % l);
plot(dchrange,DFVP,label='PFB+FFT, %i over' % over);
plot(dchrange+1,DFVP,alpha=0.3,color='k',label='PFB+FFT next channel');
ylim([-60,1]);  grid(True);  legend(loc=0)
xlabel('Frequency bins');  title('PFB and FFT comparison');   ylabel('Response, dB')
```
