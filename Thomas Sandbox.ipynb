## Phasor addition

I just needed a complex-number calculator, a bit of plotting doesn't hurt though.

```python
# dB level of the secondary signal
secondary = -6.0;

# create two signals
pri = complex(1.0, 0.0);
sec = complex(0.0, 1.0) * 10 ** (secondary/20);
res = pri+sec;

# find the phase of each
apri = degrees(angle(pri));
asec = degrees(angle(sec));
ares = degrees(angle(res))

aerror =ares-apri;
print "Worst-case phase error of a signal with a",secondary,"dB secondary signal is",aerror,"degrees"

figure();
plot([0,pri.real],[0,pri.imag],color="b");
plot([pri.real,res.real],[pri.imag,res.imag],color='r')
plot([0,res.real],[0,res.imag],color="k");
axis([-0.2,1.2,-0.3,0.7])
text(pri.real/2,pri.imag/2,str(abs(pri))+' angle '+str(degrees(apri)),color='b',horizontalalignment='center',verticalalignment='top',rotation=apri);
text(pri.real+sec.real/2,sec.imag/2,str(round(abs(res),2))+' angle '+str(round(asec,2)),color='r',horizontalalignment='right',verticalalignment='center',rotation=asec);
text(res.real/2,res.imag/2,str(round(abs(res),2))+' angle '+str(round(ares,2)),color='k',horizontalalignment='center',verticalalignment='bottom',rotation=ares);
title('Addition of a secondary signal at'+str(secondary)+'dB');
grid();
```

## Plotting the GSM signal from many files

For Simon, a method of plotting the strength of the GSM signal for many files in a directory

```python
# set up the basics
import katdal
import csv
import numpy as np
import matplotlib
import matplotlib.gridspec as gridspec
chosendir='/var/kat/archive/data/RTS/telescope_products/2014/12/03'
ch=460;
chosenvis=1;

# load and plot
D=!ls $chosendir;
numFiles=shape(D)[0];
numFiles=min(99,numFiles);

print 'Loading and plotting all ',numFiles,' files in ',chosendir
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(14,12));

for a in range(numFiles):
    d=D[a]
    print '  Loading file ',a,' filename ',d
    A=katdal.open(chosendir + '/' + d);
    numDumps=shape(A.timestamps)[0];
    print '    size=',numDumps,' Freq=',A.channel_freqs[ch]/1e6;
    GSMpower=10*log10(real(squeeze(A.vis[:,ch,chosenvis])))
    if (numDumps>100)&(mean(GSMpower)>10) :   # only plot the bigger files
        axes[0].plot(GSMpower,lw=2,alpha=0.6,label=d);
        axes[1].plot(A.channel_freqs/1e6, 10*log10(squeeze(mean(real(A.vis[:,:,chosenvis]),axis=0))),lw=2,alpha=0.6,label=a);

freqMHz = A.channel_freqs[ch]/1e6;
axes[0].set_xlabel('Time, s');  axes[0].set_ylabel('Power, dB');  axes[0].set_title('Plot of all files %f MHz' % freqMHz);  axes[0].legend(loc=1);
axes[0].set_xlim([0,400]); #axes.set_ylim([5,40]);

axes[1].set_xlabel('Freq, MHz');  axes[1].set_ylabel('Power, dB');  axes[1].set_title('Plot of all files over frequency, mean power over whole measurement');   axes[1].legend(loc=1);
axes[1].set_xlim([920,960]); #axes.set_ylim([5,40]);
```

## FFT scratching

```python
import numpy
freq=0.0996;
L=1024;
t=linspace(0.0,L-1.0,L);
f=numpy.fft.fftfreq(L);
volt=sin(2*pi*freq*t);
print "RMS voltage is ",sqrt(mean(volt**2))
fv = numpy.fft.fft(volt)/L;
plot(f,abs(fv)); xlim([-0.2,0.2])
print "RMS voltage in frequency domain is ",sqrt(mean(volt*conj(volt)))
print "Power in f=freq bins is",abs(fv[102])," and ",abs(fv[922])  # 102 and 922
```

# Load shedding

Eskom provides `http://loadshedding.eskom.co.za`  but it's hard to extract the current stage.  With a bit of scratching I find they have very helpfully provided this:  
`http://loadshedding.eskom.co.za/LoadShedding/GetStatus`  
which returns  
`3`  
today, in stage 2.

PowerAlert provides a graph of system status, including an indicator of the total load and projected load over time.  
`http://www.poweralert.co.za/`  

```python
import numpy
import urllib2
from time import time
from datetime import datetime
import matplotlib


# retrieve the load shedding status. Super-easy, thanks Eskom!
response = urllib2.urlopen('http://loadshedding.eskom.co.za/LoadShedding/GetStatus')
shed = int(response.read())-1
#print 'At time',int(time()), datetime.now().time(),'Load shedding status is ',shed


# retrieve the graph of power consumption, "grid pressure" colours and times of day
import string
response = urllib2.urlopen('http://www.poweralert.co.za/poweralert5/index.php')
html = response.read();

# do the times of day
n = string.find(html,'<td align="center" valign="middle"><table width="800" border="0" align="center" cellpadding="1" cellspacing="0" bgcolor="#000000">');
temp = html[n:(n+7500)];
T=range(64);
m=1;
for a in range(64):
    m = string.find(temp,'<td valign="top">',m+2);
    tim = temp[(m+17):(m+19)]
    if tim[1]=='<':  T[a]=int(tim[0]);
    elif tim[0]=='&':  T[a]=-1;
    else: T[a]=int(tim);
if T[0]==-1:                # sort out the half hours, two possible ways they could go
    for a in range(32):
        T[2*a]=mod(T[2*a+1]-0.5,24);
else:
    for a in range(32):
        T[2*a+1]=mod(T[2*a]+0.5,24);

# do the colours
n = string.find(html,'<tr bgcolor="#d1d1d1">');
temp = html[n:(n+7500)];
P=range(64);
C=range(64);
m=171;
for a in range(64):
    m = string.find(temp,'height=',m+5);
    hgt = temp[(m+8):(m+10)]
    col = temp[(m-28):(m-25)]
    P[a]=int(hgt);
    P[a]
    if col=='een': C[a]=0;
    if col=='nge': C[a]=1;
    if col=='red': C[a]=2;
    if col=='own': C[a]=3;
    #print a,m,hgt,col

# the 15th bar, P[14] C[14] T[14] etc, is the current system load

# find the rate of change in % points / hour
slope=1.2*(int(P[15])-int(P[14]))+0.6*(int(P[16])-int(P[14]))+0.2*(int(P[17])-int(P[14]));
print 'At', datetime.now().time(),'shedding status is',shed,' Colour is',C[14],', Load is',P[14],'%, slope is ',slope,'points/per hour'


# nice plot
collist='gyrk';
figure(figsize=[14,6]);
step(np.add(range(64),0.5),P)
step(range(64),np.multiply(C,5))
text(13.7,P[14]-3,shed,color=collist[C[a]],fontsize=14)
for a in range(64):
    text(a-0.4,P[a]+1,P[a],color=collist[C[a]],fontsize=8)
    text(a-0.4,C[a]*5+1,C[a],color=collist[C[a]],fontsize=8)
plot([13.5,13.5,14.5,14.5],[0,P[14],P[14],0]);
xlim([0,63]); ylim([-1,70]);
xticks(range(63), T, rotation='vertical')
title("At %s shedding stage %i, load=%i, slope=%3.1f/h" %(datetime.now().time(),shed,P[14],slope));
```

# Testing Interactivity

Interactivity


```python
from IPython.html.widgets import interact, interactive, fixed
from IPython.display import display
import skimage
from skimage import data, filter, io
i = data.coffee()
io.Image(i)
def edit_image(image, sigma=0.1, r=1.0, g=1.0, b=1.0):
    new_image = filter.gaussian_filter(image, sigma=sigma, multichannel=True)
    new_image[:,:,0] = r*new_image[:,:,0]
    new_image[:,:,1] = g*new_image[:,:,1]
    new_image[:,:,2] = b*new_image[:,:,2]
    new_image = io.Image(new_image)
    display(new_image)
    return new_image
lims = (0.0,1.0,0.01)
w = interactive(edit_image, image=fixed(i), sigma=(0.0,10.0,0.1), r=lims, g=lims, b=lims)
display(w)
w.result
```
